# Advancing Medical Artificial Intelligence Using a Century of Cases

# https://arxiv.org/abs/2509.12194

# üßæ Abstract

The study leverages **100 years of New England Journal of Medicine Clinicopathological Conferences (CPCs)** to systematically evaluate and advance medical AI. While earlier AI benchmarks focused only on final diagnoses, this work develops **CPC-Bench**, a physician-validated benchmark spanning 10 text-based and multimodal tasks, and introduces **Dr. CaBot**, an AI discussant capable of generating expert-style written and video presentations. Findings show that frontier LLMs outperform physicians on **differential diagnosis** and **testing plan** tasks but lag in **image interpretation** and **literature retrieval**.

---

## ‚ö†Ô∏è Problems Addressed
- **Narrow benchmarks:** Previous AI evaluations focused mainly on final diagnostic accuracy, overlooking reasoning processes and presentation skills.  
- **Educational gap:** CPCs are curated for teaching but often misused as real-life diagnostic benchmarks.  
- **Limited multimodality:** AI struggles to integrate clinical text, images, and literature in a human-like diagnostic workflow.  

---

## üéØ Purpose
- Create a robust, transparent benchmark (**CPC-Bench**) for evaluating clinical reasoning across multiple modalities.  
- Develop an AI system (**Dr. CaBot**) that emulates the expert physician discussant, providing text- and video-based differentials.  
- Track how medical AI performance evolves across both contemporary and historical cases.  

---

## üî¨ Methodology
- **Datasets:**  
  - 7,102 CPCs (1923‚Äì2025).  
  - 1,021 NEJM Image Challenges (2006‚Äì2025).  

- **Annotations:**  
  - Ten physicians annotated diagnostic *touchpoints* to validate case segmentation and reasoning steps.  

- **CPC-Bench tasks (10 total):**  
  1. Differential diagnosis  
  2. Testing plan  
  3. Literature search  
  4. Diagnostic touchpoints  
  5. QA  
  6. Reasoning evidence  
  7. Omission of findings  
  8. Visual QA  
  9. Visual differentials  
  10. Image challenge  

- **Models tested:**  
  - Proprietary: OpenAI o3, GPT-4o, Google Gemini 2.5 Pro, Anthropic Claude 4 Sonnet.  
  - Open-source: Llama-3.1-405B.  

- **AI Discussant (Dr. CaBot):**  
  - Built using OpenAI‚Äôs o3.  
  - Enhanced with literature retrieval and CPC-style emulation.  

---

## üìä Results
- **Differential diagnosis:**  
  - *o3*: 60% top-1, 84% top-10 accuracy.  
  - Physicians: 24% top-1, 45% top-10.  

- **Testing plan:**  
  - *o3*: 98% accuracy, surpassing physicians and all baselines.  

- **Literature retrieval:**  
  - *Gemini 2.5 Pro*: 49% top-10 citation accuracy.  
  - *o3*: 32%.  

- **Image tasks:**  
  - Accuracy plateaued at ~67%.  
  - Stronger in dermatology, weaker in radiology.  

- **Dr. CaBot performance:**  
  - Physicians misclassified AI vs. human discussants in 74% of trials.  
  - Rated higher for diagnostic justification, citation quality, and engagement.  

- **Historical benchmarking:**  
  - *o3* outperformed human discussants in 1940s‚Äì1950s cases.  
  - Matched or slightly exceeded experts in modern decades.  

---

## ‚úÖ Conclusions
- **Strengths of LLMs:** Superior in text-based diagnostic reasoning and testing plans.  
- **Weaknesses:** Gaps remain in multimodal integration, literature search, and image interpretation.  
- **Educational impact:** Dr. CaBot shows that AI can emulate expert presentations and act as a scalable teaching tool.  
- **Scientific contribution:** CPC-Bench and CaBot establish transparent, physician-validated benchmarks and resources for future medical AI research.  


# üìê Mathematical & Statistical Equations in the Paper

---

### 1. Sequence Probability (General LLM Formulation)

For a clinical case described as a sequence of tokens  
$ x = (x_1, x_2, \dots, x_m) $,  

the model predicts tokens  
$ y = (y_1, \dots, y_n) $ as:

$$
P(y \mid x; \theta) = \prod_{i=1}^n P(y_i \mid y_{<i}, x; \theta)
$$

This is the standard **autoregressive factorization** for large language models.

---

### 2. Accuracy Metrics (Differential Diagnosis & Test Plan)

For top-$k$ evaluation:

$$
\text{Top-}k\ \text{Accuracy} = \frac{1}{N} \sum_{i=1}^N \mathbf{1}(y_i \in \hat{Y}_i^{(k)})
$$

where:  
- $ y_i $ = ground-truth diagnosis/test,  
- $ \hat{Y}_i^{(k)} $ = model‚Äôs top-$k$ predictions,  
- $\mathbf{1}(\cdot)$ = indicator function.

---

### 3. Citation Retrieval Metric

For literature search evaluation:

$$
\text{Citation Recall@}k =
\frac{\#\{\text{ground-truth citations in top-}k\}}
{\text{Total ground-truth citations}}
$$

---

### 4. Image Task Accuracy

For NEJM Image Challenges:

$$
\text{Image Accuracy} =
\frac{\#\{\text{correctly predicted cases}\}}
{\text{Total cases}}
$$

Performance is stratified by specialty (e.g., dermatology, radiology).

---

### 5. Agreement Measurement (AI vs Human Discussant)

Physicians rated whether a discussant was AI or human. Misclassification rate:

$$
\text{Misclass Rate} =
\frac{\#\{\text{physician evaluations incorrectly identifying AI vs human}\}}
{\text{Total evaluations}}
$$

---

### 6. Historical Benchmarking

Performance gap between AI and human discussants across decades:

$$
\Delta \text{Performance}_{\text{decade}} =
\text{Accuracy}_{AI} - \text{Accuracy}_{Human}
$$

---

### 7. Statistical Testing (Bootstrap Confidence Intervals)

For metric $\theta$ estimated via bootstrapping:

$$
\hat{\theta}^* \sim \frac{1}{B} \sum_{b=1}^B \theta_b^*
$$

where $B$ = number of bootstrap resamples.

---

## ‚úÖ Summary

Unlike CNN/RNN architecture papers, this work does not introduce novel derivations.  
Instead, its math focuses on:

- Autoregressive sequence modeling.  
- **Top-$k$ accuracy** for diagnostic reasoning tasks.  
- **Recall@k** for literature retrieval.  
- **Accuracy** for image-based tasks.  
- **Misclassification rate** for AI vs human evaluations.  
- **Decade-level performance deltas** for historical benchmarking.  
- **Bootstrap confidence intervals** for statistical robustness.

```python
# ===============================================================
# Replication Lab: CPC-Bench Inspired Transformer
# Based on "Advancing Medical AI Using a Century of Cases" (2025)
# ===============================================================

import torch, torch.nn as nn, torch.optim as optim
import numpy as np, matplotlib.pyplot as plt, seaborn as sns
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction

device = "cuda" if torch.cuda.is_available() else "cpu"

# -----------------------------
# 1. Toy Medical Dataset (CPC-style)
# -----------------------------
cases = [
    ("fever cough fatigue", "influenza"),
    ("chest pain shortness breath", "myocardial_infarction"),
    ("polyuria polydipsia weight_loss", "diabetes_mellitus"),
    ("joint_pain morning_stiffness", "rheumatoid_arthritis"),
    ("abdominal_pain jaundice", "hepatitis"),
]

# Vocab
src_vocab = {"<pad>":0,"<sos>":1,"<eos>":2}
tgt_vocab = {"<pad>":0,"<sos>":1,"<eos>":2}
for s,d in cases:
    for w in s.split():
        if w not in src_vocab: src_vocab[w]=len(src_vocab)
    if d not in tgt_vocab: tgt_vocab[d]=len(tgt_vocab)
src_ivocab = {i:w for w,i in src_vocab.items()}
tgt_ivocab = {i:w for w,i in tgt_vocab.items()}

def encode(seq,vocab,max_len=6,add_eos=True):
    ids=[vocab.get(w,0) for w in seq.split()]
    if add_eos: ids.append(vocab["<eos>"])
    return ids+[0]*(max_len-len(ids))

X=torch.tensor([encode(s,src_vocab) for s,_ in cases])
Y=torch.tensor([encode(d,tgt_vocab,max_len=3) for _,d in cases])

# -----------------------------
# 2. Transformer Model
# -----------------------------
class TransformerNMT(nn.Module):
    def __init__(self,src_vocab,tgt_vocab,dim=64,heads=4,layers=2,max_len=10):
        super().__init__()
        self.src_emb=nn.Embedding(len(src_vocab),dim)
        self.tgt_emb=nn.Embedding(len(tgt_vocab),dim)
        self.pos_emb=nn.Embedding(max_len,dim)
        enc_layer=nn.TransformerEncoderLayer(d_model=dim,nhead=heads)
        dec_layer=nn.TransformerDecoderLayer(d_model=dim,nhead=heads)
        self.encoder=nn.TransformerEncoder(enc_layer,layers)
        self.decoder=nn.TransformerDecoder(dec_layer,layers)
        self.fc=nn.Linear(dim,len(tgt_vocab))
        self.dim=dim; self.max_len=max_len
    def forward(self,src,trg):
        B,Ls=src.size(); _,Lt=trg.size()
        pos_src=torch.arange(0,Ls,device=src.device).unsqueeze(0).repeat(B,1)
        pos_tgt=torch.arange(0,Lt,device=trg.device).unsqueeze(0).repeat(B,1)
        src_emb=self.src_emb(src)+self.pos_emb(pos_src)
        tgt_emb=self.tgt_emb(trg)+self.pos_emb(pos_tgt)
        memory=self.encoder(src_emb.transpose(0,1))
        out=self.decoder(tgt_emb.transpose(0,1),memory)
        return self.fc(out.transpose(0,1))

# -----------------------------
# 3. Training
# -----------------------------
model=TransformerNMT(src_vocab,tgt_vocab).to(device)
crit=nn.CrossEntropyLoss(ignore_index=0)
opt=optim.Adam(model.parameters(),1e-3)

X,Y=X.to(device),Y.to(device)
losses=[]
for ep in range(50):
    model.train(); opt.zero_grad()
    out=model(X,Y[:,:-1])
    loss=crit(out.reshape(-1,out.size(-1)),Y[:,1:].reshape(-1))
    loss.backward(); opt.step()
    losses.append(loss.item())
    if (ep+1)%10==0: print(f"Epoch {ep+1}, Loss={loss.item():.4f}")

# -----------------------------
# 4. Inference
# -----------------------------
def translate(model,sent,max_len=5):
    model.eval()
    src=torch.tensor([encode(sent,src_vocab)]).to(device)
    memory=model.encoder((model.src_emb(src)+model.pos_emb(
        torch.arange(0,src.size(1),device=device).unsqueeze(0))).transpose(0,1))
    y=torch.tensor([[tgt_vocab["<sos>"]]]).to(device)
    result=[]
    attn=[]
    for _ in range(max_len):
        pos=torch.arange(0,y.size(1),device=device).unsqueeze(0)
        tgt_emb=model.tgt_emb(y)+model.pos_emb(pos)
        out=model.decoder(tgt_emb.transpose(0,1),memory)
        next_tok=model.fc(out.transpose(0,1))[:,-1].argmax(1).item()
        if next_tok==tgt_vocab["<eos>"]: break
        result.append(tgt_ivocab[next_tok]);
        y=torch.cat([y,torch.tensor([[next_tok]]).to(device)],dim=1)
    return result

print("Prediction:",translate(model,"fever cough fatigue"))
```

```python
# -----------------------------
# 6. Visualization
# -----------------------------
plt.plot(losses); plt.xlabel("Epoch"); plt.ylabel("Loss")
plt.title("Transformer Training Convergence"); plt.show()
```
