## 1. Local Linearization of Nonlinear Systems

### Problem  
Nonlinear systems are hard to analyze, predict, or control directly.

### Jacobian Solution  
The Jacobian provides a **first-order linear approximation** around an operating point, enabling local analysis.

This allows:
- Stability analysis  
- Sensitivity analysis  
- Local prediction and control  

### Real-World Cases  
Robotics dynamics  
Aircraft control systems  
Power grids  
Chemical reactors

## 6. Invertibility and Physical Plausibility Constraints

### Problem  
Transformations may fold, collapse, or become non-physical.

### Jacobian Determinant Solution  

$$
\det J > 0 \quad \text{invertible, orientation-preserving}
$$

$$
\det J = 0 \quad \text{collapse}
$$

$$
\det J < 0 \quad \text{folding}
$$
