# Generated Faces in the Wild: Quantitative Comparison of Stable Diffusion, Midjourney and DALL-E 2

# https://arxiv.org/pdf/2210.00586


# Abstract
This study provides a quantitative evaluation of three major text-to-image models—**Stable Diffusion**, **Midjourney**, and **DALL-E 2**—focusing on their ability to generate photorealistic human faces in complex, natural scenes (“in the wild”). Using the **Fréchet Inception Distance (FID)** as the primary metric, the author introduces the **Generated Faces in the Wild (GFW)** dataset of 15,076 faces and demonstrates that Stable Diffusion achieves superior facial realism compared to the other systems.

# Problems
Despite the rapid progress in generative AI, most evaluations focus on overall image quality rather than specific object categories such as **faces** within cluttered or naturalistic scenes. Existing analyses are often anecdotal or limited to portrait-specific models (e.g., StyleGAN). A **fine-grained quantitative comparison** of diffusion-based models’ facial synthesis abilities remains lacking.

# Proposed Solutions
The study addresses this gap by:
1. Introducing a new dataset (**GFW**) of generated and real faces extracted from broader text-to-image generations.  
2. Conducting **FID-based quantitative analysis** of generated faces.  
3. Systematically comparing three major systems under standardized conditions using **COCO captions** as prompts and **MediaPipe** for face detection.

# Purpose
The primary goal is to measure how well modern diffusion-based systems synthesize photorealistic faces in unconstrained, complex scenes and to establish a **benchmark** that can inform future model development and fairness/bias assessments.

# Methodology
- **Models Evaluated:** Stable Diffusion, Midjourney, and DALL-E 2.  
- **Data Generation:** 15,076 generated faces (8,050 SD; 6,350 MJ; 676 DALL-E 2) using COCO captions containing human-related words.  
- **Real Faces:** 30,000 samples from COCO and LFW datasets.  
- **Preprocessing:** Faces filtered for quality and resized to \(100 \times 100\) pixels.  
- **Evaluation Metric:** Fréchet Inception Distance (FID) calculated 10 times over random 5K samples, with means and standard deviations reported.  

The **Fréchet Inception Distance** is given by:

$$
FID = ||\mu_r - \mu_g||_2^2 + \text{Tr}(\Sigma_r + \Sigma_g - 2(\Sigma_r \Sigma_g)^{1/2})
$$

where:
- \(\mu_r, \Sigma_r\) = mean and covariance of real image features  
- \(\mu_g, \Sigma_g\) = mean and covariance of generated image features

# Results
- **Stable Diffusion** achieved the lowest (best) FID scores, outperforming both Midjourney and DALL-E 2.  
- **DALL-E 2** underperformed possibly due to:
  1. Deepfake safeguards  
  2. Focus on single-object compositions  
  3. Limited image samples  
- **Midjourney’s** relatively poor FID likely stems from its **artistic bias** rather than photorealism.  
- Even the best model’s faces remain **distinguishable from real ones**, indicating a realism gap.  
- Common synthesis failures include **eyeglass rendering**, **facial asymmetry**, and **occlusion artifacts**.

# Conclusions
This is the **first quantitative study** comparing AI-generated faces “in the wild.” Stable Diffusion leads in realism but still falls short of human-level fidelity. The study highlights the need for:
- Larger evaluation datasets, particularly for DALL-E 2.  
- In-depth **memorization and bias analyses** of generated faces.  
- Exploration of alternative metrics (e.g., **SSIM**, **LPIPS**) and **face-specific embeddings**.  
- Future work on **expressions, demographics, and object categories** (e.g., animals, vehicles).  

The **GFW dataset** serves as a foundation for deeper investigation into **fairness**, **authenticity**, and **generalization** in generative models.


# Mathematical and Statistical Content Summary  
*(From “Generated Faces in the Wild: Quantitative Comparison of Stable Diffusion, Midjourney, and DALL-E 2” by Ali Borji, 2023)*

---

## 1. Fréchet Inception Distance (FID)

**Mathematical Concept:**  
The **FID** is a statistical measure of similarity between two sets of images, comparing their **deep feature distributions** (extracted via an Inception-v3 network). It computes the **Fréchet distance** between two multivariate Gaussian distributions:

$$
FID = ||\mu_r - \mu_g||_2^2 + \mathrm{Tr}(\Sigma_r + \Sigma_g - 2(\Sigma_r \Sigma_g)^{1/2})
$$

where:  
- \(\mu_r, \Sigma_r\): Mean and covariance of real image features.  
- \(\mu_g, \Sigma_g\): Mean and covariance of generated image features.  
- \(||\cdot||_2\): Euclidean norm.  
- \(\mathrm{Tr}\): Trace of a matrix (sum of diagonal elements).  

**Purpose:**  
FID quantifies **visual realism** — lower scores indicate that generated images are statistically closer to real ones. It serves as the core quantitative metric comparing **Stable Diffusion**, **Midjourney**, and **DALL-E 2**.

---

## 2. Random Sampling and Repetition

**Statistical Method:**  
To ensure unbiased results:
- Randomly select **5,000 faces** from each model and **5,000 real faces**.
- Compute FID **10 times**.
- Record the **mean** and **standard deviation** across runs.

**Purpose:**  
Repeated sampling reduces variance and ensures the comparison reflects **true model performance**, not random sample noise.

---

## 3. Mean and Standard Deviation

**Mathematical Definitions:**

Mean:
$$
\mu = \frac{1}{n} \sum_{i=1}^{n} x_i
$$

Standard Deviation:
$$
\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(x_i - \mu)^2}
$$

**Purpose:**  
The mean (\(\mu\)) captures the **average FID**, while the standard deviation (\(\sigma\)) measures **consistency**. Lower σ indicates more stable results across repetitions.

---

## 4. Dataset Balancing and Normalization

**Statistical Adjustment:**  
Because DALL-E 2 produced fewer samples (**676 faces**), FID was recomputed for all models using **equal sample sizes**.

**Purpose:**  
This normalization ensures **fair comparison** by removing bias introduced by unequal dataset sizes.

---

## 5. Gaussian Distributions (Implicit in FID)

**Mathematical Idea:**  
FID assumes feature embeddings follow **multivariate Gaussian distributions**:

$$
x \sim \mathcal{N}(\mu, \Sigma)
$$

**Purpose:**  
By modeling real and generated features as Gaussians, FID provides a **closed-form distance** measure, avoiding the need for pixel-wise comparison.

---

## 6. Trace Operator (\(\mathrm{Tr}\))

**Definition:**  
The trace of a square matrix \(A\) is the sum of its diagonal elements:

$$
\mathrm{Tr}(A) = \sum_{i} A_{ii}
$$

**Purpose:**  
In FID, the trace quantifies **overall variance difference** between the real and generated feature spaces.

---

## 7. Matrix Square Root \((\Sigma_r \Sigma_g)^{1/2}\)

**Mathematical Operation:**  
The matrix square root computes:

$$
(\Sigma_r \Sigma_g)^{1/2} = U \Lambda^{1/2} U^\top
$$

where \(U\) and \(\Lambda\) are the eigenvectors and eigenvalues of the covariance product.

**Purpose:**  
This operation allows **blending covariance structures** of real and generated features, capturing correlations beyond simple mean differences.

---

## 8. Alternative Quality Metrics (Discussed but Not Used)

- **SSIM (Structural Similarity Index):**  
  Evaluates perceptual similarity using luminance, contrast, and structure.  

- **LPIPS (Learned Perceptual Image Patch Similarity):**  
  Measures perceptual similarity using deep feature activations, aligning more closely with **human visual judgment**.

**Purpose:**  
These metrics provide **complementary statistical perspectives** and may enhance future facial realism evaluations.

---

All mathematical and statistical components contribute to a **rigorous, quantitative assessment** of AI-generated facial realism:

| Mathematical Tool | Function in Study |

| Dataset Balancing | Eliminates bias from unequal sample counts |

````python
# ============================================================
# GENERATED FACES IN THE WILD (EDUCATIONAL REPLICATION)
# ============================================================
# Author: Mohammed Al Abrah (Educational adaptation)
# Concept: Based on Ali Borji (2023) – "Generated Faces in the Wild"
# Goal: Simplify and replicate the *methodology* of the paper,
#       not the large-scale models (Stable Diffusion, DALL·E 2, etc.)
#       but the *evaluation pipeline*: dataset creation, training a
#       small generator, computing FID-like distance, and visualizing results.
# ============================================================

import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torchvision.utils import make_grid
import matplotlib.pyplot as plt
import numpy as np
from scipy import linalg
````

````python
# ============================================================
# 1. DATA PREPARATION (MNIST Digits as Proxy for "Faces")
# ============================================================
# We use MNIST here for teaching — similar structure: grayscale 28x28 images.
# Think of digits as "synthetic faces" for testing our generative and evaluation flow.

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))  # scale pixels to [-1, 1]
])

trainset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=128, shuffle=True)
````

````python
# ============================================================
# 2. SIMPLE GENERATOR MODEL (Tiny GAN Generator)
# ============================================================
# A small fully-connected generator network that learns to map
# random noise → fake images (28×28). This imitates the idea of generating images "in the wild".

class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(100, 256),
            nn.ReLU(),
            nn.Linear(256, 512),
            nn.ReLU(),
            nn.Linear(512, 784),
            nn.Tanh()  # output in [-1, 1]
        )

    def forward(self, z):
        img = self.model(z).view(-1, 1, 28, 28)
        return img
````

````python
# ============================================================
# 3. SIMPLE DISCRIMINATOR (Classifier of Real vs Fake)
# ============================================================
# This small CNN acts like the “evaluator” in the GAN, learning
# to distinguish real faces (MNIST digits) from generated ones.

class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Flatten(),
            nn.Linear(784, 512),
            nn.LeakyReLU(0.2),
            nn.Linear(512, 256),
            nn.LeakyReLU(0.2),
            nn.Linear(256, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.model(x)
````

````python
# ============================================================
# 4. TRAINING SETUP
# ============================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

G = Generator().to(device)
D = Discriminator().to(device)

criterion = nn.BCELoss()
opt_G = optim.Adam(G.parameters(), lr=0.0002)
opt_D = optim.Adam(D.parameters(), lr=0.0002)

# Helper arrays for plotting
loss_G, loss_D = [], []
````

````python
# ============================================================
# 5. TRAINING LOOP (GAN Optimization)
# ============================================================
epochs = 20
for epoch in range(epochs):
    for imgs, _ in trainloader:
        real_imgs = imgs.to(device)
        batch_size = real_imgs.size(0)
        real_labels = torch.ones((batch_size, 1), device=device)
        fake_labels = torch.zeros((batch_size, 1), device=device)

        # ---- Train Discriminator ----
        z = torch.randn(batch_size, 100, device=device)
        fake_imgs = G(z)
        D_real = D(real_imgs)
        D_fake = D(fake_imgs.detach())
        loss_D_real = criterion(D_real, real_labels)
        loss_D_fake = criterion(D_fake, fake_labels)
        loss_D_total = (loss_D_real + loss_D_fake) / 2
        opt_D.zero_grad()
        loss_D_total.backward()
        opt_D.step()

        # ---- Train Generator ----
        z = torch.randn(batch_size, 100, device=device)
        fake_imgs = G(z)
        D_fake = D(fake_imgs)
        loss_G_total = criterion(D_fake, real_labels)  # want fake to appear real
        opt_G.zero_grad()
        loss_G_total.backward()
        opt_G.step()

    # Record losses
    loss_G.append(loss_G_total.item())
    loss_D.append(loss_D_total.item())

    print(f"Epoch [{epoch+1}/{epochs}] | Loss_D: {loss_D_total.item():.3f} | Loss_G: {loss_G_total.item():.3f}")
````

````python
# ============================================================
# 6. VISUALIZATION: LOSS CURVES
# ============================================================
plt.figure(figsize=(8,4))
plt.plot(loss_D, label='Discriminator Loss')
plt.plot(loss_G, label='Generator Loss')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training Curves: GAN on MNIST (Faces Proxy)")
plt.legend()
plt.show()
````

````python
# ============================================================
# 7. VISUALIZATION: GENERATED "FACES"
# ============================================================
G.eval()
with torch.no_grad():
    z = torch.randn(64, 100, device=device)
    gen_imgs = G(z).cpu()
grid = make_grid(gen_imgs, nrow=8, normalize=True)
plt.figure(figsize=(8,8))
plt.imshow(np.transpose(grid, (1,2,0)))
plt.title("Generated Faces (Proxy: MNIST Digits)")
plt.axis("off")
plt.show()
````

````python
# ============================================================
# 8. EVALUATION: FID-LIKE METRIC (Simplified)
# ============================================================
# Here we use the classic Fréchet distance idea between real and fake features.
# For simplicity, we use the images directly as features (not Inception-V3 embeddings).

def compute_fid(real, fake):
    # Flatten images to vectors
    real = real.view(real.size(0), -1).cpu().numpy()
    fake = fake.view(fake.size(0), -1).cpu().numpy()

    mu_r, mu_g = real.mean(axis=0), fake.mean(axis=0)
    cov_r, cov_g = np.cov(real, rowvar=False), np.cov(fake, rowvar=False)

    diff = mu_r - mu_g
    covmean = linalg.sqrtm(cov_r.dot(cov_g))
    if np.iscomplexobj(covmean):
        covmean = covmean.real
    fid = diff.dot(diff) + np.trace(cov_r + cov_g - 2 * covmean)
    return fid

# Sample real and fake sets
real_imgs, _ = next(iter(trainloader))
real_sample = real_imgs[:1000].to(device)
z = torch.randn(1000, 100, device=device)
fake_sample = G(z).detach()

fid_score = compute_fid(real_sample, fake_sample)
print(f"Approximate FID-like score: {fid_score:.2f} (lower = more realistic)")
````

````python
# ============================================================
# 9. VISUAL SUMMARY
# ============================================================
plt.figure(figsize=(8,4))
plt.bar(['FID Score'], [fid_score], color='steelblue')
plt.title("Quantitative Evaluation (Lower = Better)")
plt.ylabel("FID-like Distance")
plt.show()
````

# Quantitative and Qualitative Analysis of GAN Training Behavior

## 1. Training Dynamics (Loss Behavior)

**Observation:**  
From the training logs and loss curves (Screenshots 3234–3235):
- **Discriminator loss (Loss_D):** stays low in the range \(0.02 \leq L_D \leq 0.5\).  
- **Generator loss (Loss_G):** fluctuates significantly between \(3\) and \(8\), showing unstable convergence.

**Academic Interpretation:**  
This pattern reflects the **adversarial instability** inherent in GAN optimization.  
A low discriminator loss indicates that \(D\) quickly learns to distinguish real from generated samples, while the generator’s oscillating loss reveals its difficulty in deceiving the discriminator.

Mathematically, this reflects the **min–max game**:

$$
\min_G \max_D V(D, G) =
\mathbb{E}_{x \sim p_{\text{data}}} [\log D(x)] +
\mathbb{E}_{z \sim p_z} [\log(1 - D(G(z)))]
$$

Here:
- \(D\) seeks to **maximize** discrimination accuracy.  
- \(G\) seeks to **minimize** this same objective by generating realistic samples.

Such oscillations are characteristic of non-convex **min–max dynamics**, often mitigated by stabilization techniques like gradient penalties, learning-rate balancing, or exponential moving averages (EMA).

**Observation (Screenshots 3236–3237):**  
Generated images resemble **fuzzy, incomplete, or distorted MNIST digits**—with visible structure but poor detail fidelity.

**Academic Interpretation:**  
The generator captures **low-frequency global patterns** (digit outlines) but fails to model **high-frequency details** (edges, curvature).  
This limitation stems from:
- **Under-parameterization** (a shallow MLP rather than convolutional or diffusion-based generator).

Conceptually, these artifacts mirror the **asymmetric and distorted synthetic faces** discussed in Borji’s “Generated Faces in the Wild.”  
In both cases, the generative model struggles with **fine-grained spatial coherence** and **realistic symmetry**, particularly under adversarial competition.

- Both studies quantify **statistical divergence** between generated and real image manifolds.  
- Both employ **FID** as an objective realism metric grounded in Gaussian statistics.  
- Both reveal a **persistent realism gap**, highlighting the challenges of bridging perceptual and statistical fidelity.

**Conclusion:**  
Through this controlled GAN experiment, the core logic of *Generated Faces in the Wild* is successfully emulated — translating subjective visual quality into **mathematically measurable** evidence of generative model performance.

### **Summary**
The related work connects *Generated Faces in the Wild* to three intersecting research domains:

1. **Foundational Generative Paradigms:**  
   Establishes the architectural lineage from GANs and VAEs to modern diffusion models, situating Stable Diffusion and DALL·E 2 within this continuum.

2. **Evaluation and Bias Studies:**  
   Highlights the evolution of realism metrics (FID, SSIM, LPIPS) and raises concerns regarding bias, generalization, and fairness in generative outputs.

3. **Interpretability and Visualization Research:**  
   Draws on visualization studies to conceptually understand how generative models encode, reconstruct, and sometimes distort human facial structures.

Together, these threads support the paper’s contribution: moving from broad qualitative assessments to **fine-grained, quantitative, and reproducible evaluation** of facial realism in modern text-to-image systems.
