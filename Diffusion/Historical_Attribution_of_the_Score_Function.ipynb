# Historical Attribution of the Score Function

The equation

$$
s(x) = \nabla_x \log p(x)
$$

was **not** introduced by Charles Stein. Its origin predates Stein’s work by several decades and lies firmly in **classical likelihood theory in statistics**.

What follows is a precise and historically accurate attribution.

---

## Who First Introduced the Score Function?

### Primary Origin: Ronald A. Fisher

The score function originates in the work of **Ronald A. Fisher** during the period **1915–1925**, as part of his foundational development of **maximum likelihood estimation** and **Fisher information**.

---

## Fisher’s Original Contribution

Fisher introduced the *score* as the derivative of the log-likelihood with respect to a **parameter** $$\theta$$:

$$
U(\theta) = \frac{\partial}{\partial \theta} \log p(x \mid \theta).
$$

This quantity was named the *score* because:

- It measures the sensitivity of the likelihood to changes in parameters.
- It vanishes at maximum likelihood estimates.
- Its variance defines the Fisher information.

This object is the **direct ancestor** of the modern score function.

---

## How the Modern Form Emerged

Subsequent developments generalized Fisher’s original concept in two essential directions.

### 1. From Parameters to Variables

Instead of differentiating with respect to parameters $$\theta$$, later work considered derivatives with respect to the **random variable** $$x$$ itself:

$$
s(x) = \nabla_x \log p(x).
$$

This transformation moves the score from parameter space to sample space.

---

### 2. From Estimation to Geometry and Dynamics

The score came to be understood as:

- A vector field over the sample space,
- A local descriptor of probability geometry,
- A driving quantity in stochastic dynamics.

This conceptual shift emerged gradually through work in:

- Statistical physics,
- Information geometry,
- Stochastic process theory.

---

## Historical Timeline

| $$\text{Period}$$ | $$\text{Contribution}$$ |
|------------------|-------------------------|
| $$1915\text{–}1925$$ | $$\text{Fisher defines the score in likelihood theory}$$ |
| $$1940\text{–}1960\text{s}$$ | $$\text{Score used in asymptotic statistics and efficiency theory}$$ |
| $$1972$$ | $$\text{Stein embeds score-like derivatives in distribution-characterizing identities}$$ |
| $$2005$$ | $$\text{Hyvärinen formalizes score matching}$$ |
| $$2019\text{–}2021$$ | $$\text{Score becomes central in diffusion and score-based generative models}$$ |

---

## What Stein Did Differently (Key Distinction)

Charles Stein **did not invent** the score function.

His contribution was fundamentally different:

- He used score-related derivatives inside **distribution-defining operators**.
- He transformed the score into a tool for **distributional comparison**.
- He made it central to **normal approximation and error bounds**.

This is why the term *Stein score function* exists.

It refers to:

- The **role** of the score within Stein’s method,

and **not** to its historical invention.

---

## Definitive Statement

The equation

$$
s(x) = \nabla_x \log p(x)
$$

originates from **Ronald Fisher’s likelihood theory**; **Charles Stein** later repurposed this quantity within **distribution-characterizing identities**, giving rise to the retrospective term *Stein score function*.


# Primary Papers by Ronald A. Fisher Related to the Score Function

This section provides a precise and historically correct account of **Ronald A. Fisher’s primary papers** that introduced and formalized the **score function**, together with an explanation of how the modern form emerged.

---

## 1. The Original Introduction (Most Important)

### Fisher, Ronald A. (1922)  
**Title:** *On the Mathematical Foundations of Theoretical Statistics*  
**Venue:** *Philosophical Transactions of the Royal Society of London, Series A*

### Why This Paper Matters

This paper is the **authoritative and canonical source** for the score function. It establishes the mathematical foundations of likelihood theory and introduces:

- The likelihood and log-likelihood,
- Differentiation of the log-likelihood,
- Fisher information as a variance quantity,
- The score as a central inferential object.

### Definition Introduced by Fisher

In this paper, Fisher defines the score as

$$
U(\theta) = \frac{\partial}{\partial \theta} \log p(x \mid \theta).
$$

This derivative is explicitly named the **score** by Fisher.

### Conceptual Role

- Measures sensitivity of the likelihood to changes in $$\theta$$,
- Vanishes at maximum likelihood estimates,
- Has variance equal to the Fisher information.

This paper is universally cited as the **mathematical foundation of likelihood theory**.

---

## 2. Earlier Conceptual Development

### Fisher, Ronald A. (1912)  
**Title:** *On an Absolute Criterion for Fitting Frequency Curves*  
**Venue:** *Messenger of Mathematics*

### Contribution

- Introduces early likelihood-based reasoning,
- Uses optimization principles that prefigure likelihood differentiation,
- Lays conceptual groundwork for later formal definition of the score.

While the score is not yet explicitly formalized, this paper represents Fisher’s **first systematic move toward likelihood-based inference**.

---

## 3. Asymptotic and Inferential Role of the Score

### Fisher, Ronald A. (1925)  
**Title:** *Theory of Statistical Estimation*  
**Venue:** *Proceedings of the Cambridge Philosophical Society*

### Contribution

This paper establishes the **statistical properties** of the score:

- Zero expectation:
  $$
  \mathbb{E}[U(\theta)] = 0,
  $$

- Variance equal to Fisher information:
  $$
  \mathcal{I}(\theta) = \mathbb{E}\!\left[ U(\theta)^2 \right],
  $$

- Central role in:
  - Efficiency,
  - Consistency,
  - Asymptotic normality of estimators.

This work firmly places the score at the center of modern statistical inference.

---

## 4. What Fisher Actually Defined

Fisher’s original definition of the score is **parameter-based**:

$$
U(\theta) = \frac{\partial}{\partial \theta} \log p(x \mid \theta).
$$

This is the **original score function**.

---

## 5. How This Became the Modern Score  
### $$\nabla_x \log p(x)$$

Later developments generalized Fisher’s concept in a new direction:

- **Fisher:** differentiation with respect to parameters $$\theta$$,
- **Modern theory:** differentiation with respect to variables $$x$$.

The modern form is

$$
s(x) = \nabla_x \log p(x).
$$

### Interpretation Shift

- Same mathematical structure,
- Different object of differentiation,
- Different conceptual role.

This generalization emerged through:

- Information geometry,
- Statistical physics,
- Stochastic processes,
- Stein’s method,
- Score matching and diffusion models.

---

## 6. Definitive Academic Statement

The score function was introduced by **Ronald A. Fisher** in his 1922 paper *On the Mathematical Foundations of Theoretical Statistics*, where it appears as the derivative of the log-likelihood and forms the basis of Fisher information.

This is the **correct and definitive citation** when tracing the origin of the score function.
