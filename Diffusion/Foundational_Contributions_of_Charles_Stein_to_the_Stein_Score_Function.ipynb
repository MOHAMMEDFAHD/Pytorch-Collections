# Foundational Contributions of Charles Stein to the Stein Score Function

## 1. Primary Origin of Stein’s Method

**Stein, Charles (1972)**  
**Title:** *A bound for the error in the normal approximation to the distribution of a sum of dependent random variables*  
**Venue:** Proceedings of the Sixth Berkeley Symposium on Mathematical Statistics and Probability  

This seminal paper marks the formal introduction of **Stein’s method**, a novel probabilistic technique for assessing the accuracy of distributional approximations, particularly normal approximations, in settings involving dependence. Rather than relying on characteristic functions or moment matching, Stein proposed an operator-based approach in which a target distribution is uniquely characterized by a differential or difference equation.

A central conceptual innovation of this work is the idea that a probability distribution can be characterized through identities involving expectations of suitably chosen operators. These operators implicitly involve derivatives of the log-density of the target distribution, laying the theoretical groundwork for what would later be recognized as the **Stein score function**. Although the term itself does not appear in this paper, the mathematical structure introduced here constitutes the foundational origin of the concept.

---

## 2. Book-Length Formalization and Systematization

**Stein, Charles (1986)**  
**Title:** *Approximate Computation of Expectations*  
**Publisher:** Institute of Mathematical Statistics (IMS)  

In this monograph, Stein provides a comprehensive and systematic exposition of Stein’s method, significantly expanding its scope and applicability. The book formalizes Stein identities, Stein operators, and the general methodology for approximating expectations under complex probability distributions.

Within this framework, derivatives of log-probability densities appear explicitly and repeatedly as fundamental analytical objects. These quantities—now commonly referred to as **score functions**—play a central role in defining Stein operators and in establishing distributional characterizations. This work therefore solidifies the mathematical role of the score function within Stein’s method, even though modern terminology had not yet been standardized.

---

## 3. Important Historical Clarification

The **score function** is defined as
$$
s(x) = \nabla_x \log p(x)
$$

This quantity was well known in statistics prior to Stein’s work, particularly through classical likelihood theory and Fisher information. What distinguishes Stein’s contribution is **not** the invention of the score function itself, but rather its **conceptual repurposing**.

Stein’s key contributions were:

- Embedding the score function within *distribution-characterizing identities*, and  
- Elevating it to a central role in *distribution comparison, approximation, and convergence analysis*.

As a result, the modern term **“Stein score function”** should be understood as a *retrospective designation*, reflecting how the score function operates within the Stein framework, rather than the title or explicit terminology of any original paper by Stein.

---

## 4. Later Works Making the Stein Score Explicit

Subsequent research—primarily in modern statistics and machine learning—made the connection between Stein’s method and the score function explicit and computationally operational. Notable examples include:

- **Hyvärinen (2005)**  
  *Estimation of Non-Normalized Statistical Models by Score Matching*  
  Introduced score matching as a practical estimation technique for unnormalized models.

- **Liu, Lee, and Jordan (2016)**  
  *A Kernelized Stein Discrepancy*  
  Defined a computable discrepancy measure based on Stein operators and score functions.

- **Gorham and Mackey (2017)**  
  *Measuring Sample Quality with Stein’s Method*  
  Used Stein discrepancies to assess the quality of samples from complex distributions.

These works collectively formalized the Stein score as a **computational object**, bridging classical probability theory with modern machine learning and generative modeling.

---

## Definitive Statement

There is **no single paper** titled *“Stein Score Function”*.  
The concept originates implicitly from Charles Stein’s 1972 work on normal approximation, is developed systematically across his contributions to Stein’s method, and is made explicit and computational in later statistical and machine learning literature.
