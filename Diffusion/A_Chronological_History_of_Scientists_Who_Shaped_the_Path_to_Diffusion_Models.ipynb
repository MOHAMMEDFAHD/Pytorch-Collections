## Isaac Newton

## (1643–1727)

### Contribution

- Co-invented **calculus**, introducing derivatives and integrals as fundamental mathematical tools.  
- Established **deterministic differential equations** as the formal language of classical mechanics.  
- Provided the mathematical foundation later extended to **ordinary**, **partial**, and **stochastic differential equations**.

Classical Newtonian dynamics can be written as deterministic evolution equations:
$$
\frac{d x(t)}{d t} = v(t), \qquad
\frac{d v(t)}{d t} = \frac{F(x(t))}{m}.
$$

### Relevance to Diffusion Models

Diffusion models can be viewed as a **generalization of Newtonian dynamics**.  
While Newton’s framework describes perfectly deterministic motion, diffusion introduces **stochasticity** to account for uncertainty, noise, and microscopic fluctuations.

This transition replaces deterministic dynamics with stochastic dynamics:
$$
d x(t) = f(x(t), t)\, dt + g(t)\, dW_t,
$$
where \( dW_t \) denotes a Wiener (Brownian motion) increment.

In this sense, modern diffusion models inherit Newton’s mathematical language but extend it from **deterministic physical motion** to **probabilistic evolution**, forming a core foundation of contemporary generative modeling.


Gottfried Wilhelm Leibniz.jpg

## Gottfried Wilhelm Leibniz

## (1646–1716)

### Contribution

- Independent co-founder of **calculus**, alongside Isaac Newton.  
- Introduced the **notation and formalism** (such as \( d \), \( \int \)) that became standard for derivatives and integrals.  
- Provided a symbolic language that made **differential equations** systematic, expressive, and extensible.

For example, Leibniz’s notation expresses differentiation as:
$$
\frac{d y}{d x},
$$
and integration as:
$$
\int f(x)\, dx,
$$
which remain the dominant mathematical notation today.

### Relevance to Diffusion Models

Without calculus in Leibniz’s formal sense, **partial differential equations (PDEs)**, **stochastic differential equations (SDEs)**, and **diffusion equations** could not be formulated or analyzed.

Diffusion models rely fundamentally on equations of the form:
$$
\frac{\partial p(x,t)}{\partial t}
= \mathcal{L}\, p(x,t),
$$
where \( \mathcal{L} \) is a differential operator.

Leibniz’s notation and calculus framework are therefore a **prerequisite mathematical infrastructure** for all modern diffusion-based generative models.


Thomas Bayes.jpg

## Thomas Bayes

## (1701–1761)

### Contribution

- Introduced **Bayes’ theorem**, providing a formal rule for probabilistic inference under uncertainty.
- Established the mathematical relationship between prior beliefs, observed data, and posterior beliefs.

Bayes’ theorem is given by:
$$
p(x \mid y) = \frac{p(y \mid x)\, p(x)}{p(y)}.
$$

This equation formalizes how knowledge is updated when new evidence is observed.

### Role in Diffusion Models

Bayes’ rule forms the **inferential foundation** of diffusion models, especially for **conditional generation** and **inverse problems**.

In score-based diffusion models:

- Bayes’ rule is applied **in score space**, rather than directly in probability space.
- Posterior scores are obtained by combining gradients:
$$
\nabla_x \log p(x \mid y)
=
\nabla_x \log p(x)
+
\nabla_x \log p(y \mid x).
$$
- The prior score \( \nabla_x \log p(x) \) is provided by the trained diffusion model.
- The likelihood score \( \nabla_x \log p(y \mid x) \) encodes the observation or measurement process.

This framework enables powerful conditional tasks **without retraining**, including:
- Image inpainting  
- Super-resolution  
- MRI and CT reconstruction  
- General inverse problems  

Bayes’ theorem therefore provides the **probabilistic backbone** that allows diffusion models to move from unconditional generation to flexible, data-consistent inference.


Leonhard Euler.jpg

## Leonhard Euler

## (1707–1783)

### Contribution

- Developed the theory of **ordinary differential equations (ODEs)**, establishing a systematic framework for describing continuous-time dynamical systems.  
- Introduced **numerical integration schemes**, most notably the Euler method, enabling approximate solutions to differential equations when closed-form solutions are unavailable.

The classical Euler method for an ODE
$$
\frac{d x(t)}{d t} = f(x(t), t)
$$
is given by the discrete update:
$$
x_{t+\Delta t} = x_t + f(x_t, t)\,\Delta t.
$$

### Role in Diffusion Models

Euler’s numerical schemes form the **deterministic computational backbone** of modern diffusion models.

In diffusion modeling:

- Deterministic Euler integration is extended to **stochastic integration**, leading to methods such as **Euler–Maruyama**:
$$
x_{t+\Delta t} = x_t + f(x_t, t)\,\Delta t + g(t)\,\sqrt{\Delta t}\,\varepsilon,
$$
where \( \varepsilon \sim \mathcal{N}(0, I) \).
- These stochastic solvers are used to simulate both **forward diffusion** (noise injection) and **reverse-time diffusion** (generation).

Thus, Euler’s work provides the **algorithmic foundation** upon which stochastic solvers for diffusion and score-based generative models are built.


Carl Friedrich Gauss.jpg

## Carl Friedrich Gauss

## (1777–1855)

### Contribution

- Introduced the **Gaussian (normal) distribution**, providing a principled mathematical model for natural variability and random error.  
- Formalized **error theory** and probabilistic modeling, establishing the normal distribution as the central object in statistics and measurement theory.  
- Laid the groundwork for modern probability theory, statistical inference, and stochastic processes.

The Gaussian (normal) distribution for a scalar variable \( x \) is defined as:
$$
p(x) = \frac{1}{\sqrt{2\pi\sigma^2}}
\exp\!\left(-\frac{(x-\mu)^2}{2\sigma^2}\right),
$$
where \( \mu \) is the mean and \( \sigma^2 \) is the variance.

For the multivariate case \( x \in \mathbb{R}^d \):
$$
p(x) = \frac{1}{(2\pi)^{d/2} |\Sigma|^{1/2}}
\exp\!\left(
-\frac{1}{2}(x-\mu)^\top \Sigma^{-1} (x-\mu)
\right),
$$
where \( \mu \in \mathbb{R}^d \) and \( \Sigma \) is the covariance matrix.

### Relevance to Diffusion Models

Gaussian noise is the **fundamental noise process** used in diffusion models.

In diffusion-based generative modeling:

- The **forward diffusion process** incrementally corrupts data by adding Gaussian noise:
$$
x_t = \sqrt{\alpha_t}\, x_0 + \sqrt{1-\alpha_t}\, \varepsilon,
\qquad \varepsilon \sim \mathcal{N}(0, I).
$$
- The choice of Gaussian noise ensures analytical tractability, stability, and well-defined score functions:
$$
\nabla_x \log \mathcal{N}(x;\mu,\Sigma).
$$

Gauss’s formulation of the normal distribution therefore provides the **probabilistic substrate** upon which diffusion processes, score matching, and modern generative diffusion models are built.

This observed motion could not be explained by classical deterministic mechanics alone and suggested the presence of intrinsic randomness in physical systems.

Robert Brown’s observation supplied the **empirical foundation** for diffusion theory.

- The random particle motion he documented motivated the development of **probabilistic descriptions of dynamics**.  
- It directly led to mathematical models of diffusion based on random fluctuations rather than deterministic trajectories.

In modern terms, Brownian motion is modeled as a stochastic process \( W_t \), forming the core noise source in stochastic dynamics:

d x(t) = f(x(t), t)\, dt + g(t)\, dW_t.
$$

Thus, Brown’s experimental discovery provided the **physical phenomenon** that ultimately inspired stochastic differential equations, diffusion equations, and the noise-driven mechanisms used in contemporary diffusion-based generative models.


Josiah Willard Gibbs.webp

## Josiah Willard Gibbs

## (1839–1903)

### Contribution

- Founded **statistical mechanics**, providing a probabilistic description of large systems composed of many particles.  
- Introduced an **energy-based view of probability**, where probabilities arise from energy functions rather than explicit normalization by counting.  
- Established deep connections between **entropy**, **probability**, and **thermodynamics**, formalizing equilibrium behavior.

The Gibbs (Boltzmann–Gibbs) distribution is defined as:
$$
p(x) = \frac{1}{Z}\, \exp\!\left(-\frac{E(x)}{k_B T}\right),
$$
where:
- \( E(x) \) is the energy of state \( x \),
- \( T \) is temperature,
- \( k_B \) is Boltzmann’s constant,
- \( Z = \int \exp(-E(x)/(k_B T))\, dx \) is the partition function.

Gibbs also formalized entropy as:
$$
H(p) = - \int p(x)\, \log p(x)\, dx.
$$

### Role in Diffusion Models

Gibbs established the **equilibrium and energy-based perspective** that underlies diffusion and modern generative modeling.

In diffusion models:

- Target data distributions are implicitly treated as **energy-based models**:
$$
p(x) \propto \exp(-E(x)).
$$
- Diffusion processes can be interpreted as systems that evolve toward or sample from an equilibrium distribution governed by an energy landscape.
- The reverse-time diffusion process corresponds to **relaxation toward equilibrium**, analogous to thermodynamic systems minimizing free energy.

Thus, Gibbs’s framework provides the **conceptual bridge** between thermodynamics, probability, diffusion processes, and contemporary energy-based and score-based generative models.


## Emergence of Vector Fields in Modern Form

Vector fields in their modern mathematical form emerged in the **late 19th century**, primarily through the independent and complementary work of **Josiah Willard Gibbs** and **Oliver Heaviside**. Their motivation was pragmatic rather than abstract: to simplify and clarify **James Clerk Maxwell’s equations of electromagnetism**, which were originally expressed using **quaternions**.

Although earlier scientists such as **Isaac Newton** and **Leonhard Euler** had already worked implicitly with vector-like quantities—forces, velocities, and fluid flows—their formulations lacked a unified and explicit vector calculus framework.

---

### The Gibbs–Heaviside Vector Analysis Revolution

Gibbs and Heaviside made a decisive conceptual break by:

- **Separating scalars and vectors** from quaternion algebra  
- Treating vectors as first-class mathematical objects  
- Introducing a compact and intuitive algebra for physical laws  

They established the core operations and operators that define vector fields today.

---

### Fundamental Vector Operations

The following algebraic operations were formalized and standardized:

- **Dot product** (measuring projection and alignment):
$$
\mathbf{a} \cdot \mathbf{b}
$$

- **Cross product** (measuring oriented area and rotational structure):
$$
\mathbf{a} \times \mathbf{b}
$$

These operations replaced cumbersome quaternion expressions and made physical interpretation transparent.

---

### Differential Operators on Vector Fields

Gibbs and Heaviside also introduced the differential operators that govern spatial variation:

- **Gradient** (direction of steepest increase of a scalar field):
$$
\nabla \phi
$$

- **Divergence** (local expansion or compression of a vector field):
$$
\nabla \cdot \mathbf{v}
$$

- **Curl** (local rotation of a vector field):
$$
\nabla \times \mathbf{v}
$$

These operators are the backbone of modern physics, fluid dynamics, and electromagnetism.

---

### Conceptual Significance

This reformulation achieved several lasting outcomes:

- Maxwell’s equations were reduced to a compact and elegant vector form  
- Physical laws became **operator-based** rather than coordinate-dependent  
- Vector fields became the natural language for describing continuous systems  

This shift laid the groundwork for later developments in:

- Fluid mechanics  
- Statistical mechanics  
- Partial differential equations  
- Diffusion and transport theory  

---

### One-Sentence Synthesis

Gibbs and Heaviside transformed scattered notions of forces and flows into a unified **vector field calculus**, replacing quaternions with gradients, divergences, and curls—and in doing so, established the mathematical language that underlies modern physics, diffusion theory, and operator-based modeling.


Oliver Heaviside.jpg

## Oliver Heaviside

## (1850 - 1925)

### Core Contribution

Oliver Heaviside was a self-taught mathematician and electrical engineer who introduced **operational calculus**—a radically new way of reasoning about differential equations by treating differentiation as an **algebraic operator** rather than a purely analytic limit process.

His central idea was to represent differentiation symbolically as an operator:
$$
D = \frac{d}{dt},
$$
and to manipulate differential equations algebraically, decades before full mathematical rigor was established.

---

### What Heaviside Actually Introduced

#### Operator-Based View of Differential Equations

Heaviside rewrote differential equations as **algebraic equations in operators**, shifting attention from explicit solutions to operator structure.

Conceptually:
$$
(D^2 + aD + b)\,x(t) = f(t).
$$

This perspective emphasized *how* systems act on signals, not how to solve each equation from scratch.

#### Linear Systems and Impulse Response

Heaviside introduced the idea that a system is fully characterized by its **response to an impulse**.  
This directly led to the modern concept of **Green’s functions**, where solutions are constructed from impulse responses.

#### Step and Impulse Functions

He introduced the **Heaviside step function**, later formalized within distribution theory, enabling compact representations of time-evolving systems:
$$
H(t) =
\begin{cases}
0, & t < 0, \\
1, & t \ge 0.
\end{cases}
$$

This allowed differential equations to be treated operationally rather than pointwise.

---

### Why This Matters for Diffusion Models

Although Heaviside worked long before probabilistic diffusion models, his ideas are **structurally foundational**.

#### 1. Operator View of Dynamics

Diffusion models rely fundamentally on operators:

- Gradients and divergences  
- Laplacians  
- Fokker–Planck operators governing probability evolution  
- Reverse-time operators in generative sampling  

Heaviside’s philosophy can be summarized as:

> Understand systems through operators, not explicit solutions.

This is exactly the mindset behind modern diffusion theory.

#### 2. From Operators to Stein’s Method

There is a clear conceptual lineage:

- **Heaviside**: differential operators as fundamental objects  
- **Stein**: probability distributions characterized by operator identities  
- **Hyvärinen / Song**: learning these operators (scores) with neural networks  

A Stein operator takes the form:
$$
\mathcal{T}_p f(x)
=
\nabla_x \log p(x)\, f(x)
+
\nabla_x f(x),
$$
which is philosophically **Heaviside’s operational calculus applied to probability distributions**.

#### 3. Linear Response to Diffusion Kernels

Diffusion processes rely on:

- Green’s functions  
- Heat kernels  
- Gaussian transition densities  

These are direct descendants of Heaviside’s **impulse response** and **convolution-based system evolution** viewpoint.

---

### Role in the Diffusion Intellectual Timeline

| Scientist | Key Idea |
|---------|----------|
| Euler | Differential equations |
| Heaviside | Operator calculus |
| Gibbs | Energy and equilibrium |
| Stein | Operator-based probability |
| Song | Learned score operators |

Heaviside forms the **operator-theoretic bridge** between classical differential equations and modern probabilistic operators.

---

### What Heaviside Did *Not* Do (Important Clarification)

- He did not work directly on probability theory  
- He did not define stochastic processes  
- His methods were initially non-rigorous and later formalized via distribution theory and functional analysis  

Despite this, his intuition anticipated modern mathematics by decades.

---

### One-Sentence Synthesis

Oliver Heaviside introduced the **operator way of thinking** that ultimately allowed probability distributions, diffusion processes, and score functions to be defined and manipulated through differential operators rather than explicit formulas.


Andrey Markov.jpg

## Andrey Markov

## (1856–1922)

### Contribution

- Introduced **Markov chains**, providing a formal mathematical framework for stochastic processes that evolve through a sequence of states.  
- Formalized the concept of **memoryless (Markov) processes**, where the future depends only on the present state, not on the full past history.

The Markov property is expressed as:
$$
p(x_{t+1} \mid x_t, x_{t-1}, \ldots, x_0)
=
p(x_{t+1} \mid x_t).
$$

In continuous time, this idea underlies Markov stochastic processes such as diffusion:
$$
p(x_{t+\Delta t} \mid x_t).
$$

### Relevance to Diffusion Models

Diffusion models are fundamentally **Markovian generative processes**.

- The **forward diffusion process** adds noise in a step-by-step Markov manner:
$$
p(x_t \mid x_{t-1}).
$$
- The **reverse-time generative process** also assumes Markov structure:
$$
p(x_{t-1} \mid x_t).
$$
- This property enables tractable training, sampling, and likelihood factorization over time.

Markov’s formalization of memoryless stochastic evolution therefore provides the **structural backbone** for diffusion models, allowing complex distributions to be generated through a sequence of simple local transitions.


Louis Bachelier.jpg

## Louis Bachelier

## (1870–1946)

### Contribution

- Developed the **first mathematical model of random walks** in continuous time (1900).  
- Introduced **Brownian motion** as a mathematical object prior to its physical explanation by Einstein.  
- Modeled randomness using continuous-time stochastic processes, particularly in the context of financial markets.

Bachelier modeled price dynamics as a continuous stochastic process of the form:
$$
X(t) = X(0) + \sigma W_t,
$$
where \( W_t \) denotes Brownian motion and \( \sigma \) controls the scale of fluctuations.

### Role in Diffusion Models

Bachelier’s work represents the **first formal introduction of continuous-time stochastic processes** as models of randomness.

- His formulation anticipated the mathematical structure of **diffusion processes**.  
- It established the idea that randomness can be described by smooth-time stochastic trajectories rather than discrete jumps.
- This viewpoint directly underlies modern diffusion modeling, where data evolves through continuous-time noise injection and removal.

Thus, Bachelier provided the **mathematical precursor** to stochastic differential equations and diffusion-based generative models by framing randomness as a continuous-time process.


Albert Einstein.webp

## Albert Einstein

## (1879–1955)

### Contribution

- Provided the first **rigorous statistical theory of Brownian motion** (1905).  
- Established a quantitative link between **microscopic random motion** of particles and **macroscopic diffusion phenomena**.  
- Gave physical meaning to randomness by deriving observable diffusion behavior from molecular fluctuations.

Einstein showed that the probability density of particle positions satisfies the diffusion equation:
$$
\frac{\partial p(x,t)}{\partial t}
=
D \nabla^2 p(x,t),
$$
where \( D \) is the diffusion coefficient.

He also derived the mean-squared displacement relation:
$$
\mathbb{E}\!\left[\|x(t) - x(0)\|^2\right]
=
2 d D t,
$$
where \( d \) is the spatial dimension.

### Relevance to Diffusion Models

Einstein’s work established the **physical interpretation of diffusion as noise-driven evolution**.

- Random microscopic fluctuations aggregate into smooth macroscopic diffusion.  
- This principle justifies modeling diffusion as a stochastic process rather than a deterministic one.
- Modern diffusion models mirror this idea by interpreting noise injection as a principled physical process rather than arbitrary corruption.

Thus, Einstein provided the **physical and statistical justification** for diffusion processes, forming a cornerstone of both classical diffusion theory and modern diffusion-based generative modeling.


Paul Langevin.jpg

## Paul Langevin

## (1872–1946)

### Contribution

- Introduced the **Langevin equation**, providing a dynamical description of systems influenced by both deterministic forces and random fluctuations.  
- Decomposed motion into two components:
  - **Deterministic drift**, representing systematic physical forces.
  - **Stochastic noise**, representing random microscopic interactions.

The Langevin equation is written as:
$$
d x(t) = f(x(t))\, dt + \sigma\, dW_t,
$$
where:
- \( f(x(t)) \) is the deterministic drift term,
- \( \sigma \) controls the noise intensity,
- \( dW_t \) is a Brownian motion (Wiener process) increment.

### Role in Diffusion Models

Langevin dynamics is the **direct conceptual and mathematical ancestor** of modern diffusion model sampling.

- Diffusion models use learned drift terms (scores) combined with Gaussian noise:
$$
d x(t) = \nabla_x \log p_t(x)\, dt + \sqrt{2}\, dW_t.
$$
- Reverse-time diffusion and score-based generative models can be interpreted as **Langevin dynamics guided by learned score functions**.
- Sampling in diffusion models corresponds to iteratively applying drift–noise updates until equilibrium is reached.

Langevin’s formulation therefore provides the **core dynamical template** that connects classical stochastic physics to contemporary diffusion-based generative modeling.


Ronald Fisher.jpg

## Ronald A. Fisher

## (1890–1962)

### Contribution

- Introduced **likelihood theory**, establishing a systematic framework for statistical inference from data.  
- Defined the **score function**, the gradient of the log-likelihood with respect to parameters or variables.  
- Introduced **Fisher information**, quantifying the amount of information a random variable carries about unknown parameters.

For a probability density \( p(x;\theta) \), the **score function** is defined as:
$$
s(x;\theta)
=
\nabla_\theta \log p(x;\theta).
$$

The **Fisher information** is given by:
$$
\mathcal{I}(\theta)
=
\mathbb{E}_{x \sim p(x;\theta)}
\left[
\left(\nabla_\theta \log p(x;\theta)\right)
\left(\nabla_\theta \log p(x;\theta)\right)^\top
\right].
$$

These concepts form the mathematical core of classical statistical estimation and asymptotic theory.

### Role in Diffusion Models

Fisher’s score function is the **central quantity learned in score-based diffusion models**.

In modern diffusion modeling:

- The score is defined with respect to the data variable rather than parameters:
$$
s(x,t) = \nabla_x \log p_t(x).
$$
- Neural networks are trained to approximate this score at different noise levels \( t \).  
- The learned score replaces explicit likelihoods and normalization constants, enabling scalable generative modeling.

Sampling is then performed by integrating stochastic dynamics driven by the score:
$$
d x(t) = s(x,t)\, dt + dW_t.
$$

Thus, Fisher’s introduction of the score function provides the **statistical foundation** upon which score matching, Langevin dynamics, and modern diffusion-based generative models are built.


Kiyosi Itô.jpg

## Kiyosi Itô

## (1915–2008)

### Contribution

- Invented **Itô calculus**, a mathematical framework for integration and differentiation with respect to stochastic processes.  
- Provided the first rigorous and complete formalization of **stochastic differential equations (SDEs)**.  
- Introduced fundamental tools such as **Itô integrals**, **Itô’s lemma**, and martingale-based analysis.

A general Itô stochastic differential equation is written as:
$$
dX_t = b(X_t, t)\, dt + \sigma(X_t, t)\, dW_t,
$$
where:
- \( b(X_t, t) \) is the drift term,
- \( \sigma(X_t, t) \) is the diffusion coefficient,
- \( W_t \) is a Wiener (Brownian motion) process.

Itô’s lemma, the stochastic analogue of the chain rule, states that for a smooth function \( f(X_t, t) \):
$$
df(X_t, t)
=
\left(
\frac{\partial f}{\partial t}
+
b \frac{\partial f}{\partial x}
+
\frac{1}{2}\sigma^2 \frac{\partial^2 f}{\partial x^2}
\right) dt
+
\sigma \frac{\partial f}{\partial x}\, dW_t.
$$

### Role in Diffusion Models

Itô’s work made **diffusion processes mathematically rigorous and analyzable**.

In diffusion modeling:

- Forward and reverse diffusion processes are explicitly defined as **SDEs**.  
- The evolution of probability densities is derived using Itô calculus, leading to equations such as the **Fokker–Planck equation**:
$$
\frac{\partial p(x,t)}{\partial t}
=
-\nabla \cdot (b(x,t)p(x,t))
+
\frac{1}{2}\nabla^2 \big(\sigma^2(x,t)p(x,t)\big).
$$
- Reverse-time diffusion and score-based models rely on Itô’s framework to ensure correctness of stochastic simulation and sampling.

Kiyosi Itô therefore provided the **mathematical backbone** that transformed diffusion from a physical intuition into a precise, controllable, and modern tool for probabilistic modeling and generative AI.


Gisiro Maruyama.jpg

## Gisiro Maruyama

## (1916–1986)

### Contribution

- Co-developed the **Euler–Maruyama method**, the stochastic extension of Euler’s method.  
- Enabled practical **numerical simulation of stochastic differential equations (SDEs)**.  
- Bridged rigorous stochastic calculus with implementable algorithms.

For an Itô SDE
$$
dX_t = b(X_t,t)\,dt + \sigma(X_t,t)\,dW_t,
$$
the Euler–Maruyama discretization with step size \( \Delta t \) is:
$$
X_{t+\Delta t}
=
X_t
+
b(X_t,t)\,\Delta t
+
\sigma(X_t,t)\,\sqrt{\Delta t}\,\varepsilon_t,
\qquad
\varepsilon_t \sim \mathcal{N}(0, I).
$$

### Relevance to Diffusion Models

The Euler–Maruyama method makes **reverse-time diffusion sampling computationally feasible**.

In diffusion and score-based generative models:

- The reverse-time SDE is simulated numerically using Euler–Maruyama updates.  
- Learned drift terms (scores) are combined with Gaussian noise at each step.  
- Sampling reduces to a sequence of simple, stable updates over discretized time.

Thus, Maruyama’s contribution provides the **algorithmic workhorse** that turns continuous-time diffusion theory into efficient, practical generative model sampling.


Solomon Kullback.jpg

## Solomon Kullback

## (1907–1994)

### Contribution

- Co-introduced the **Kullback–Leibler (KL) divergence**, a fundamental measure of discrepancy between probability distributions.  
- Established a core concept in **information theory** that quantifies how one distribution deviates from another.

The KL divergence between two distributions \( p \) and \( q \) is defined as:
$$
\mathrm{KL}(p \,\|\, q)
=
\int p(x)\, \log \frac{p(x)}{q(x)} \, dx.
$$

KL divergence is non-negative and asymmetric, and it measures information loss when \( q \) is used to approximate \( p \).

### Role in Diffusion Models

KL divergence is the **implicit objective** underlying likelihood-based training of diffusion models.

In particular:

- The **DDPM evidence lower bound (ELBO)** decomposes into a **sum of KL divergences** across timesteps:
$$
\mathcal{L}_{\text{ELBO}}
=
\sum_t
\mathrm{KL}\!\left(
q(x_{t-1}\mid x_t, x_0)
\;\|\;
p_\theta(x_{t-1}\mid x_t)
\right)
+
\text{const}.
$$
- **Likelihood-weighted score matching** can be shown to minimize a KL divergence between true and model-implied path distributions.
- **Probability flow ODEs** enable exact likelihood computation, making the KL-based interpretation explicit:
$$
\log p_\theta(x_0)
=
\log p_T(x_T)
-
\int_0^T
\mathrm{KL}\text{-related drift terms}\, dt.
$$

Through these mechanisms, Kullback’s work provides the unifying principle connecting:
- **Information theory**  
- **Variational inference**  
- **Diffusion model training objectives**

KL divergence thus serves as the **information-theoretic backbone** of diffusion-based generative modeling.


Richard Leibler.webp

## Richard Leibler

## (1914 - 2003)

### Contribution

Richard Leibler, together with Solomon Kullback, introduced the **Kullback–Leibler (KL) divergence** in their seminal 1951 paper *“On Information and Sufficiency.”*  
KL divergence provides a fundamental, asymmetric measure of how one probability distribution differs from another.

It is defined as:
$$
\mathrm{KL}(p \,\|\, q)
=
\mathbb{E}_{p}
\left[
\log \frac{p(x)}{q(x)}
\right]
=
\int p(x)\, \log \frac{p(x)}{q(x)} \, dx.
$$

This work established a central pillar of **information theory**, **statistical inference**, and **variational methods**, formalizing the notion of information loss when approximating one distribution with another.

---

### Why This Matters for Diffusion Models

Although diffusion models do not always explicitly optimize KL divergence in their final training loss, **KL divergence is the hidden objective throughout diffusion theory**.

#### Training Objectives

- **Maximum likelihood training** is equivalent to minimizing a KL divergence between the data distribution and the model distribution:
$$
\min_\theta \; \mathrm{KL}\!\left(p_{\text{data}} \,\|\, p_\theta\right).
$$

- In **DDPMs**, the evidence lower bound (ELBO) decomposes into a **sum of KL divergences** between successive conditional distributions:
$$
\mathcal{L}_{\text{ELBO}}
=
\sum_t
\mathrm{KL}
\left(
q(x_{t-1}\mid x_t, x_0)
\;\|\;
p_\theta(x_{t-1}\mid x_t)
\right)
+
\text{const}.
$$

- In **score-based diffusion**, likelihood-weighted score matching can be shown to minimize a KL divergence between **true and model-induced path distributions**.

#### Probability Flow ODE

- The **probability flow ODE** enables exact likelihood computation.
- Likelihood evaluation relies on a change-of-variables formula whose optimality and consistency are measured in **KL divergence terms**.

#### Variational Interpretation

- Diffusion models can be viewed as **variational inference over stochastic trajectories**, where KL divergence quantifies approximation error at each time step.
- Each forward–reverse transition implicitly minimizes a KL gap between true and learned processes.

#### Model Comparison and Evaluation

- Many theoretical guarantees in diffusion models (convergence, optimality, consistency) are stated using **KL divergence** or closely related information-theoretic quantities.

## Stanisław Ulam

## (1909–1984)

### Contribution

- Co-invented **Monte Carlo methods**, introducing randomized simulation as a systematic computational tool.  
- Formalized **stochastic simulation** for analyzing complex, high-dimensional systems that are analytically intractable.  
- Established randomness as a legitimate and powerful mechanism for numerical computation.

In Monte Carlo estimation, expectations are approximated via samples:
$$
\mathbb{E}_{p(x)}[f(x)]
\;\approx\;
\frac{1}{N}\sum_{i=1}^N f(x^{(i)}),
\qquad x^{(i)} \sim p(x).
$$

### Role in Diffusion Models

Diffusion models can be interpreted as **large-scale Monte Carlo samplers** guided by learned vector fields.

Specifically:

- **Reverse diffusion** corresponds to *guided Monte Carlo sampling* from a complex target distribution:
$$
x_{t-1} \sim p_\theta(x_{t-1}\mid x_t).
$$
- **Langevin dynamics** is a form of **Markov Chain Monte Carlo (MCMC)** with score guidance:
$$
x_{k+1} = x_k + \eta \nabla_x \log p(x_k) + \sqrt{2\eta}\,\varepsilon_k,
\qquad \varepsilon_k \sim \mathcal{N}(0,I).
$$
- **Predictor–Corrector samplers** combine numerical SDE solvers (predictor) with Langevin MCMC steps (corrector), forming advanced Monte Carlo schemes for efficient sampling.

Ulam’s contribution thus provides the **computational philosophy** behind diffusion sampling:  
complex distributions are explored not by closed-form solutions, but by **guided stochastic simulation**, which lies at the heart of modern diffusion-based generative models.


Charles M. Stein.jpg

## Charles M. Stein

## (1920–2016)

### Contribution

- Introduced **Stein’s method**, a powerful framework for assessing the distance between probability distributions.  
- Showed that a probability distribution can be **characterized by operator identities** involving expectations.  
- Embedded the **score function** (gradient of the log-density) into distribution-comparison theory through what are now called **Stein operators**.

For a target distribution \( p(x) \) with score
$$
s_p(x) = \nabla_x \log p(x),
$$
a canonical Stein operator takes the form:
$$
\mathcal{A}_p f(x)
=
s_p(x)^\top f(x) + \nabla_x \cdot f(x),
$$
with the defining property:
$$
\mathbb{E}_{x \sim p}\!\left[\mathcal{A}_p f(x)\right] = 0
\quad \text{for suitable test functions } f.
$$

### Role in Diffusion Models

Stein’s work provides the **theoretical justification for comparing distributions using scores alone**.

In the context of diffusion and score-based models:

- One can compare or match distributions **without access to normalized densities**, relying only on score functions.  
- This insight underlies **score matching**, where models are trained to approximate \( \nabla_x \log p(x) \) rather than \( p(x) \) itself.  
- It legitimizes learning and evaluating generative models through **local gradient information** instead of global likelihoods.

Stein’s method thus supplies the **theoretical bridge** connecting classical probability theory, score matching, and modern diffusion-based generative modeling.


Brian Anderson.jpg

## Brian D. O. Anderson

## (born 1941)

### Contribution

- Formalized **reverse-time stochastic differential equations** (1982).  
- Proved precise conditions under which **diffusion processes can be reversed in time**.  
- Established how the drift of a forward diffusion must be modified to obtain a valid reverse-time stochastic process.

For a forward Itô SDE
$$
dX_t = b(X_t,t)\,dt + \sigma(t)\,dW_t,
$$
Anderson showed that the corresponding **reverse-time SDE** takes the form:
$$
dX_t =
\Big[
b(X_t,t)
-
\sigma^2(t)\,\nabla_x \log p_t(X_t)
\Big] dt
+
\sigma(t)\, d\bar W_t,
$$
where \( p_t(x) \) is the marginal density at time \( t \), and \( \bar W_t \) is a reverse-time Brownian motion.

### Relevance to Diffusion Models

Anderson’s work provides the **mathematical justification for sampling data from noise**.

In diffusion-based generative modeling:

- The **forward process** gradually destroys structure by adding noise.  
- The **reverse-time process**, guaranteed by Anderson’s theory, reconstructs structure by following a corrected drift involving the **score function**:
$$
\nabla_x \log p_t(x).
$$
- This result underpins **reverse diffusion sampling**, making it theoretically sound rather than heuristic.

Thus, Anderson’s formalization supplies the **rigorous bridge** that allows modern diffusion models to start from pure noise and reliably generate structured data.

## Geoffrey Hinton

## (born 1947)

### Contribution

- Developed **energy-based models**, framing probability distributions in terms of energy functions rather than explicit normalized densities.  
- Popularized **unnormalized probabilistic modeling**, where learning does not require computing the partition function explicitly.  
- Introduced **contrastive divergence**, a practical algorithm for training energy-based models by comparing nearby distributions.

In energy-based modeling, a distribution is defined implicitly as:
$$
p(x) \propto \exp(-E(x)),
$$
where \( E(x) \) is an energy function and the normalization constant
$$
Z = \int \exp(-E(x))\, dx
$$
is typically intractable.

Contrastive divergence replaces exact likelihood gradients with approximate updates based on short-run dynamics.

### Relevance to Diffusion Models

Hinton’s work laid the **conceptual foundation for learning probability distributions without explicit normalization**.

In diffusion and score-based models:

- Models learn **local information** (scores or energy gradients) instead of normalized densities.  
- Training avoids direct computation of partition functions, similar in spirit to contrastive divergence.  
- The score function can be viewed as the negative gradient of an implicit energy:
$$
\nabla_x \log p(x) = -\nabla_x E(x).
$$

This perspective directly connects energy-based modeling to diffusion models, where generation is achieved by **following learned energy or score fields** rather than evaluating explicit likelihoods.

Hinton’s contributions therefore provide a **conceptual bridge** between classical energy-based learning and modern diffusion-based generative modeling.

## Aapo Hyvärinen

## (born 1967)

### Contribution

- Introduced **score matching** (2005), a learning principle for estimating probability distributions **without computing normalization constants**.  
- Showed that models can be trained by matching **score functions** rather than probability densities themselves.  
- Eliminated the need for access to the **true data distribution** or its normalizing constant.

The score function of a distribution \( p(x) \) is defined as:
$$
s(x) = \nabla_x \log p(x).
$$

Score matching minimizes the **Fisher divergence** between the data distribution \( p_{\text{data}} \) and the model \( p_\theta \):
$$
\mathcal{J}(\theta)
=
\mathbb{E}_{p_{\text{data}}(x)}
\left[
\left\|
\nabla_x \log p_\theta(x)
-
\nabla_x \log p_{\text{data}}(x)
\right\|^2
\right].
$$

Crucially, Hyvärinen showed this objective can be rewritten so that
\( \nabla_x \log p_{\text{data}}(x) \) **never needs to be known explicitly**.

### Relevance to Diffusion Models

Hyvärinen’s score matching made **score-based learning computationally tractable**, directly enabling modern diffusion models.

In diffusion modeling:

- Neural networks are trained to approximate
$$
s_\theta(x,t) \approx \nabla_x \log p_t(x),
$$
at multiple noise levels \( t \).  
- Training relies on **denoising score matching**, a direct extension of Hyvärinen’s original formulation.  
- This bypasses explicit likelihoods and partition functions while remaining statistically principled.

Thus, Hyvärinen’s work provides the **key learning mechanism** that allows diffusion models to estimate complex, high-dimensional distributions using only local gradient information.

## Jascha Sohl-Dickstein

### Core Contribution (Why He Is Pivotal)

Jascha Sohl-Dickstein is the **direct modern originator of diffusion-based generative models in machine learning**.

He was the first to explicitly formulate **generative modeling as the reversal of a gradual diffusion (noising) process**, creating the conceptual bridge between:

- **Statistical physics diffusion**
- **Modern deep generative models**

This single idea defines the modern diffusion paradigm.

---

### 1. Diffusion as a Generative Model (2015)

#### Key Work

**“Deep Unsupervised Learning using Nonequilibrium Thermodynamics” (2015)**

#### What He Introduced

- A **forward diffusion process** that gradually destroys data by adding noise:
$$
q(x_t \mid x_{t-1})
$$
- A **learned reverse process** that reconstructs data from noise:
$$
p_\theta(x_{t-1} \mid x_t)
$$
- A **variational objective (ELBO)** to train the reverse process:
$$
\log p_\theta(x_0) \ge \mathbb{E}_q[\text{ELBO}]
$$

This was the **first complete diffusion-based generative framework** in machine learning.

---

### 2. Thermodynamic Perspective (Crucial Insight)

Sohl-Dickstein framed generative modeling using ideas from **nonequilibrium thermodynamics**:

- Forward diffusion corresponds to **entropy increase**
- Learned reverse dynamics correspond to **entropy reduction**

This interpretation:

- Explained why gradual noising is *reversible*
- Provided theoretical grounding for **stable training**
- Justified the use of many small noise steps

This is why modern diffusion models:

- Use long chains of small perturbations  
- Are stable compared to GANs  
- Avoid mode collapse  

---

### 3. Markovian Structure of Diffusion Models

He formalized diffusion as:

- A **Markov chain**
- With **tractable forward transitions**
- And **learnable reverse transitions**

This structure is inherited by:

- DDPMs  
- Score-based diffusion models  
- SDE-based diffusion models  

The defining property:
$$
p(x_{t-1} \mid x_t, x_{t+1}, \dots) = p(x_{t-1} \mid x_t)
$$

---

### 4. Relationship to Score-Based Models (Later Unification)

Initially:

- Sohl-Dickstein trained diffusion via **variational bounds**
- Score-based models were trained via **score matching**

Later work (with Song, Kingma, Ermon, Poole, Kumar) showed:

- Both approaches are **mathematically equivalent**
- Both are discretizations of the **same underlying SDE**
- Both learn the same object:
$$
\nabla_x \log p_t(x)
$$

This culminated in:

**“Score-Based Generative Modeling through Stochastic Differential Equations” (ICLR 2021)**

## Yang Song

### Contribution

- Unified **score matching**, **diffusion processes**, and **stochastic differential equations (SDEs)** into a single, coherent generative modeling framework.  
- Reformulated diffusion models as **continuous-time SDEs**, showing that many seemingly different generative models are mathematically equivalent.  
- Introduced **score-based generative modeling via SDEs**, providing a general theory that subsumes DDPMs, NCSN, and related methods.

In this framework, the forward diffusion process is written as an SDE:
$$
dX_t = f(X_t,t)\,dt + g(t)\,dW_t,
$$
with marginal densities \( p_t(x) \).

The corresponding reverse-time SDE is:
$$
dX_t =
\big[
f(X_t,t) - g^2(t)\,\nabla_x \log p_t(X_t)
\big] dt
+ g(t)\, d\bar W_t.
$$

### Relevance to Diffusion Models

Yang Song’s work provides the **modern unifying theory of diffusion models**.

Specifically:

- **Score matching** supplies the learning objective for estimating
$$
\nabla_x \log p_t(x).
$$
- **Diffusion processes** define how data is progressively corrupted into noise.  
- **Reverse-time SDEs** define how samples are generated from noise using the learned score.

This unification enables:
- A principled interpretation of diffusion models as **probability flow systems**.  
- Flexible sampler design via SDE solvers and ODE solvers.  
- A clean theoretical bridge between classical stochastic processes and modern deep generative models.

Yang Song’s contribution completes the diffusion journey by turning earlier ideas into a **single, mathematically consistent generative modeling framework**.
