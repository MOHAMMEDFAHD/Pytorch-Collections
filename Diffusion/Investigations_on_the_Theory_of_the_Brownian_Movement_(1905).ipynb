# Albert Einstein  
## *Investigations on the Theory of the Brownian Movement* (1905)

---

## Abstract

In *Investigations on the Theory of the Brownian Movement*, Albert Einstein provides the first rigorous theoretical explanation of Brownian motion grounded in the molecular–kinetic theory of heat. He demonstrates that the irregular motion of microscopic particles suspended in a liquid arises from incessant thermal collisions with surrounding molecules. By deriving precise quantitative relationships between diffusion, temperature, viscosity, and particle size, Einstein establishes Brownian motion as direct empirical evidence for the atomic nature of matter and introduces new experimental methods for determining fundamental physical constants.

---

## Problems

Prior to Einstein’s work, classical thermodynamics and continuum mechanics were unable to account for the observable erratic motion of microscopic particles in liquids. Brownian motion was well documented experimentally but lacked a quantitative theoretical explanation. There was no clear link between macroscopic observables such as diffusion and viscosity and the underlying microscopic dynamics of molecules. This gap weakened experimental validation of the molecular–kinetic theory of heat and prevented direct measurement of atomic-scale quantities.

---

## Proposed Solutions

Einstein models suspended particles as bodies subjected to random thermal forces resulting from molecular collisions in the surrounding fluid. He treats these particles statistically, extending thermodynamic concepts—particularly osmotic pressure and diffusion—from dissolved molecules to visible suspended particles. By framing particle motion in terms of probability distributions, he derives diffusion laws and equilibrium conditions that are fully consistent with kinetic theory.

---

## Purpose

The primary purpose of the paper is to demonstrate that Brownian motion is an inevitable consequence of the molecular–kinetic theory of heat, thereby providing strong evidence for the physical existence of atoms and molecules. A secondary objective is to derive experimentally testable relations that enable the determination of fundamental constants, including Avogadro’s number, as well as microscopic particle dimensions.

---

## Methodology

Einstein employs statistical mechanics and thermodynamic reasoning to analyze systems of suspended particles. He derives an expression for the osmotic pressure exerted by suspended particles, analogous to that of dissolved molecules, and formulates equilibrium conditions in the presence of external forces. By examining particle displacements over small time intervals and invoking probabilistic arguments, he derives the diffusion equation and relates the diffusion coefficient to macroscopic fluid properties using hydrodynamic considerations.

The diffusion equation obtained is

$$
\frac{\partial p(x,t)}{\partial t} = D \nabla^2 p(x,t),
$$

where \( p(x,t) \) denotes the probability density of particle position and \( D \) is the diffusion coefficient.

Einstein further establishes a relation between diffusion and viscosity:

$$
D = \frac{kT}{6\pi \eta a},
$$

where \( k \) is Boltzmann’s constant, \( T \) the absolute temperature, \( \eta \) the fluid viscosity, and \( a \) the particle radius.

---

## Results

The paper shows that suspended particles exert an osmotic pressure equivalent to that predicted for dissolved molecules, confirming the applicability of thermodynamic principles at microscopic scales. Einstein derives the diffusion coefficient for spherical particles as a function of temperature, viscosity, and particle size. He demonstrates that the mean squared displacement of a particle grows linearly with time:

$$
\langle x^2(t) \rangle = 2Dt,
$$

implying that the characteristic displacement scales with the square root of time. These theoretical predictions are shown to be in quantitative agreement with experimental observations of Brownian motion.

---

## Conclusions

Einstein concludes that Brownian motion constitutes compelling empirical evidence for the molecular–kinetic theory of heat and the existence of atoms. His analysis transforms Brownian motion from a qualitative phenomenon into a precise quantitative tool for probing microscopic physics. The derived relations enable direct experimental determination of atomic dimensions and fundamental constants, marking a foundational advance in statistical physics and providing the conceptual basis for later developments in stochastic processes, diffusion theory, and modern statistical mechanics.

## Key Academic Insight

Einstein resolves foundational gaps in physics by replacing deterministic, continuum-based descriptions with a probabilistic kinetic framework. In doing so, he transforms Brownian motion into a quantitative probe of microscopic physics and establishes diffusion as a fundamental bridge between molecular dynamics and observable macroscopic phenomena.

## Scholarly Context

Einstein’s theory synthesizes experimental observation (Brown), diffusion theory (Fourier), statistical mechanics (Boltzmann and Gibbs), osmotic pressure theory (van ’t Hoff), and hydrodynamics (Stokes) into a unified probabilistic framework. This synthesis transformed Brownian motion from an empirical curiosity into decisive evidence for atomic theory and laid the intellectual foundation for modern stochastic processes and diffusion-based modeling.
