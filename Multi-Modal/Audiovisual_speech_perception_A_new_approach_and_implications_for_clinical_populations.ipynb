## Abstract

This paper presents a selective review of audiovisual (AV) speech perception, examining
how visible articulatory information influences auditory speech processing. The authors
survey the neurobiological and behavioral evidence underpinning AV integration, review
findings across clinical populations (autism spectrum disorder, developmental language
disorder, and hearing loss), and propose a novel paradigm for assessing AV speech
perception that avoids the methodological limitations of auditory noise and cross-category
mismatch stimuli.

---

## Problems

1. **Methodological Limitations of Existing Paradigms:** The two dominant methods for
   studying AV speech — speech-in-noise (visual gain) and the McGurk effect — carry
   significant constraints:
   - Auditory noise is particularly disruptive for individuals with developmental
     disabilities.
   - McGurk stimuli produce illusory percepts rated as poor category exemplars, potentially
     introducing decision-level confounds, particularly for populations with executive
     functioning deficits (e.g., ASD).
   - High individual variability in response to McGurk stimuli limits generalizability.
   - McGurk stimuli typically rely on low-level, meaningless syllables (CV, CVCV), reducing
     ecological validity.

2. **Gaps in Clinical Population Research:** Existing AV speech studies in clinical
   populations do not adequately control for confounds such as gaze behavior, making it
   unclear whether reduced AV integration reflects true perceptual deficits or attentional
   differences.

3. **Incomplete Developmental Account:** No mechanistic model adequately explains the
   observed U-shaped developmental function in visual influence on heard speech across
   the lifespan.

---

## Proposed Solutions

1. **Novel AV Speech Paradigm — Visual Phonemic Restoration:** A new assessment
   method using auditory token degradation rather than noise or mismatch:
   - Two auditory conditions are presented: clear tokens (e.g., $/ba/$) and reduced tokens
     (formant transitions systematically flattened so the consonant is not detected,
     perceived as $/a/$).
   - The visual signal remains constant (speaker producing $/ba/$).
   - Visual influence is measured as restoration of the weakened auditory signal:
     $$P(\text{perceived as } /ba/ \mid \text{reduced auditory} + \text{visual } /ba/) > P(\text{perceived as } /ba/ \mid \text{reduced auditory alone})$$
   - This within-category strengthening design avoids overt intersensory conflict.

2. **Concurrent Eye-Tracking Controls:** Integrating gaze recording during AV speech
   tasks to isolate true integration deficits from attentional or gaze-behavior confounds,
   particularly in ASD populations.

---

## Purpose

- To synthesize the existing literature on AV speech perception across typical and
  atypical populations.
- To identify the neural substrates, developmental trajectory, and individual-difference
  factors governing AV integration.
- To introduce a methodologically cleaner paradigm applicable to clinical populations who
  cannot reliably be assessed with noise-based or McGurk-type stimuli.

---

## Methodology

### Review Scope
The paper constitutes a selective literature review organized across five domains:

| Domain | Key Topics Covered |
|---|---|
| Visual Gain | Speech-in-noise intelligibility improvement via lip-reading |
| McGurk Effect | Cross-category AV mismatch and perceptual fusion |
| Gaze Behavior | Eye-fixation patterns to speaking faces across age and language |
| Neurobiology | fMRI, EEG/ERP localization and timing of AV processing |
| Clinical Populations | ASD, developmental language disorder, hearing loss |

### Novel Paradigm Design
- **Stimuli:** Synthesized CV tokens (/ba/) with systematically reduced formant
  transitions to create a perceptual continuum from clear to consonant-absent.
- **Visual component:** Natural face video of speaker producing /ba/, constant across
  conditions.
- **Behavioral measure:** Listener identification ratings of /ba/ category strength under
  auditory-only vs. AV conditions.
- **Electrophysiological measure:** Oddball ERP paradigm measuring mismatch negativity
  (MMN) to infrequent reduced /ba/ deviants embedded among intact /ba/ standards,
  all paired with a /ba/ face — enabling passive assessment suitable for minimally
  verbal participants.

### Key Variables
- **Independent:** Presence/absence of congruent visual speech; auditory signal clarity
- **Dependent:** Phoneme identification accuracy; MMN amplitude; eye-fixation location

---

## Results

### Behavioral and Neural Evidence (Reviewed Literature)

- **Visual Gain:** Visible articulation provides significant intelligibility benefit in noise,
  quantified as improvement in speech reception threshold.
- **McGurk Effect:** Robust cross-modal fusion occurs even when perceivers are aware of
  the manipulation, across mismatched talker gender, and within temporal asynchrony
  windows.
- **Neurobiology:**
  - fMRI identifies the posterior superior temporal sulcus (STS) as the primary site of
    AV integration in both adults and children.
  - ERP data (Bernstein et al., 2008) reveal very early ($< 100$ ms) activation of
    supramarginal gyrus, angular gyrus, intraparietal sulcus, and inferior frontal gyrus,
    preceding STS activation detected by fMRI.
  - Congruent visual speech attenuates N1/P2 amplitude and shortens latency, indicating
    facilitated auditory processing:
    $$\text{AV amplitude} < \text{A-only amplitude (N1/P2)}$$

### Clinical Population Findings

| Population | Key Finding |
|---|---|
| ASD | Reduced visual influence on heard speech even when fixated on speaker's face; gaze directed to non-focal regions (ears, cheeks, forehead) rather than mouth |
| Developmental Language Disorder | Weaker McGurk effect; poorer speechreading and speech-in-noise performance vs. typically developing peers |
| Hearing Loss / Cochlear Implants | Enhanced visual phonetic sensitivity; AV speech particularly beneficial post-implantation, with effects modulated by age of implantation |

### Novel Paradigm Findings (Preliminary)
- Typically developing adults showed significantly stronger /ba/ category ratings for
  reduced auditory tokens when paired with visual /ba/ versus auditory-only reduced
  tokens — a **within-category strengthening effect**.
- A more robust visual effect was observed for reduced acoustic signals than for
  visual-only (no acoustic) stimuli, confirming AV interaction rather than pure visual
  dominance.

---

## Conclusions

1. AV speech perception is a robust, cross-cultural phenomenon emerging early in
   development and engaging a well-defined set of neural circuits centered on the STS,
   with earlier-acting parietal and frontal components revealed by ERP.

2. Clinical populations — particularly those with ASD — show genuine AV integration
   deficits beyond what can be explained by reduced face-looking behavior alone, pointing
   toward atypical multisensory binding as a contributing mechanism in language and
   communication difficulties.

3. The proposed visual phonemic restoration paradigm represents a methodologically
   superior alternative to noise-based and McGurk-based approaches, enabling
   within-category measurement of visual influence without overt intersensory conflict or
   noise-related disruption.

4. The passive ERP oddball version of the new paradigm extends applicability to minimally
   verbal or low-functioning clinical participants, offering a neurophysiological index of
   AV integration where behavioral methods are inadequate.

5. Findings across clinical populations suggest that AV speech training may constitute a
   viable intervention target, particularly for individuals whose auditory-only speech
   processing is impaired.

# Mathematical and Statistical Content Summary

## Overview

This paper is primarily a literature review and methodological proposal in cognitive
psychology and speech science. It contains minimal formal mathematics. However, several
implicit quantitative concepts, statistical constructs, and signal-processing principles
underlie the experimental designs and findings discussed. These are identified and explained
below.

---

## 1. Visual Gain (Signal Detection Framework)

### Concept
Visual gain refers to the measurable improvement in speech intelligibility when visible
articulatory information accompanies an auditory signal presented in noise, relative to
auditory-only conditions.

### Implicit Formulation
$$\text{Visual Gain} = \text{Intelligibility}_{AV} - \text{Intelligibility}_{A\text{-only}}$$

### Role in Paper
This is the foundational quantitative effect studied since Sumby and Pollack (1954). It
establishes that the visual modality carries redundant phonetic information that
supplements degraded auditory input, producing a measurable performance benefit.

---

## 2. Auditory Noise Manipulation (Signal-to-Noise Ratio)

### Concept
Speech-in-noise studies manipulate the signal-to-noise ratio (SNR) to degrade auditory
intelligibility:

$$\text{SNR} = 10 \log_{10} \left( \frac{P_{\text{signal}}}{P_{\text{noise}}} \right) \text{ (dB)}$$

### Role in Paper
SNR manipulation is the standard method used in visual gain paradigms. By lowering SNR,
the auditory signal becomes insufficient alone, allowing researchers to measure how much
the visual signal compensates.

---

## 3. Formant Transition Manipulation (Acoustic Signal Processing)

### Concept
The novel paradigm introduces reduced auditory tokens by systematically flattening the
formant transitions of a naturally produced $/ba/$ syllable. Formants are resonant
frequency peaks $F_1, F_2, F_3, \ldots$ in the vocal tract transfer function. A
transition is the time-varying trajectory of a formant at syllable onset:

$$F_k(t) \rightarrow \text{constant} \quad \text{(flattened transition)}$$

where $F_k(t)$ denotes the $k$-th formant frequency as a function of time $t$.

### Role in Paper
Flattening transitions removes the primary acoustic cue for consonant identity, producing
a stimulus perceived as $/a/$ rather than $/ba/$. This degradation is the engine of the
new paradigm — the visual signal then restores consonant perception without any
cross-category conflict.

---

## 4. Visual Phonemic Restoration (Conditional Probability Framework)

### Concept
The core measurement in the new paradigm is whether visual information increases the
probability of correct consonant perception for a degraded auditory token:

$$P\left(\text{perceived } /ba/ \mid \text{reduced } /a/ + \text{visual } /ba/\right)
> P\left(\text{perceived } /ba/ \mid \text{reduced } /a/ \text{ alone}\right)$$

### Role in Paper
This inequality defines the within-category strengthening effect. If it holds, visual
information has restored the weakened auditory cue, analogous to the classical phonemic
restoration effect (Warren, 1970) but driven by visual rather than contextual information.

---

## 5. Within-Category Strengthening (Rating Scale Measurement)

### Concept
Participants rate perceived tokens on a category goodness scale. The statistical comparison
of interest is:

$$\bar{r}_{AV,\text{reduced}} > \bar{r}_{A\text{-only},\text{reduced}}$$

where $\bar{r}$ denotes the mean category exemplar rating for the reduced token under
AV versus auditory-only conditions.

### Role in Paper
This operationalizes the visual influence as a graded, continuous measure rather than a
binary forced-choice response, providing greater sensitivity for detecting subtle perceptual
effects, particularly in clinical populations.

---

## 6. AX Same-Different Discrimination Task (Psychophysical Method)

### Concept
In the AX paradigm, two stimuli (A and X) are presented sequentially. The listener judges
whether they are the same or different. Sensitivity is typically quantified using
discriminability:

$$d' = z(H) - z(FA)$$

where $H$ is the hit rate, $FA$ is the false alarm rate, and $z(\cdot)$ is the inverse
normal cumulative distribution function.

### Role in Paper
The authors report a more robust visual effect for the AV reduced condition than for the
visual-only condition using this task, confirming that the visual signal interacts with
residual acoustic information rather than simply replacing it.

---

## 7. Mismatch Negativity (MMN) — Oddball Paradigm

### Concept
In the ERP oddball paradigm, a deviant stimulus is embedded among frequent standard
stimuli. The MMN is the difference waveform:

$$\text{MMN}(t) = \text{ERP}_{\text{deviant}}(t) - \text{ERP}_{\text{standard}}(t)$$

It reflects automatic auditory change detection and is typically a negative deflection
peaking between 100–250 ms post-stimulus onset.

### Role in Paper
A larger (more negative) MMN for the reduced $/ba/$ deviant in the absence of visual
influence is predicted for individuals with weak AV integration, providing a passive
neurophysiological index:

$$|\text{MMN}|_{\text{weak AV integration}} > |\text{MMN}|_{\text{strong AV integration}}$$

This is the key predicted outcome for assessing AV integration in minimally verbal
clinical participants such as low-functioning children with ASD.

---

## 8. N1/P2 Amplitude Modulation (ERP Components)

### Concept
The auditory N1 (negative peak ~100 ms) and P2 (positive peak ~200 ms) components are
modulated by the presence of congruent visual speech:

$$A_{N1/P2}^{AV} < A_{N1/P2}^{A\text{-only}}$$
$$\text{Latency}_{N1/P2}^{AV} < \text{Latency}_{N1/P2}^{A\text{-only}}$$

### Role in Paper
These inequalities summarize the consistent finding across Van Wassenhove et al. (2005)
and Pilling (2009) that congruent visual speech dampens and accelerates early auditory
cortical responses, indicating that visual input facilitates and speeds auditory neural
processing rather than simply adding to it.

---

## 9. Developmental U-Shaped Function

### Concept
The observed developmental trajectory of visual influence on heard speech follows an
approximate U-shape across the lifespan:

$$\text{Visual Influence}(age) \approx \text{low (infancy)} \rightarrow
\text{high (adults)} \rightarrow \text{lower (older adults)}$$

### Role in Paper
The authors note this non-monotonic function as an unresolved empirical observation
lacking a mechanistic mathematical model. It motivates future quantitative developmental
work incorporating maturation, attentional capacity, and sensory acuity as covariates.

---

## Summary Table

| Concept | Mathematical Form | Role |
|---|---|---|
| Visual Gain | $\text{Intelligibility}_{AV} - \text{Intelligibility}_{A}$ | Quantifies benefit of visible speech in noise |
| SNR | $10\log_{10}(P_s/P_n)$ | Controls auditory signal degradation |
| Formant Flattening | $F_k(t) \to \text{constant}$ | Creates reduced auditory tokens |
| Visual Restoration | $P(/ba/ \mid \text{reduced}+\text{visual}) > P(/ba/ \mid \text{reduced})$ | Core measure of new paradigm |
| Rating Comparison | $\bar{r}_{AV} > \bar{r}_{A}$ | Within-category strengthening effect |

| N1/P2 Modulation | $A^{AV} < A^{A\text{-only}}$; $\text{Lat}^{AV} < \text{Lat}^{A\text{-only}}$ | Visual facilitation of auditory processing |
| Developmental Function | Non-monotonic across lifespan | Unresolved; motivates future modeling |

```python
# ------------------------------------------------------------
# SECTION 2: SYNTHETIC VISUAL MODALITY — MOUTH SHAPE TEMPLATES
#
# The paper discusses how visible articulatory information
# (lip shape, mouth movements) provides additional cues for
# speech perception. Here we create simplified proxy images
# representing different mouth shapes for each phoneme class.
#
# Each class gets a unique deterministic grayscale 64x64 image
# that simulates a different mouth articulation shape.
# ------------------------------------------------------------

def generate_mouth_template(label_idx, img_size=IMG_SIZE):
    """
    Generate a synthetic grayscale mouth-shape image for a given class.

    In real AV speech research, these would be frames from a video
    of a speaker's face. Here we use geometric primitives as a
    simplified educational proxy for visible articulation.

    Args:
        label_idx: Integer class index (0–5)
        img_size:  Output image size (height and width)

    Returns:
        Tensor of shape (1, img_size, img_size) — single-channel image
    """
    # Create a blank black canvas
    img = np.zeros((img_size, img_size), dtype=np.float32)

    # Center coordinates
    cx, cy = img_size // 2, img_size // 2

    # Create coordinate grids for drawing shapes
    Y, X = np.ogrid[:img_size, :img_size]

    if label_idx == 0:
        # "yes" → Wide open oval mouth (affirmative, open vowel)
        mask = ((X - cx)**2 / (22**2) + (Y - cy)**2 / (14**2)) <= 1
        img[mask] = 0.9
        # Add teeth line (horizontal stripe)
        img[cy-2:cy+2, cx-18:cx+18] = 0.3

    elif label_idx == 1:
        # "no" → Closed/tight horizontal line (negative, closed bilabial)
        img[cy-4:cy+4, cx-20:cx+20] = 0.85
        img[cy-2:cy+2, cx-18:cx+18] = 0.4  # Lip shadow

    elif label_idx == 2:
        # "up" → Small round opening (upward lip rounding)
        mask = ((X - cx)**2 + (Y - (cy-4))**2) <= 10**2
        img[mask] = 0.8
        # Raised upper lip
        img[cy-14:cy-10, cx-12:cx+12] = 0.6

    elif label_idx == 3:
        # "down" → Downward-curved mouth (open, drooping corners)
        mask = ((X - cx)**2 / (18**2) + (Y - (cy+4))**2 / (12**2)) <= 1
        img[mask] = 0.85
        # Lowered jaw indicator
        img[cy+12:cy+18, cx-16:cx+16] = 0.5

    elif label_idx == 4:
        # "left" → Asymmetric mouth shifted left (lateral gesture)
        mask = ((X - (cx-6))**2 / (20**2) + (Y - cy)**2 / (10**2)) <= 1
        img[mask] = 0.88
        img[cy-2:cy+2, cx-24:cx+4] = 0.35

    elif label_idx == 5:
        # "right" → Asymmetric mouth shifted right (lateral gesture)
        mask = ((X - (cx+6))**2 / (20**2) + (Y - cy)**2 / (10**2)) <= 1
        img[mask] = 0.88
        img[cy-2:cy+2, cx-4:cx+24] = 0.35

    # Add a subtle face-oval outline (represents the speaker's face boundary)
    face_mask = ((X - cx)**2 / (30**2) + (Y - cy)**2 / (28**2)) <= 1
    outline = ((X - cx)**2 / (30**2) + (Y - cy)**2 / (28**2))
    img[np.abs(outline - 1.0) < 0.08] = np.maximum(
        img[np.abs(outline - 1.0) < 0.08], 0.5
    )

    # Normalize to [0, 1]
    img = np.clip(img, 0, 1)

    # Return as (1, H, W) tensor — single channel (grayscale)
    return torch.tensor(img, dtype=torch.float32).unsqueeze(0)


# Pre-generate all mouth templates (one per class)
# These are fixed templates — the visual modality is deterministic per class
MOUTH_TEMPLATES = {
    idx: generate_mouth_template(idx) for idx in range(NUM_CLASSES)
}
print(f"[INFO] Generated {len(MOUTH_TEMPLATES)} mouth shape templates.")
print(f"[INFO] Visual template shape: {MOUTH_TEMPLATES[0].shape}")
```

```python
# ------------------------------------------------------------
# SECTION 3: AUDIO PREPROCESSING
#
# MFCC (Mel-Frequency Cepstral Coefficients) are the standard
# features used in speech processing. They capture the spectral
# envelope of the voice in a compact representation.
#
# Steps:
# 1. Load waveform from dataset
# 2. Resample to 16kHz
# 3. Convert stereo to mono
# 4. Compute MFCC
# 5. Pad or truncate to fixed length
# ------------------------------------------------------------

# MFCC transform — converts raw waveform to MFCC feature matrix
mfcc_transform = T.MFCC(
    sample_rate=SAMPLE_RATE,
    n_mfcc=N_MFCC,               # Number of cepstral coefficients
    melkwargs={
        'n_fft':    400,          # FFT window size
        'hop_length': 160,        # Step between windows (10ms at 16kHz)
        'n_mels':   40,           # Number of Mel filterbanks
        'f_min':    0.0,
        'f_max':    8000.0,
    }
)


def preprocess_audio(waveform, sample_rate):
    """
    Convert a raw audio waveform into fixed-size MFCC features.

    The MFCC matrix has shape (N_MFCC, T) where T varies by
    clip duration. We pad or truncate T to MAX_LEN.

    Args:
        waveform:    Raw audio tensor (channels, samples)
        sample_rate: Original sample rate of the audio

    Returns:
        mfcc: Tensor of shape (1, N_MFCC, MAX_LEN) — ready for CNN
    """
    # Step 1: Resample to 16kHz if necessary
    if sample_rate != SAMPLE_RATE:
        resampler = T.Resample(orig_freq=sample_rate, new_freq=SAMPLE_RATE)
        waveform = resampler(waveform)

    # Step 2: Convert to mono (average channels)
    if waveform.shape[0] > 1:
        waveform = waveform.mean(dim=0, keepdim=True)

    # Step 3: Compute MFCC — output shape: (1, N_MFCC, T)
    mfcc = mfcc_transform(waveform)  # shape: (1, N_MFCC, T)

    # Step 4: Fix time dimension to MAX_LEN
    T_current = mfcc.shape[-1]
    if T_current < MAX_LEN:
        # Pad with zeros on the right
        pad_amount = MAX_LEN - T_current
        mfcc = F.pad(mfcc, (0, pad_amount))
    else:
        # Truncate to MAX_LEN
        mfcc = mfcc[:, :, :MAX_LEN]

    # Step 5: Normalize MFCC values to approximately [-1, 1]
    mean = mfcc.mean()
    std  = mfcc.std() + 1e-8
    mfcc = (mfcc - mean) / std

    return mfcc  # shape: (1, N_MFCC, MAX_LEN)


def degrade_audio(mfcc, noise_std=NOISE_STD, mask_ratio=MASK_RATIO):
    """
    Degrade an MFCC tensor to simulate impoverished auditory input.

    This implements the paper's concept of "reduced tokens" where
    acoustic cues are weakened to the point where the consonant
    is not reliably detected.

    Two degradation strategies (both applied):
    1. Additive Gaussian noise — simulates SNR reduction
    2. Random frame masking    — simulates missing acoustic information

    Args:
        mfcc:       Clean MFCC tensor (1, N_MFCC, MAX_LEN)
        noise_std:  Standard deviation of added Gaussian noise
        mask_ratio: Fraction of time frames to zero out

    Returns:
        degraded: Degraded MFCC tensor, same shape as input
    """
    degraded = mfcc.clone()

    # Strategy 1: Add Gaussian noise
    noise = torch.randn_like(degraded) * noise_std
    degraded = degraded + noise

    # Strategy 2: Random temporal masking (zero out frames)
    num_frames = degraded.shape[-1]
    num_masked = int(num_frames * mask_ratio)
    mask_indices = torch.randperm(num_frames)[:num_masked]
    degraded[:, :, mask_indices] = 0.0

    return degraded
```

```python
# ------------------------------------------------------------
# SECTION 4: DATASET CLASS
#
# The MultimodalSpeechDataset class:
# - Loads audio from the Hugging Face Speech Commands dataset
# - Generates the synthetic visual modality (mouth shape)
# - Creates one of four experimental conditions per sample
# - Returns (audio_tensor, visual_tensor, label, condition)
# ------------------------------------------------------------

class MultimodalSpeechDataset(Dataset):
    """
    Multimodal dataset combining audio features and synthetic visual
    mouth-shape templates for AV speech perception experiments.

    Each sample is assigned one of four conditions (randomly during
    training to expose the model to all conditions):
        0 — Audio Clear:       clean MFCC + zero visual
        1 — Audio Degraded:    noisy MFCC + zero visual
        2 — AV Clear:          clean MFCC + mouth template
        3 — AV Restored:       noisy MFCC + mouth template

    The model learns to classify BOTH the speech class AND the
    experimental condition simultaneously.
    """

    def __init__(self, hf_dataset, label_map, augment=False):
        """
        Args:
            hf_dataset: Hugging Face dataset split (already filtered)
            label_map:  Dict mapping class name -> integer label
            augment:    If True, randomly assign conditions (training)
                        If False, use round-robin assignment (validation)
        """
        self.dataset   = hf_dataset
        self.label_map = label_map
        self.augment   = augment

        # Pre-compute indices and labels for quick access
        self.items = []
        for i, sample in enumerate(hf_dataset):
            label_name = sample['label']
            # Handle both integer and string labels
            if isinstance(label_name, int):
                # Convert integer label to string name
                label_name = hf_dataset.features['label'].int2str(label_name)
            if label_name in label_map:
                self.items.append((i, label_map[label_name]))

    def __len__(self):
        return len(self.items)

    def __getitem__(self, idx):
        """
        Returns a single (audio, visual, label, condition) tuple.
        """
        dataset_idx, label = self.items[idx]
        sample = self.dataset[dataset_idx]

        # ---- Extract raw audio ----
        audio_data  = sample['audio']
        waveform    = torch.tensor(audio_data['array'], dtype=torch.float32)
        sample_rate = audio_data['sampling_rate']

        # Ensure waveform has channel dimension
        if waveform.dim() == 1:
            waveform = waveform.unsqueeze(0)  # (1, T)

        # ---- Compute clean MFCC ----
        mfcc_clean = preprocess_audio(waveform, sample_rate)

        # ---- Assign experimental condition ----
        if self.augment:
            # Training: random condition for each sample
            condition = np.random.randint(0, NUM_CONDITIONS)
        else:
            # Validation: cycle through conditions deterministically
            condition = idx % NUM_CONDITIONS

        # ---- Build audio tensor based on condition ----
        if condition in [0, 2]:
            # Conditions 0 (Audio Clear) and 2 (AV Clear): use clean audio
            audio_tensor = mfcc_clean
        else:
            # Conditions 1 (Audio Degraded) and 3 (AV Restored): use noisy audio
            audio_tensor = degrade_audio(mfcc_clean)

        # ---- Build visual tensor based on condition ----
        if condition in [2, 3]:
            # Conditions 2 and 3 include visual information
            # Visual input = the mouth template for this speech class
            visual_tensor = MOUTH_TEMPLATES[label].clone()
        else:
            # Conditions 0 and 1: no visual input (zero tensor)
            # This represents eyes-closed or audio-only conditions
            visual_tensor = torch.zeros(1, IMG_SIZE, IMG_SIZE)

        return (
            audio_tensor,   # (1, N_MFCC, MAX_LEN)
            visual_tensor,  # (1, IMG_SIZE, IMG_SIZE)
            label,          # Integer in [0, NUM_CLASSES-1]
            condition       # Integer in [0, NUM_CONDITIONS-1]
        )
```

```python
# ------------------------------------------------------------
# SECTION 5: LOAD AND PREPARE DATASET
#
# Load Google Speech Commands from Hugging Face.
# We use "speech_commands" from the "tensorflow-datasets" mirror
# which is a verified Parquet-based dataset — no loading script,
# no trust_remote_code needed.
# Dataset: "Mxode/google-speech-commands-v2"
# ------------------------------------------------------------

print("\n[INFO] Loading Google Speech Commands dataset from Hugging Face...")
print("[INFO] This may take a moment on first run...")

# ---- Attempt multiple verified Parquet-based mirrors ----
# We try each source in order until one succeeds.
# This makes the code robust to dataset availability changes.

DATASET_CANDIDATES = [
    # Candidate 1: Mxode's Parquet mirror of Speech Commands v2
    dict(path="Mxode/google-speech-commands-v2",   config=None),
    # Candidate 2: speech_commands via datasets library built-in
    dict(path="speech_commands",                    config="v0.02"),
    # Candidate 3: Lightweight fallback — SpeechBrain mini split
    dict(path="SpeechBrain/google_speech_commands", config=None),
]

ds = None
used_source = None

for candidate in DATASET_CANDIDATES:
    try:
        print(f"[INFO] Trying: {candidate['path']} ...")
        if candidate['config']:
            ds = load_dataset(candidate['path'], candidate['config'],
                              split=None)
        else:
            ds = load_dataset(candidate['path'], split=None)
        used_source = candidate['path']
        print(f"[INFO] Successfully loaded from: {used_source}")
        break
    except Exception as e:
        print(f"[WARN] Failed ({candidate['path']}): {type(e).__name__}: "
              f"{str(e)[:80]}")
        continue

# ---- Hard fallback: synthesize a minimal dataset from scratch ----
# If ALL remote sources fail (network issues, dataset removed),
# we build a fully synthetic dataset so the rest of the pipeline
# always runs without modification.
if ds is None:
    print("\n[WARN] All remote dataset sources failed.")
    print("[INFO] Generating fully synthetic Speech Commands dataset...")
    print("[INFO] Synthetic audio = white noise bursts, one per class.")

    import random
    from datasets import Dataset as HFDataset, DatasetDict, Audio
    from datasets import Features, Value, ClassLabel, Sequence

    SR        = 16000
    DURATION  = 1        # seconds
    N_SAMPLES = SR * DURATION

    # Build synthetic records
    def make_split(n_per_class=500):
        records = []
        for cls_idx, cls_name in enumerate(CLASSES):
            for _ in range(n_per_class):
                # Unique noise seed per class → different spectral character
                rng   = np.random.RandomState(cls_idx * 1000 +
                                              len(records))
                # Low-frequency sine + noise — crude phoneme proxy
                t     = np.linspace(0, 1, N_SAMPLES)
                freq  = 200 + cls_idx * 150       # class-specific pitch
                wave  = (0.4 * np.sin(2 * np.pi * freq * t) +
                         0.6 * rng.randn(N_SAMPLES)).astype(np.float32)
                wave /= (np.abs(wave).max() + 1e-8)
                records.append({
                    'audio': {
                        'array':         wave,
                        'sampling_rate': SR,
                        'path':          f'{cls_name}_{len(records)}.wav'
                    },
                    'label': cls_name,
                })
        random.shuffle(records)
        return records

    train_records = make_split(n_per_class=400)
    val_records   = make_split(n_per_class=100)

    # Convert to Hugging Face DatasetDict
    train_hf = HFDataset.from_list(train_records)
    val_hf   = HFDataset.from_list(val_records)
    ds       = DatasetDict({'train': train_hf, 'validation': val_hf})
    used_source = 'SYNTHETIC'
    print(f"[INFO] Synthetic dataset created: "
          f"{len(train_hf)} train / {len(val_hf)} val samples")

# ---- Inspect splits ----
print(f"\n[INFO] Source         : {used_source}")
print(f"[INFO] Available splits: {list(ds.keys())}")
print(f"[INFO] Columns        : {ds[list(ds.keys())[0]].column_names}")

# ---- Select train / validation splits ----
if 'train' in ds:
    train_raw = ds['train']
else:
    # Some datasets use 'training' or only expose one split
    first_split = list(ds.keys())[0]
    train_raw   = ds[first_split]

if 'validation' in ds:
    val_raw = ds['validation']
elif 'test' in ds:
    val_raw = ds['test']
else:
    print("[INFO] No val/test split found — splitting train 90/10...")
    split   = train_raw.train_test_split(test_size=0.1, seed=42)
    train_raw = split['train']
    val_raw   = split['test']

print(f"[INFO] Raw train size : {len(train_raw)}")
print(f"[INFO] Raw val   size : {len(val_raw)}")

# ---- Inspect one sample to discover field names ----
sample_record = train_raw[0]
print(f"\n[INFO] Sample record keys  : {list(sample_record.keys())}")
print(f"[INFO] Sample label value  : {sample_record.get('label', '?')}")

# ---- Auto-discover label field ----
LABEL_FIELD = None
for candidate in ['label', 'word', 'transcription', 'text', 'command']:
    if candidate in train_raw.column_names:
        LABEL_FIELD = candidate
        break
if LABEL_FIELD is None:
    raise ValueError(f"Cannot find label column. "
                     f"Columns: {train_raw.column_names}")
print(f"[INFO] Label field         : '{LABEL_FIELD}'")

# ---- Auto-discover audio field ----
AUDIO_FIELD = None
for candidate in ['audio', 'speech', 'file', 'wav', 'sound']:
    if candidate in train_raw.column_names:
        AUDIO_FIELD = candidate
        break
if AUDIO_FIELD is None:
    raise ValueError(f"Cannot find audio column. "
                     f"Columns: {train_raw.column_names}")
print(f"[INFO] Audio field         : '{AUDIO_FIELD}'")

# ---- Discover how labels are stored (string vs int) ----
feat          = train_raw.features.get(LABEL_FIELD)
HAS_INT2STR   = hasattr(feat, 'int2str')
print(f"[INFO] Label type          : "
      f"{'ClassLabel (int→str)' if HAS_INT2STR else 'raw string'}")

# ---- Helper: resolve label to string ----
def resolve_label(raw, feat, has_int2str):
    if isinstance(raw, int) and has_int2str:
        return feat.int2str(raw)
    return str(raw)

# ---- Discover available class names in dataset ----
print("\n[INFO] Sampling label values from dataset to verify class names...")
sample_labels = set()
for i in range(min(500, len(train_raw))):
    raw = train_raw[i][LABEL_FIELD]
    sample_labels.add(resolve_label(raw, feat, HAS_INT2STR))
print(f"[INFO] Sample label names  : {sorted(sample_labels)[:15]} ...")

# ---- Verify our CLASSES exist in the dataset ----
missing = [c for c in CLASSES if c not in sample_labels]
if missing:
    # Try to find near-matches (case-insensitive)
    lower_map = {s.lower(): s for s in sample_labels}
    remapped  = {}
    for c in missing:
        if c.lower() in lower_map:
            remapped[c] = lower_map[c.lower()]
    if remapped:
        print(f"[INFO] Remapping classes: {remapped}")
        CLASSES   = [remapped.get(c, c) for c in CLASSES]
        LABEL_MAP = {cls: idx for idx, cls in enumerate(CLASSES)}
    else:
        print(f"[WARN] Classes not found in dataset: {missing}")
        print(f"[INFO] Using available classes instead...")
        available = sorted(sample_labels)
        CLASSES   = available[:NUM_CLASSES]
        LABEL_MAP = {cls: idx for idx, cls in enumerate(CLASSES)}
        print(f"[INFO] Updated CLASSES    : {CLASSES}")

print(f"[INFO] Final CLASSES       : {CLASSES}")
print(f"[INFO] Final LABEL_MAP     : {LABEL_MAP}")

# ---- Filter dataset to target classes ----
def filter_to_classes(dataset, classes, label_field, feat, has_int2str):
    """
    Filter HuggingFace dataset to only target class samples.
    Handles ClassLabel integers, plain strings, and mixed types.
    """
    target_set = set(classes)

    def is_target(example):
        raw = example[label_field]
        return resolve_label(raw, feat, has_int2str) in target_set

    return dataset.filter(is_target)

print(f"\n[INFO] Filtering to classes: {CLASSES}")
train_filtered = filter_to_classes(
    train_raw, CLASSES, LABEL_FIELD, feat, HAS_INT2STR)
val_filtered   = filter_to_classes(
    val_raw,   CLASSES, LABEL_FIELD, feat, HAS_INT2STR)
print(f"[INFO] Filtered train size : {len(train_filtered)}")
print(f"[INFO] Filtered val   size : {len(val_filtered)}")

if len(train_filtered) == 0:
    raise RuntimeError(
        "Filtered training set is empty.\n"
        f"  Label field used: '{LABEL_FIELD}'\n"
        f"  Target classes  : {CLASSES}\n"
        f"  Sample labels   : {sorted(sample_labels)[:20]}\n"
        "Check that class names exactly match the dataset's label strings."
    )

# ---- Multimodal Dataset class (field-name aware) ----
class MultimodalSpeechDataset(Dataset):
    """
    Multimodal dataset for AV speech perception experiments.

    Combines:
      - MFCC audio features from Speech Commands waveforms
      - Synthetic mouth-shape visual templates (one per class)
      - Four experimental conditions (Clear/Degraded × Audio/AV)

    Compatible with any Parquet-based Speech Commands variant
    from Hugging Face regardless of column naming conventions.
    """

    def __init__(self, hf_dataset, label_map, augment=False,
                 label_field='label', audio_field='audio',
                 feat=None, has_int2str=False):
        self.dataset      = hf_dataset
        self.label_map    = label_map
        self.augment      = augment
        self.label_field  = label_field
        self.audio_field  = audio_field
        self.feat         = feat
        self.has_int2str  = has_int2str

        # Pre-index valid samples only
        self.items = []
        for i, sample in enumerate(hf_dataset):
            raw        = sample[label_field]
            label_name = resolve_label(raw, feat, has_int2str)
            if label_name in label_map:
                self.items.append((i, label_map[label_name]))

        print(f"[INFO] MultimodalSpeechDataset: {len(self.items)} samples "
              f"| augment={augment}")

    def __len__(self):
        return len(self.items)

    def __getitem__(self, idx):
        dataset_idx, label = self.items[idx]
        sample = self.dataset[dataset_idx]

        # ---- Audio ----
        audio_data  = sample[self.audio_field]
        waveform    = torch.tensor(
            np.array(audio_data['array']), dtype=torch.float32
        )
        sr          = audio_data['sampling_rate']
        if waveform.dim() == 1:
            waveform = waveform.unsqueeze(0)

        mfcc_clean = preprocess_audio(waveform, sr)

        # ---- Condition ----
        condition = (np.random.randint(0, NUM_CONDITIONS)
                     if self.augment else idx % NUM_CONDITIONS)

        # ---- Build tensors ----
        audio_tensor  = (mfcc_clean if condition in [0, 2]
                         else degrade_audio(mfcc_clean))
        visual_tensor = (MOUTH_TEMPLATES[label].clone() if condition in [2, 3]
                         else torch.zeros(1, IMG_SIZE, IMG_SIZE))

        return audio_tensor, visual_tensor, label, condition


# ---- Instantiate datasets ----
train_dataset_full = MultimodalSpeechDataset(
    train_filtered, LABEL_MAP, augment=True,
    label_field=LABEL_FIELD, audio_field=AUDIO_FIELD,
    feat=feat, has_int2str=HAS_INT2STR
)
val_dataset_full = MultimodalSpeechDataset(
    val_filtered, LABEL_MAP, augment=False,
    label_field=LABEL_FIELD, audio_field=AUDIO_FIELD,
    feat=feat, has_int2str=HAS_INT2STR
)

# ---- Apply size constraints ----
train_size_actual = min(TRAIN_SIZE, len(train_dataset_full))
val_size_actual   = min(VAL_SIZE,   len(val_dataset_full))
train_dataset     = Subset(train_dataset_full, list(range(train_size_actual)))
val_dataset       = Subset(val_dataset_full,   list(range(val_size_actual)))

print(f"\n[INFO] Training samples   : {len(train_dataset)}")
print(f"[INFO] Validation samples : {len(val_dataset)}")

# ---- DataLoaders ----
train_loader = DataLoader(
    train_dataset, batch_size=BATCH_SIZE, shuffle=True,
    num_workers=NUM_WORKERS, pin_memory=(DEVICE.type == 'cuda'),
    drop_last=False
)
val_loader = DataLoader(
    val_dataset, batch_size=BATCH_SIZE, shuffle=False,
    num_workers=NUM_WORKERS, pin_memory=(DEVICE.type == 'cuda'),
    drop_last=False
)

# ---- Sanity checks ----
print("\n" + "="*55)
print("SANITY CHECKS")
print("="*55)
s_audio, s_visual, s_label, s_cond = train_dataset[0]
print(f"Dataset source         : {used_source}")
print(f"Audio tensor shape     : {s_audio.shape}")
print(f"Visual tensor shape    : {s_visual.shape}")
print(f"Example label          : {s_label} → {CLASSES[s_label]}")
print(f"Example condition      : {s_cond} → {CONDITIONS[s_cond]}")
print(f"Label range            : [0, {NUM_CLASSES-1}]")
print(f"Condition range        : [0, {NUM_CONDITIONS-1}]")
print(f"Batch size             : {BATCH_SIZE}")
print(f"Num training batches   : {len(train_loader)}")
print(f"Num val batches        : {len(val_loader)}")
print(f"Classes                : {CLASSES}")
print(f"Conditions             : {CONDITIONS}")
print("="*55 + "\n")
```

```python
# ------------------------------------------------------------
# SECTION 6: MODEL ARCHITECTURE
#
# The AV speech perception model has three components:
#
# 1. AUDIO ENCODER (CNN on MFCC spectrograms)
#    Input:  (B, 1, N_MFCC, MAX_LEN) = (B, 1, 40, 128)
#    Output: (B, 64) audio embedding
#
# 2. VISUAL ENCODER (CNN on mouth images)
#    Input:  (B, 1, IMG_SIZE, IMG_SIZE) = (B, 1, 64, 64)
#    Output: (B, 64) visual embedding
#
# 3. FUSION MODULE + CLASSIFIER
#    Input:  concatenated [audio_emb, visual_emb] = (B, 128)
#    Output A: class logits    (B, NUM_CLASSES=6)
#    Output B: condition logits (B, NUM_CONDITIONS=4)
#
# The dual-output design teaches the model BOTH what was said
# AND in what perceptual condition (clear/degraded/AV).
# ------------------------------------------------------------

class AudioEncoder(nn.Module):
    """
    Small CNN that processes MFCC spectrograms.

    The MFCC is treated like a 2D image:
    - Height dimension = frequency (N_MFCC coefficients)
    - Width  dimension = time      (MAX_LEN frames)

    Convolutional layers learn to detect audio patterns
    (formant transitions, voice onset, etc.).
    """

    def __init__(self, n_mfcc=N_MFCC, max_len=MAX_LEN, embed_dim=64):
        super(AudioEncoder, self).__init__()

        # Convolutional feature extractor
        # Input: (B, 1, 40, 128)
        self.conv_layers = nn.Sequential(
            # Block 1: Low-level spectral features
            nn.Conv2d(1, 16, kernel_size=3, padding=1),  # (B, 16, 40, 128)
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2),                  # (B, 16, 20, 64)

            # Block 2: Mid-level temporal-spectral patterns
            nn.Conv2d(16, 32, kernel_size=3, padding=1), # (B, 32, 20, 64)
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2),                  # (B, 32, 10, 32)

            # Block 3: High-level phonetic features
            nn.Conv2d(32, 64, kernel_size=3, padding=1), # (B, 64, 10, 32)
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((4, 4))                  # (B, 64, 4, 4)
        )

        # Projection to fixed embedding dimension
        self.projection = nn.Sequential(
            nn.Flatten(),                # (B, 64*4*4) = (B, 1024)
            nn.Linear(64 * 4 * 4, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, embed_dim),  # (B, embed_dim)
            nn.ReLU()
        )

    def forward(self, x):
        """
        Args:
            x: Audio tensor (B, 1, N_MFCC, MAX_LEN)
        Returns:
            embedding: (B, embed_dim)
        """
        features = self.conv_layers(x)
        embedding = self.projection(features)
        return embedding


class VisualEncoder(nn.Module):
    """
    Small CNN that processes synthetic mouth-shape images.

    In real AV speech research, this would process video frames
    of a speaker's face. Here we use synthetic 64x64 grayscale
    mouth templates as a simplified proxy for visible articulation.

    The visual encoder learns to extract articulatory shape features
    from these simplified face representations.
    """

    def __init__(self, img_size=IMG_SIZE, embed_dim=64):
        super(VisualEncoder, self).__init__()

        # Convolutional feature extractor
        # Input: (B, 1, 64, 64)
        self.conv_layers = nn.Sequential(
            # Block 1: Edge and shape detection
            nn.Conv2d(1, 16, kernel_size=3, padding=1),  # (B, 16, 64, 64)
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2),                  # (B, 16, 32, 32)

            # Block 2: Mouth shape features
            nn.Conv2d(16, 32, kernel_size=3, padding=1), # (B, 32, 32, 32)
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2),                  # (B, 32, 16, 16)

            # Block 3: Articulatory configuration features
            nn.Conv2d(32, 64, kernel_size=3, padding=1), # (B, 64, 16, 16)
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((4, 4))                  # (B, 64, 4, 4)
        )

        # Projection to fixed embedding dimension
        self.projection = nn.Sequential(
            nn.Flatten(),                # (B, 1024)
            nn.Linear(64 * 4 * 4, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, embed_dim),  # (B, embed_dim)
            nn.ReLU()
        )

    def forward(self, x):
        """
        Args:
            x: Visual tensor (B, 1, IMG_SIZE, IMG_SIZE)
        Returns:
            embedding: (B, embed_dim)
        """
        features = self.conv_layers(x)
        embedding = self.projection(features)
        return embedding


class AudioVisualFusionModel(nn.Module):
    """
    Full audiovisual speech perception model.

    Architecture:
        AudioEncoder  ──┐
                        ├──► Concatenate ──► FusionMLP ──► ClassHead
        VisualEncoder ──┘                              └──► ConditionHead

    The model learns a joint representation of audio and visual
    speech signals, demonstrating:
    1. How multimodal fusion combines two information streams
    2. How visual information can compensate for degraded audio
    3. How the brain might integrate seen and heard speech
    """

    def __init__(self, audio_embed_dim=64, visual_embed_dim=64,
                 fusion_dim=128, num_classes=NUM_CLASSES,
                 num_conditions=NUM_CONDITIONS):
        super(AudioVisualFusionModel, self).__init__()

        self.audio_embed_dim  = audio_embed_dim
        self.visual_embed_dim = visual_embed_dim
        fusion_input_dim = audio_embed_dim + visual_embed_dim  # 128

        # ---- Encoders ----
        self.audio_encoder  = AudioEncoder(embed_dim=audio_embed_dim)
        self.visual_encoder = VisualEncoder(embed_dim=visual_embed_dim)

        # ---- Fusion Module ----
        # Takes concatenated [audio_emb | visual_emb] and creates a
        # unified multimodal representation
        self.fusion_mlp = nn.Sequential(
            nn.Linear(fusion_input_dim, fusion_dim),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(fusion_dim, fusion_dim // 2),
            nn.ReLU(),
            nn.Dropout(0.2),
        )

        # ---- Classification Head ----
        # Primary task: identify WHAT was said (which word)
        self.class_head = nn.Linear(fusion_dim // 2, num_classes)

        # ---- Condition Head ----
        # Auxiliary task: identify the perceptual condition
        # (clear/degraded/AV) — this teaches the model to be
        # aware of the quality of its inputs
        self.condition_head = nn.Linear(fusion_dim // 2, num_conditions)

    def forward(self, audio, visual):
        """
        Forward pass through the full AV perception model.

        Args:
            audio:  Audio tensor  (B, 1, N_MFCC, MAX_LEN)
            visual: Visual tensor (B, 1, IMG_SIZE, IMG_SIZE)

        Returns:
            class_logits:     (B, num_classes)
            condition_logits: (B, num_conditions)
            audio_emb:        (B, audio_embed_dim)  — for analysis
            visual_emb:       (B, visual_embed_dim) — for analysis
        """
        # Step 1: Encode each modality independently
        audio_emb  = self.audio_encoder(audio)    # (B, 64)
        visual_emb = self.visual_encoder(visual)  # (B, 64)

        # Step 2: Multimodal fusion
        # Concatenation is the simplest fusion strategy:
        # the model sees both representations simultaneously
        fused = torch.cat([audio_emb, visual_emb], dim=1)  # (B, 128)

        # Step 3: Learn joint AV representation
        fused_repr = self.fusion_mlp(fused)  # (B, 64)

        # Step 4: Dual classification
        class_logits     = self.class_head(fused_repr)     # (B, 6)
        condition_logits = self.condition_head(fused_repr) # (B, 4)

        return class_logits, condition_logits, audio_emb, visual_emb


# ---- Instantiate model ----
model = AudioVisualFusionModel().to(DEVICE)

# ---- Count parameters ----
audio_params   = sum(p.numel() for p in model.audio_encoder.parameters()
                     if p.requires_grad)
visual_params  = sum(p.numel() for p in model.visual_encoder.parameters()
                     if p.requires_grad)
total_params   = sum(p.numel() for p in model.parameters() if p.requires_grad)

print("="*55)
print("MODEL ARCHITECTURE SUMMARY")
print("="*55)
print(model)
print(f"\nAudio encoder parameters:  {audio_params:,}")
print(f"Visual encoder parameters: {visual_params:,}")
print(f"Total trainable parameters:{total_params:,}")
print("="*55 + "\n")
```

```python
# ------------------------------------------------------------
# SECTION 7: LOSS FUNCTION AND OPTIMIZER
#
# We use two losses:
# 1. Class loss:     CrossEntropy on speech class prediction
# 2. Condition loss: CrossEntropy on condition prediction
#
# Combined loss = class_loss + lambda * condition_loss
# The condition loss acts as auxiliary supervision, encouraging
# the model to learn about the quality of its inputs.
# ------------------------------------------------------------

class_criterion     = nn.CrossEntropyLoss()
condition_criterion = nn.CrossEntropyLoss()
CONDITION_LAMBDA    = 0.3  # Weight for the auxiliary condition loss

optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE, weight_decay=1e-4)

# Learning rate scheduler: reduce LR when val loss plateaus
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=2, gamma=0.5)
```

```python
# ------------------------------------------------------------
# SECTION 8: TRAINING AND VALIDATION LOOPS
# ------------------------------------------------------------

def train_one_epoch(model, loader, optimizer, device):
    """
    Run one full pass over the training set.

    Returns:
        metrics: Dict with loss values and accuracy
    """
    model.train()

    total_loss     = 0.0
    class_loss_sum = 0.0
    cond_loss_sum  = 0.0
    correct_class  = 0
    correct_cond   = 0
    total_samples  = 0

    for batch_idx, (audio, visual, labels, conditions) in enumerate(loader):
        # Move data to device (GPU/CPU)
        audio      = audio.to(device)
        visual     = visual.to(device)
        labels     = labels.to(device)
        conditions = conditions.to(device)

        # ---- Forward pass ----
        optimizer.zero_grad()
        class_logits, cond_logits, _, _ = model(audio, visual)

        # ---- Compute losses ----
        loss_class = class_criterion(class_logits, labels)
        loss_cond  = condition_criterion(cond_logits, conditions)
        loss_total = loss_class + CONDITION_LAMBDA * loss_cond

        # ---- Backward pass ----
        loss_total.backward()

        # Gradient clipping prevents exploding gradients
        nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()

        # ---- Track metrics ----
        bs = labels.size(0)
        total_loss     += loss_total.item() * bs
        class_loss_sum += loss_class.item() * bs
        cond_loss_sum  += loss_cond.item()  * bs

        pred_class = class_logits.argmax(dim=1)
        pred_cond  = cond_logits.argmax(dim=1)
        correct_class += (pred_class == labels).sum().item()
        correct_cond  += (pred_cond  == conditions).sum().item()
        total_samples += bs

    n = total_samples
    return {
        'total_loss': total_loss / n,
        'class_loss': class_loss_sum / n,
        'cond_loss':  cond_loss_sum / n,
        'class_acc':  correct_class / n,
        'cond_acc':   correct_cond  / n,
    }


def validate(model, loader, device):
    """
    Evaluate model on validation set.

    Returns:
        metrics: Dict with losses, accuracy, and per-class/condition stats
    """
    model.eval()

    total_loss     = 0.0
    class_loss_sum = 0.0
    cond_loss_sum  = 0.0
    total_samples  = 0

    all_labels     = []
    all_conditions = []
    all_pred_class = []
    all_pred_cond  = []
    all_confidences = []

    with torch.no_grad():
        for audio, visual, labels, conditions in loader:
            audio      = audio.to(device)
            visual     = visual.to(device)
            labels     = labels.to(device)
            conditions = conditions.to(device)

            class_logits, cond_logits, _, _ = model(audio, visual)

            loss_class = class_criterion(class_logits, labels)
            loss_cond  = condition_criterion(cond_logits, conditions)
            loss_total = loss_class + CONDITION_LAMBDA * loss_cond

            bs = labels.size(0)
            total_loss     += loss_total.item() * bs
            class_loss_sum += loss_class.item() * bs
            cond_loss_sum  += loss_cond.item()  * bs
            total_samples  += bs

            # Softmax probabilities for confidence scores
            probs = F.softmax(class_logits, dim=1)
            confidence, pred_class = probs.max(dim=1)
            pred_cond = cond_logits.argmax(dim=1)

            all_labels.extend(labels.cpu().numpy())
            all_conditions.extend(conditions.cpu().numpy())
            all_pred_class.extend(pred_class.cpu().numpy())
            all_pred_cond.extend(pred_cond.cpu().numpy())
            all_confidences.extend(confidence.cpu().numpy())

    n = total_samples
    all_labels     = np.array(all_labels)
    all_conditions = np.array(all_conditions)
    all_pred_class = np.array(all_pred_class)
    all_pred_cond  = np.array(all_pred_cond)
    all_confidences = np.array(all_confidences)

    # Per-class accuracy
    per_class_acc = {}
    for c in range(NUM_CLASSES):
        mask = all_labels == c
        if mask.sum() > 0:
            per_class_acc[c] = (all_pred_class[mask] == c).mean()
        else:
            per_class_acc[c] = 0.0

    # Per-condition accuracy
    per_cond_acc = {}
    for c in range(NUM_CONDITIONS):
        mask = all_conditions == c
        if mask.sum() > 0:
            per_cond_acc[c] = (all_pred_class[mask] == all_labels[mask]).mean()
        else:
            per_cond_acc[c] = 0.0

    return {
        'total_loss':     total_loss / n,
        'class_loss':     class_loss_sum / n,
        'cond_loss':      cond_loss_sum  / n,
        'class_acc':      (all_pred_class == all_labels).mean(),
        'cond_acc':       (all_pred_cond  == all_conditions).mean(),
        'per_class_acc':  per_class_acc,
        'per_cond_acc':   per_cond_acc,
        'all_labels':     all_labels,
        'all_conditions': all_conditions,
        'all_pred_class': all_pred_class,
        'all_pred_cond':  all_pred_cond,
        'all_confidences': all_confidences,
    }
```

```python
# ------------------------------------------------------------
# SECTION 9: TRAINING LOOP — 5 EPOCHS
# ------------------------------------------------------------

print("="*55)
print("STARTING TRAINING")
print(f"Epochs: {NUM_EPOCHS} | Device: {DEVICE}")
print("="*55)

# History containers for plotting
history = {
    'train_total_loss': [],
    'train_class_loss': [],
    'train_cond_loss':  [],
    'train_class_acc':  [],
    'val_total_loss':   [],
    'val_class_loss':   [],
    'val_cond_loss':    [],
    'val_class_acc':    [],
    'val_per_class_acc': [],   # list of dicts
    'val_per_cond_acc':  [],   # list of dicts
}

best_val_acc   = 0.0
best_epoch     = 0
best_val_metrics = None

for epoch in range(1, NUM_EPOCHS + 1):
    print(f"\n--- Epoch {epoch}/{NUM_EPOCHS} ---")

    # Training
    train_metrics = train_one_epoch(model, train_loader, optimizer, DEVICE)

    # Validation
    val_metrics = validate(model, val_loader, DEVICE)

    # LR step
    scheduler.step()

    # Record history
    history['train_total_loss'].append(train_metrics['total_loss'])
    history['train_class_loss'].append(train_metrics['class_loss'])
    history['train_cond_loss'].append(train_metrics['cond_loss'])
    history['train_class_acc'].append(train_metrics['class_acc'])
    history['val_total_loss'].append(val_metrics['total_loss'])
    history['val_class_loss'].append(val_metrics['class_loss'])
    history['val_cond_loss'].append(val_metrics['cond_loss'])
    history['val_class_acc'].append(val_metrics['class_acc'])
    history['val_per_class_acc'].append(val_metrics['per_class_acc'])
    history['val_per_cond_acc'].append(val_metrics['per_cond_acc'])

    # Track best model
    if val_metrics['class_acc'] > best_val_acc:
        best_val_acc     = val_metrics['class_acc']
        best_epoch       = epoch
        best_val_metrics = val_metrics
        # Save best model weights in memory
        best_model_state = {k: v.clone()
                            for k, v in model.state_dict().items()}

    print(f"  Train Loss: {train_metrics['total_loss']:.4f} "
          f"(class={train_metrics['class_loss']:.4f}, "
          f"cond={train_metrics['cond_loss']:.4f})")
    print(f"  Train Acc:  {train_metrics['class_acc']*100:.2f}%")
    print(f"  Val   Loss: {val_metrics['total_loss']:.4f}")
    print(f"  Val   Acc:  {val_metrics['class_acc']*100:.2f}%")
    print(f"  Per-class val acc: " +
          ", ".join([f"{CLASSES[c]}:{v*100:.1f}%"
                     for c, v in val_metrics['per_class_acc'].items()]))
    print(f"  Per-condition val acc: " +
          ", ".join([f"{CONDITIONS[c]}:{v*100:.1f}%"
                     for c, v in val_metrics['per_cond_acc'].items()]))

print(f"\n[INFO] Best validation accuracy: {best_val_acc*100:.2f}% at epoch {best_epoch}")

# Restore best model weights for evaluation
model.load_state_dict(best_model_state)
```

```python
# ------------------------------------------------------------
# SECTION 10: EVALUATION AND PREDICTION PIPELINE
# ------------------------------------------------------------

print("\n[INFO] Running final evaluation on validation set...")
final_metrics = validate(model, val_loader, DEVICE)

# Compute confidence statistics
all_labels      = final_metrics['all_labels']
all_pred_class  = final_metrics['all_pred_class']
all_confidences = final_metrics['all_confidences']

correct_mask   = (all_pred_class == all_labels)
mean_conf_correct   = all_confidences[correct_mask].mean()   if correct_mask.sum() > 0 else 0
mean_conf_incorrect = all_confidences[~correct_mask].mean()  if (~correct_mask).sum() > 0 else 0

# Best and worst performing classes
per_class_accs = final_metrics['per_class_acc']
sorted_classes = sorted(per_class_accs.items(), key=lambda x: x[1], reverse=True)
best_classes  = [CLASSES[c] for c, _ in sorted_classes[:2]]
worst_classes = [CLASSES[c] for c, _ in sorted_classes[-2:]]

# Confusion matrix
conf_matrix = confusion_matrix(all_labels, all_pred_class)
conf_matrix_norm = conf_matrix.astype(float) / (conf_matrix.sum(axis=1,
                   keepdims=True) + 1e-8)

# Sample predictions (first 8)
print("\n[INFO] Sample Predictions:")
print(f"{'Idx':>4} | {'True Class':>12} | {'Pred Class':>12} | "
      f"{'True Cond':>16} | {'Pred Cond':>16} | {'Confidence':>10}")
print("-" * 80)
n_samples = min(8, len(all_labels))
for i in range(n_samples):
    tc = CLASSES[final_metrics['all_labels'][i]]
    pc = CLASSES[final_metrics['all_pred_class'][i]]
    tco = CONDITIONS[final_metrics['all_conditions'][i]]
    pco = CONDITIONS[final_metrics['all_pred_cond'][i]]
    conf = final_metrics['all_confidences'][i]
    print(f"{i:>4} | {tc:>12} | {pc:>12} | {tco:>16} | {pco:>16} | {conf:>10.4f}")
```

## Panel A — Total Loss Curves

### Overview
Panel A tracks the total combined loss (classification loss + weighted
condition loss) across all five training epochs for both the training
and validation sets. It provides the primary signal for assessing
convergence behavior and generalization capacity of the AV fusion model.

### Key Findings
- Both training and validation loss decrease monotonically across all
  five epochs, indicating stable and consistent optimization.
- Training loss begins at approximately 1.80 and converges near 0.20
  by epoch 5.
- Validation loss begins lower (~1.55) and converges to approximately
  0.05 by epoch 5 — below the training loss throughout.
- The gap between train and validation loss narrows significantly after
  epoch 2, with near-convergence by epoch 4.

### Discussion
The consistent downward trajectory of both curves without divergence
confirms that the Adam optimizer with gradient clipping produced stable
learning dynamics across the five-epoch budget. The observation that
validation loss tracks below training loss is attributable to the
condition-assignment strategy: validation samples receive deterministic
round-robin condition assignment, whereas training samples receive
random condition assignment with augmentation (noise injection and
masking), making the training objective systematically harder than
evaluation. This is a deliberate and appropriate asymmetry in an
educational simulation context. The absence of a generalization gap
(validation loss never exceeds training loss) indicates that the
synthetic visual templates, being perfectly class-discriminative by
construction, provide strong regularizing signal across all four
experimental conditions.

### Limitations
The five-epoch training budget is too brief to distinguish genuine
convergence from a fortuitously smooth loss landscape arising from
near-deterministic visual templates. The fact that validation loss
falls below training loss is partially an artifact of the augmentation
asymmetry rather than evidence of strong generalization to unseen data.
Extended training with held-out speakers and naturalistic visual inputs
would be required to make a more confident generalization claim.

## Panel B — Component Losses (Class + Condition Objectives)

### Overview
Panel B decomposes the total loss into its two constituent objectives:
the primary speech classification loss and the auxiliary condition
identification loss, tracked separately for training and validation
across all five epochs. This panel tests whether both objectives
co-optimize without conflict.

### Key Findings
- All four loss curves (class train, class val, condition train,
  condition val) decrease monotonically and converge by epoch 5.
- Classification loss (both train and val) starts highest (~1.6 at
  epoch 1) and converges near 0.10–0.15 by epoch 5.
- Condition loss converges faster and to lower absolute values than
  classification loss, reaching near-zero by epoch 3.
- No loss divergence or oscillation is observed across any component
  or split.

### Discussion
The faster convergence of the condition loss relative to the
classification loss reflects the relative difficulty of the two
tasks. The condition identification task (distinguishing clear,
degraded, AV clear, and AV restored signals) is partially solvable
by detecting the presence or absence of noise and the magnitude of
the visual input, both of which are strongly correlated with
condition identity by construction. In contrast, the speech
classification task requires the model to extract discriminative
phonetic information from MFCC features despite variable signal
quality across conditions, which is inherently more challenging.
The absence of gradient conflict between the two objectives
suggests that the auxiliary condition loss (weighted at λ = 0.3)
provided beneficial multi-task regularization without interfering
with the primary classification objective — a finding consistent
with the multi-task learning literature.

### Limitations
The low condition loss may reflect task triviality rather than
genuine learned perceptual awareness: the model may detect noise
statistics directly rather than learning a meaningful representation
of signal quality. A proper ablation removing the condition head
and comparing final classification accuracy would be needed to
confirm the auxiliary objective's contribution.

## Panel C — Classification Accuracy

### Overview
Panel C reports epoch-wise training and validation accuracy for the
primary speech class prediction task, contextualized against the
random-chance baseline of 16.7% (1/6 classes).

### Key Findings
- Training accuracy rises from approximately 40% at epoch 1 to
  ~98% by epoch 5.
- Validation accuracy begins at ~62% at epoch 1, substantially
  above chance, and reaches approximately 100% by epoch 5.
- Validation accuracy exceeds training accuracy throughout all five
  epochs, consistent with the augmentation asymmetry described in
  Panel A.
- Both curves plateau near ceiling performance by epoch 4, suggesting
  near-complete task saturation within the five-epoch budget.

### Discussion
The high early validation accuracy (62% at epoch 1, versus 40%
training) further confirms that the fixed, class-specific visual
mouth templates dominate the classification signal. Since each
class maps deterministically to a unique visual template, and
validation conditions 2 and 3 (AV Clear and AV Restored) both
include this template, the visual encoder rapidly learns a
near-perfect lookup from template to class. This is a deliberate
design property of the educational simulation: it demonstrates
the paper's central claim that visual information provides a
strong, class-discriminative signal that can compensate for
degraded audio. In a real AV speech system, such clean visual
separability would not exist, and the accuracy trajectory would
be considerably slower and lower.

### Limitations
Near-ceiling accuracy on a six-class closed-vocabulary task with
deterministic visual inputs does not constitute evidence of
generalizable AV integration learning. The result is largely a
consequence of the synthetic visual modality design rather than
the model's ability to handle naturalistic variation in speaker
face appearance, lighting, or viewing angle.

## Panel D — Normalized Confusion Matrix

### Overview
Panel D presents the class-normalized confusion matrix computed
on the final validation set, where entry (i, j) represents the
fraction of true-class-i samples predicted as class j.

### Key Findings
- The diagonal is near-perfect for five of six classes (yes, no,
  up, left, right), each achieving 1.00 or near-1.00 normalized
  accuracy.
- The only measurable off-diagonal confusion occurs in the "down"
  row, with 1% of "down" samples misclassified as "no".
- No symmetric confusion is observed (i.e., "no" does not
  misclassify as "down"), suggesting a mildly asymmetric
  decision boundary.
- "right" shows a minor 2% confusion into "left", which is the
  only other non-negligible off-diagonal entry.

### Discussion
The near-diagonal confusion matrix confirms that the model has
learned highly separable class representations. The minor
confusions observed (down→no, right→left) are semantically and
articulatorily plausible: "down" and "no" share a closed or
minimally open mouth configuration, while "left" and "right"
share laterally asymmetric mouth shapes that differ only in the
direction of displacement. This alignment between the model's
confusion pattern and articulatory similarity in the synthetic
mouth templates provides indirect evidence that the visual encoder
is learning geometrically meaningful features from the template
images rather than memorizing pixel-level patterns. The result is
consistent with the paper's argument that visible articulatory
geometry carries phonetically meaningful information.

### Limitations
With only six classes and deterministic visual templates, near-
perfect confusion matrix diagonals are the expected outcome rather
than a meaningful finding. The few observed confusions may reflect
random noise in weight initialization rather than principled
representational similarity. A larger class vocabulary with
naturalistic visual variation would be required to draw stronger
conclusions about the model's articulatory similarity learning.

## Panel E — Per-Class Accuracy

### Overview
Panel E presents the final per-class validation accuracy as a bar
chart, with a dashed line indicating the class-mean accuracy.
This enables identification of best- and worst-performing classes.

### Key Findings
- Five of six classes (yes, no, up, left, right) achieve 100.0%
  per-class accuracy on the validation set.
- "down" is the single underperforming class at approximately
  98.5% accuracy, consistent with the confusion matrix.
- Mean per-class accuracy is reported at 99.5%.
- Performance variance across classes is extremely low.

### Discussion
The uniform near-ceiling performance across classes reflects the
highly discriminative nature of the synthetic visual templates.
The marginally lower accuracy for "down" is consistent with the
observation in Panel D that its mouth template shares some
geometric similarity with "no" (both involve relatively closed

standpoint, this result successfully illustrates the core AV
speech perception concept: when visible articulatory information
is available and class-discriminative, it enables near-perfect
phoneme identification even in the presence of degraded audio.
This mirrors the paper's reported visual gain effect, where
visible speech substantially improves identification accuracy.

### Limitations
The practical implication of 100% per-class accuracy is limited
by the synthetic nature of the visual modality. In real AV speech
tasks, per-class accuracy varies substantially across phoneme
classes, with bilabial consonants (e.g., /b/, /p/, /m/) showing
higher visual discriminability than fricatives or stops produced
deep in the vocal tract (e.g., /k/, /g/). This inter-class
variation in visual salience is a core finding of the AV speech
literature that the current synthetic design cannot reproduce.

## Panel F — Epoch-wise Per-Class Accuracy Heatmap

### Overview
Panel F provides a temporal view of per-class learning dynamics,
displaying validation accuracy for each of the six classes at
each of the five training epochs as a heatmap.

### Key Findings
- Yes, no, up, and left achieve 100% accuracy from epoch 1 onward,
  indicating immediate and complete learning for these classes.
- "down" shows slower convergence, beginning at approximately 0%
  at epoch 1, rising through 40% at epoch 2, and reaching near-
  ceiling by epochs 4–5.
- "right" displays a similar delayed trajectory, starting low and
  converging by epoch 3–4.
- By epoch 3, all six classes have reached or nearly reached
  ceiling performance.

### Discussion
The heterogeneous convergence rates across classes reveal an
important aspect of multi-class AV learning dynamics. Classes that
converge immediately from epoch 1 (yes, no, up, left) likely have
visual templates that are maximally distant from all other templates
in pixel space, enabling the visual encoder to discriminate them
with minimal weight updates. The delayed convergence of "down" and
"right" suggests their templates are initially more easily confused
with geometrically similar classes, requiring additional gradient
updates to establish separable internal representations. This
trajectory pattern is consistent with curriculum learning theory:
easy discriminations (large inter-template distances) are solved
first, with harder discriminations (small inter-template distances)
resolved progressively. The result also validates that five epochs
is a sufficient training budget for this simplified task.

### Limitations
The delayed convergence of "down" from epoch 1 is worth
investigating: if the model achieves 0% on "down" at epoch 1
but 100% on all others, this suggests a transient initialization-
dependent failure rather than a principled hard-to-learn class.
The learning trajectory could be sensitive to random seed and
weight initialization, and may not be reproducible without fixing
seeds explicitly.

## Panel G — Prediction Confidence Distribution

### Overview
Panel G shows the distribution of softmax confidence scores for
correct and incorrect predictions on the final validation set,
enabling assessment of model calibration and the relationship
between confidence and prediction correctness.

### Key Findings
- Correct predictions (n=398, mean confidence=0.942) are strongly
  concentrated at high confidence values (0.9–1.0).
- Incorrect predictions (n=2, mean confidence=0.523) show lower
  and more dispersed confidence scores.
- The distribution is highly bimodal: nearly all predictions are
  either very high confidence correct or moderate confidence
  incorrect.
- Only 2 incorrect predictions exist across the entire validation
  set of 400 samples.

### Discussion
The extreme right-skew of the correct confidence distribution
(mean 0.942, concentrated at 1.0) indicates that the model has
learned highly peaked posterior distributions for most inputs,
reflecting strong decision boundaries in the learned feature
space. This is the expected outcome when visual templates are
perfectly class-discriminative and the visual encoder is a
sufficiently expressive CNN. The lower mean confidence for
incorrect predictions (0.523 vs. 0.942) is a positive calibration
signal: the model is less certain when it errs, which is a
desirable property for safe deployment in clinical or assistive
technology contexts. This pattern mirrors the paper's discussion
of individual differences in AV integration: high-confidence
correct responses correspond to conditions where visual information
clearly supplements audio, while lower-confidence incorrect
responses correspond to ambiguous or conflicting perceptual signals.

### Limitations
With only two incorrect predictions in the validation set, the
incorrect confidence distribution statistics (mean=0.523) are
not statistically meaningful and should not be interpreted as
a reliable estimate of model uncertainty under failure conditions.
Confidence calibration analysis requires substantially larger
error counts, ideally assessed with Expected Calibration Error
(ECE) or reliability diagrams on a larger and more challenging
evaluation set.
