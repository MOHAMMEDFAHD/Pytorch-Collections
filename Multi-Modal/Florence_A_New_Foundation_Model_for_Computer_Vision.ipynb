# Florence: A New Foundation Model for Computer Vision

# https://arxiv.org/pdf/2111.11432

## Abstract

Florence introduces a general-purpose computer vision foundation model designed to support a wide spectrum of vision tasks across images, videos, object-level understanding, and vision-language applications. Unlike prior vision-language models such as CLIP and ALIGN, which primarily learn image-text alignment for classification and retrieval, Florence expands representation learning across three dimensions:

1. Space: from scene-level understanding to object-level understanding.
2. Time: from static images to dynamic videos.
3. Modality: from RGB images to multimodal reasoning involving language and depth.

The model is trained on a large-scale dataset of 900 million image-text pairs and demonstrates state-of-the-art performance across classification, retrieval, object detection, visual question answering, image captioning, and video understanding tasks.

---

# Problems

Existing vision foundation models suffer from several limitations:

### 1. Limited Task Coverage

Models such as CLIP and ALIGN are primarily optimized for:

- Image classification
- Image-text retrieval

They do not naturally support:

- Object detection
- Dense prediction
- Video understanding
- Fine-grained vision-language reasoning

### 2. Lack of Object-Level Representations

Most image-text pretraining approaches learn global image representations but fail to capture detailed object-level information required for:

- Detection
- Segmentation
- Localization

### 3. Weak Support for Temporal Understanding

Traditional vision-language models operate on images only and cannot effectively process videos or temporal relationships.

### 4. Fragmented Vision Systems

Different tasks often require separate architectures and training pipelines, making deployment expensive and inefficient.

---

# Proposed Solutions

Florence proposes a unified computer vision foundation model consisting of:

### 1. Large-Scale Vision-Language Pretraining

Training on:

- 900 million image-text pairs (FLD-900M)

to learn universal visual representations.

### 2. Unified Contrastive Learning (UniCL)

Instead of treating every text description as unique, Florence groups identical descriptions and treats all associated images as positive examples.

Training objective:

$$
L = L_{i2t} + L_{t2i}
$$

where:

- \(L_{i2t}\): image-to-text contrastive loss
- \(L_{t2i}\): text-to-image contrastive loss

This combines:

- Supervised learning
- Contrastive learning

within a unified framework.

### 3. Hierarchical Vision Transformer

Florence uses:

- CoSwin Transformer

which extends Swin Transformer using convolutional embedding layers for improved scalability and dense prediction capabilities.

### 4. Specialized Adaptation Modules

#### Object-Level Adapter

Dynamic Head adapter:

- Extends scene representations to object representations.
- Enables strong object detection performance.

#### Vision-Language Adapter

METER adapter:

- Enables fine-grained multimodal reasoning.
- Supports VQA and image captioning.

#### Video Adapter

Video CoSwin:

- Converts image representations into video representations.
- Reuses pretrained image weights efficiently.

- Generalize across diverse vision tasks.

# Mathematical and Statistical Content in *Florence: A New Foundation Model for Computer Vision*

## Overview

The mathematical content of the paper is concentrated in the **Unified Image-Text Contrastive Learning (UniCL)** framework, representation learning, similarity computation, optimization methods, and evaluation metrics.

Unlike many theoretical papers, Florence is primarily an engineering and large-scale empirical work. Therefore, most mathematics appears in the training objective and statistical evaluation methodology rather than in new mathematical theory.

---

# 1. Feature Normalization

Before computing similarities, image and text representations are normalized.

Given:

- Image encoder:

$$
f_\theta(x)
$$

- Text encoder:

$$
f_\phi(t)
$$

The normalized feature vectors are:

$$
u=\frac{f_\theta(x)}{\|f_\theta(x)\|}
$$

$$
v=\frac{f_\phi(t)}{\|f_\phi(t)\|}
$$

### Meaning

This converts feature vectors into unit vectors.

### Purpose

Normalization ensures that:

- Only direction matters.
- Vector magnitude does not influence similarity.

This allows cosine similarity to become a simple dot product.

---

# 2. Unified Contrastive Learning (UniCL)

The main mathematical contribution of the paper.

The total training loss is:

$$
L = L_{i2t} + L_{t2i}
$$

where:

- \(L_{i2t}\) = image-to-text loss
- \(L_{t2i}\) = text-to-image loss

### Meaning

The model learns in two directions simultaneously:

1. Match images to correct text.
2. Match text to correct images.

### Purpose

Creates a shared representation space where related images and texts become close together.

---

# 3. Image-to-Text Contrastive Loss

The paper defines:

$$
L_{i2t}
=
-
\sum_{i\in B}
\frac{1}{|P(i)|}
\sum_{k\in P(i)}
\log
\frac{
\exp(\tau u_i v_k)
}
{
\sum_{j\in B}
\exp(\tau u_i v_j)
}
$$

### Components

#### Batch

$$
B
$$

A mini-batch of training examples.

#### Positive Set

$$
P(i)
=
\{k \mid y_k=y_i\}
$$

Images sharing the same text label are treated as positive examples.

#### Temperature Parameter

$$
\tau
$$

A learnable scaling factor.

---

### Meaning

The model is rewarded when:

- Correct image-text pairs have high similarity.

The model is penalized when:

- Incorrect image-text pairs appear similar.

---

### Purpose

Learns semantic alignment between images and language.

---

# 4. Text-to-Image Contrastive Loss

The reverse objective:

$$
L_{t2i}
=
-
\sum_{j\in B}
\frac{1}{|Q(j)|}
\sum_{k\in Q(j)}
\log
\frac{
\exp(\tau u_k v_j)
}
{
\sum_{i\in B}
\exp(\tau u_i v_j)
}
$$

where

$$
Q(j)
=
\{k \mid y_k=y_j\}
$$

---

### Meaning

For each text:

- Correct images should receive higher scores.
- Incorrect images should receive lower scores.

---

### Purpose

Creates symmetry in learning and improves retrieval quality.

---

# 5. Exponential Similarity Function

The loss uses:

$$
\exp(\tau u_i v_j)
$$

### Meaning

Transforms similarity scores into positive values.

### Purpose

Acts like a probability distribution over possible matches.

This is the same principle used in:

- Softmax classifiers
- InfoNCE contrastive learning
- CLIP training

---

# 6. Cosine Similarity

For retrieval and zero-shot classification, Florence computes similarity between image and text embeddings.

Conceptually:

$$
\text{Similarity}
=
u^T v
$$

Since vectors are normalized:

$$
u^T v
=
\cos(\theta)
$$

---

### Meaning

Measures semantic closeness between:

- Image representation
- Text representation

---

### Purpose

Used for:

- Zero-shot classification
- Image retrieval
- Text retrieval
- Open-vocabulary recognition

---

# 7. Hash-Based Label Grouping

A statistical idea introduced by UniCL.

Instead of assuming every caption is unique:

$$
(x,t,y)
$$

where:

- \(x\) = image
- \(t\) = text description
- \(y\) = hash label

Images sharing identical descriptions receive the same label.

---

### Meaning

Multiple images can correspond to the same concept.

### Purpose

Reduces false negatives during contrastive learning.

This is one of the main improvements over CLIP-style training.

---

# 8. Data Augmentation Through Prompt Templates

Short labels are expanded into sentences such as:

> "A photo of a dog"

instead of:

> "dog"

---

### Statistical Purpose

Improves:

- Linguistic diversity
- Generalization
- Robustness

without collecting additional data.

---

# 11. Evaluation Metrics

The paper uses numerous statistical performance metrics.

---

## Classification Accuracy

### Top-1 Accuracy

$$
\text{Top-1}
=
\frac{\text{Correct Predictions}}
{\text{Total Samples}}
$$

### Meaning

Prediction is correct only if the highest-ranked class is correct.

---

## Top-5 Accuracy

$$
\text{Top-5}
=
\frac{
\text{Ground Truth Appears in Top 5}
}
{
\text{Total Samples}
}
$$

### Meaning

Measures ranking quality.

---

# 12. Recall@K (Retrieval)

Used for image-text retrieval.

### Recall@1

$$
R@1
$$

Correct result appears first.

### Recall@5

$$
R@5
$$

Correct result appears within top five.

---

### Purpose

Measures retrieval effectiveness.

Higher values indicate better semantic alignment.

---

# 13. Mean Average Precision (mAP)

Used in object detection.

### Definition

Average precision is computed over different recall levels.

Mean Average Precision:

$$
mAP
=
\frac{1}{N}
\sum_{i=1}^{N}
AP_i
$$

---

### Meaning

Measures both:

- Detection accuracy
- Localization quality

---

### Purpose

Primary evaluation metric for:

- COCO
- Object365

---

# 14. AP50

Used in Visual Genome evaluation.

A detection is considered correct when:

$$
IoU \ge 0.5
$$

where

$$
IoU
=
\frac{
\text{Intersection}
}
{
\text{Union}
}
$$

---

### Meaning

Measures overlap between:

- Predicted box
- Ground-truth box

---

### Purpose

Evaluates localization quality.

---

# 15. Few-Shot Learning Statistics

Experiments use:

- 5-shot
- 20-shot
- 50-shot

settings.

### Meaning

Only a small number of labeled examples per class are available.

### Purpose

Tests adaptation ability under limited supervision.

---

# 16. Zero-Shot Learning

No task-specific training examples are provided.

Prediction is performed through:

$$
\arg\max
\left(
u^T v
\right)
$$

choosing the text label with highest similarity.

---

### Purpose

Measures true transferability of the learned representation.

   $$

3. **Large-scale statistical learning**

```python
# Florence-Inspired Educational Implementation on FashionMNIST

## Important Note

# The original Florence model cannot realistically be reproduced in a notebook because it requires:

# - 900M image-text pairs
# - 893M parameters
# - 512 A100 GPUs
# - Multi-modal image-text training
# - Object detection adapters
# - Video adapters
# - Vision-language adapters

# Instead, this educational lab reproduces the **core Florence idea**:

# > Learn a shared image-text embedding space using contrastive learning.

# We will:

# 1. Load FashionMNIST.
# 2. Convert class labels into text descriptions.
# 3. Build:
#    - Image Encoder (CNN)
#    - Text Encoder (Embedding + GRU)
# 4. Learn a shared embedding space.
# 5. Train with CLIP/Florence-style contrastive learning.
# 6. Evaluate retrieval accuracy.
# 7. Visualize:
#    - Training loss
#    - Retrieval accuracy
#    - Embedding space (t-SNE)
#    - Sample predictions
#    - Similarity heatmaps

# Everything is contained in a **single code cell**.

# ============================================================
# FLORENCE-STYLE IMAGE-TEXT FOUNDATION MODEL
# EDUCATIONAL IMPLEMENTATION USING FASHION-MNIST
# ============================================================

# ------------------------------------------------------------
# Imports
# ------------------------------------------------------------

import torch
import torch.nn as nn
import torch.nn.functional as F

import torchvision
import torchvision.transforms as transforms

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.manifold import TSNE
from sklearn.metrics import accuracy_score

from torch.utils.data import DataLoader

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print("Device:", device)
```

```python
# ============================================================
# FASHION-MNIST LABELS AS TEXT
# ============================================================

class_names = [
    "T-shirt",
    "Trouser",
    "Pullover",
    "Dress",
    "Coat",
    "Sandal",
    "Shirt",
    "Sneaker",
    "Bag",
    "Ankle boot"
]

text_descriptions = [
    "a photo of a t shirt",
    "a photo of a trouser",
    "a photo of a pullover",
    "a photo of a dress",
    "a photo of a coat",
    "a photo of a sandal",
    "a photo of a shirt",
    "a photo of a sneaker",
    "a photo of a bag",
    "a photo of an ankle boot"
]
```

```python
# ============================================================
# TOKENIZER
# ============================================================

vocab = {}

for sentence in text_descriptions:
    for word in sentence.lower().split():
        if word not in vocab:
            vocab[word] = len(vocab) + 1

vocab_size = len(vocab) + 1

def tokenize(text):
    return [vocab[w] for w in text.lower().split()]

tokenized_texts = [tokenize(t) for t in text_descriptions]

max_len = max(len(x) for x in tokenized_texts)

text_tensor = []

for tokens in tokenized_texts:
    padded = tokens + [0]*(max_len-len(tokens))
    text_tensor.append(padded)

text_tensor = torch.tensor(text_tensor)
```

```python
# ============================================================
# DATASET
# ============================================================

transform = transforms.Compose([
    transforms.ToTensor()
])

train_dataset = torchvision.datasets.FashionMNIST(
    root="./data",
    train=True,
    download=True,
    transform=transform
)

test_dataset = torchvision.datasets.FashionMNIST(
    root="./data",
    train=False,
    download=True,
    transform=transform
)

train_loader = DataLoader(
    train_dataset,
    batch_size=256,
    shuffle=True
)

test_loader = DataLoader(
    test_dataset,
    batch_size=256,
    shuffle=False
)
```

```python
# ============================================================
# IMAGE ENCODER
# Florence uses a large Vision Transformer.
# For education we use a simple CNN.
# ============================================================

class ImageEncoder(nn.Module):

    def __init__(self, embedding_dim=128):
        super().__init__()

        self.features = nn.Sequential(

            nn.Conv2d(1,32,3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(32,64,3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(64,128,3,padding=1),
            nn.ReLU(),

            nn.AdaptiveAvgPool2d((1,1))
        )

        self.fc = nn.Linear(128,embedding_dim)

    def forward(self,x):

        x = self.features(x)

        x = x.view(x.size(0),-1)

        x = self.fc(x)

        x = F.normalize(x,dim=1)

        return x
```

```python
# ============================================================
# TEXT ENCODER
# Florence uses a Transformer.
# For education we use Embedding + GRU.
# ============================================================

class TextEncoder(nn.Module):

    def __init__(self,vocab_size,embedding_dim=128):

        super().__init__()

        self.embedding = nn.Embedding(
            vocab_size,
            64,
            padding_idx=0
        )

        self.gru = nn.GRU(
            64,
            embedding_dim,
            batch_first=True
        )

    def forward(self,x):

        x = self.embedding(x)

        _,h = self.gru(x)

        h = h.squeeze(0)

        h = F.normalize(h,dim=1)

        return h
```

```python
# ============================================================
# FLORENCE-LIKE MODEL
# Shared image-text embedding space
# ============================================================

class FlorenceMini(nn.Module):

    def __init__(self,vocab_size):

        super().__init__()

        self.image_encoder = ImageEncoder()

        self.text_encoder = TextEncoder(vocab_size)

        self.temperature = nn.Parameter(
            torch.tensor(0.07)
        )

    def forward(self,images,texts):

        image_features = self.image_encoder(images)

        text_features = self.text_encoder(texts)

        logits = (
            image_features @ text_features.T
        ) / self.temperature.exp()

        return logits


model = FlorenceMini(vocab_size).to(device)
```

```python
# ============================================================
# LOSS
# Florence uses bidirectional contrastive learning.
# ============================================================

optimizer = torch.optim.Adam(
    model.parameters(),
    lr=1e-3
)

epochs = 10

train_losses = []
retrieval_accs = []
```

```python
# ============================================================
# TRAINING
# ============================================================

for epoch in range(epochs):

    model.train()

    running_loss = 0

    for images,labels in train_loader:

        images = images.to(device)

        texts = text_tensor[labels].to(device)

        logits = model(images,texts)

        targets = torch.arange(
            len(images)
        ).to(device)

        image_loss = F.cross_entropy(
            logits,
            targets
        )

        text_loss = F.cross_entropy(
            logits.T,
            targets
        )

        loss = (
            image_loss + text_loss
        ) / 2

        optimizer.zero_grad()

        loss.backward()

        optimizer.step()

        running_loss += loss.item()

    print(f"Loss: {running_loss}")

    avg_loss = running_loss / len(train_loader)

    train_losses.append(avg_loss)
```

```python
# --------------------------------------------------------
# Retrieval Accuracy
# --------------------------------------------------------

model.eval()

correct = 0
total = 0

with torch.no_grad():

    all_text_features = model.text_encoder(
        text_tensor.to(device)
    )

    for images,labels in test_loader:

        images = images.to(device)

        labels = labels.to(device)

        image_features = (
            model.image_encoder(images)
        )

        similarities = (
            image_features @
            all_text_features.T
        )

        preds = similarities.argmax(1)

        correct += (
            preds == labels
        ).sum().item()

        total += len(labels)

acc = correct/total

retrieval_accs.append(acc)

print(
    f"Epoch {epoch+1}/{epochs}"
    f" | Loss={avg_loss:.4f}"
    f" | Retrieval Acc={acc:.4f}"
)
```

```python
# ============================================================
# VISUALIZATION 2
# RETRIEVAL ACCURACY
# ============================================================

plt.figure(figsize=(6,4))
plt.plot(retrieval_accs,marker="o")
plt.title("Retrieval Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.grid()
plt.show()
```
