# Neural networks and physical systems with emergent collective computational abilities

# https://pmc.ncbi.nlm.nih.gov/articles/PMC346238/
# Abstract

This paper introduces a neural network model demonstrating that computational abilities, such as associative memory and categorization, can emerge from the collective dynamics of many simple binary neurons. The model shows how networks can store and recall patterns robustly, using asynchronous parallel updating, and suggests applications in both neuroscience and fault-tolerant hardware design.  
*(Hopfield, 1982, "Neural networks and physical systems with emergent collective computational abilities")*

---

# Problems

- Understanding whether large collections of simple neurons can spontaneously exhibit computational capabilities beyond those of small circuits.  
- Addressing the limitations of previous models like perceptrons, which required strict synchrony and lacked robustness in recurrent connections.  
- Bridging the gap between biological plausibility and computational utility, particularly in memory, categorization, and error correction.  
*(Hopfield, 1982)*

---

# Proposed Solutions

Hopfield proposed a recurrent neural network with symmetric (and later asymmetric) connections, where neurons update asynchronously. The system is governed by an **energy function** similar to physical spin systems, leading to stable attractor states that serve as stored memories. Hebbian learning rules enable memory storage and retrieval.  

Energy function:  

$$
E = -\frac{1}{2}\sum_{i \neq j} w_{ij} s_i s_j + \sum_i \theta_i s_i
$$  

*(Hopfield, 1982)*

---

# Purpose

The purpose was to demonstrate that important computational functions—such as content-addressable memory, generalization, and familiarity recognition—could emerge as collective phenomena in neural-like systems, independent of detailed biological mechanisms.  
*(Hopfield, 1982)*

---

# Methodology

- Defined neurons as binary units with thresholds, updating asynchronously.  
- Used **Hebbian learning** to encode memory states into the weight matrix:  

  $$
  w_{ij} = \frac{1}{N} \sum_{\mu=1}^P \xi_i^\mu \xi_j^\mu
  $$  

- Analyzed system dynamics via an energy function analogous to the Ising model.  
- Performed Monte Carlo simulations with networks of 30–100 neurons to study stability, memory capacity, cycles, and error behavior.  
- Investigated robustness under noisy inputs, failed connections, and overload conditions.  
*(Hopfield, 1982)*

---

# Results

- Networks converge to stable attractor states that represent stored memories.  
- Partial or noisy inputs are completed into full patterns, demonstrating **associative memory**.  
- The capacity is approximately **0.15N** patterns for **N neurons** before recall becomes unreliable.  
- The model supports categorization, error correction, and recognition of novel inputs.  
- Cycles and chaotic wandering can occur, but most dynamics stabilize to memory states.  
- The system exhibits **“soft failure”**: performance degrades gradually under overload or connection failures.  
*(Hopfield, 1982)*

---

# Conclusions

Hopfield’s model showed that collective dynamics of simple neurons naturally yield robust computational properties. These findings provided a **unifying framework** connecting neural computation, statistical physics, and parallel processing. The work laid the foundation for **energy-based models** and established the Hopfield network as a fundamental architecture for understanding memory and emergent computation in neural systems.  
*(Hopfield, 1982)*


# Mathematical Foundations of the Hopfield Network (1982)

## 1. Binary Neuron Model

Each neuron \(V_i\) has two possible states:

$$
V_i \in \{0,1\} \quad \text{or equivalently} \quad V_i \in \{-1,+1\}.
$$

- \(V_i = 1\): neuron firing.  
- \(V_i = 0\): neuron not firing.  

The state of the system is a binary vector of length \(N\).  

**Role:** Simple, abstract representation of neurons suitable for analysis.

---

## 2. Connection Weights (Synapses)

Each connection between neuron \(i\) and \(j\) has strength \(T_{ij}\):

$$
T_{ij} =
\begin{cases}
0 & \text{if no connection exists}, \\
0 & \text{if } i=j.
\end{cases}
$$

**Role:** Encodes how neurons influence each other.

---

## 3. Update Rule (Dynamics)

Each neuron checks its input sum:

$$
h_i = \sum_j T_{ij} V_j,
$$

and flips to \(+1\) or \(-1\) depending on whether \(h_i\) crosses threshold \(U_i\).  
Updating is **asynchronous and stochastic**.  

**Role:** Defines the time evolution and convergence.

---

## 4. Hebbian Storage Rule

To store memory states \(V^s, \; s=1,\dots,n\):

$$
T_{ij} = \sum_{s=1}^n (2V_i^s - 1)(2V_j^s - 1), \quad T_{ii} = 0.
$$

**Role:** Implements Hebb’s rule — *neurons that fire together, wire together*.

---

## 5. Energy Function (Lyapunov Function)

For symmetric weights \(T_{ij} = T_{ji}\):

$$
E = -\frac{1}{2}\sum_{i \neq j} T_{ij} V_i V_j.
$$

When a neuron updates:

$$
\Delta E = -\Delta V_i \sum_j T_{ij} V_j.
$$

**Role:** Ensures system moves toward lower energy and stabilizes at attractors.

---

## 6. Relation to Statistical Physics

- Analogy to **Ising spin systems** and **spin glasses**.  
- Stable memories correspond to **energy minima**.  

**Role:** Connects neural computation to physics.

---

## 7. Memory Capacity

Storage limit:

$$
n_{\text{max}} \approx 0.15N.
$$

Noise term from overlapping patterns:

$$
\text{rms noise} \sim \sqrt{\frac{n-1}{N}}.
$$

**Role:** Quantifies maximum reliable storage.

---

## 8. Error Probability

For large \(n, N\), noise is Gaussian. Probability of bit error:

$$
P = \frac{1}{\sqrt{2\pi}\sigma} \int_{N/2}^\infty e^{-\frac{x^2}{2\sigma^2}} \, dx,
$$

where \(\sigma^2\) is noise variance.  

**Role:** Provides statistical estimate of recall reliability.

---

## 9. Hamming Distance

Distance between binary states \(A, B\):

$$
d(A,B) = \# \{ i : A_i \neq B_i \}.
$$

**Role:** Measures similarity of input to stored patterns.

---

## 10. Entropy of Chaotic Wandering

Entropy of visited states:

$$
M = -\sum_i p_i \ln p_i,
$$

where \(p_i\) is probability of visiting state \(i\).  

**Role:** Quantifies randomness of chaotic wandering.

---

## 11. Categorization and Generalization

- Mixed/partial inputs converge to nearest attractor.  
- Attractors dominate state space.  

**Role:** Explains categorization and error correction.

---

## 12. Sequence Memory

With asymmetric weights:

$$
\Delta T_{ij} = \sum_s (2V_i^{s+1} - 1)(2V_j^s - 1).
$$

**Role:** Enables recall of short ordered sequences.

---

# Summary

Hopfield’s 1982 model is grounded in:

- Binary states and Hebbian weights.  
- An energy function ensuring convergence.  
- Physics analogies (Ising model, spin glasses).  
- Capacity analysis (\(\sim 0.15N\) patterns).  
- Error probability and entropy estimates.  
- Tools like Hamming distance for pattern closeness.  
- Categorization, generalization, and limited sequential recall.

**Conclusion:** Stable collective computation — memory recall, error correction, categorization — emerges naturally from recurrent stochastic binary neurons.

````python
import torch, numpy as np, matplotlib.pyplot as plt

# --- Hopfield Network Class ---
class HopfieldNet:
    def __init__(self, n_units: int):
        self.n_units = n_units
        self.W = torch.zeros((n_units, n_units), dtype=torch.float32)

    def store_patterns(self, patterns: torch.Tensor):
        self.W = torch.zeros((self.n_units, self.n_units), dtype=torch.float32)
        for p in patterns:
            self.W += torch.outer(p, p)
        self.W.fill_diagonal_(0)
        self.W /= patterns.shape[0]

    def recall(self, state: torch.Tensor, steps=10, synchronous=False):
        s, history = state.clone(), [state.clone()]
        for _ in range(steps):
            if synchronous:
                s = torch.sign(self.W @ s)
            else:
                for i in torch.randperm(self.n_units):
                    raw = torch.dot(self.W[i], s)
                    s[i] = 1.0 if raw >= 0 else -1.0
            history.append(s.clone())
        return torch.stack(history)

# --- Helpers ---
def generate_patterns(n_units, n_patterns, seed=42):
    rng = torch.Generator().manual_seed(seed)
    return (torch.randint(0, 2, (n_patterns, n_units), generator=rng)*2 - 1).float()

def plot_pattern(vec, title="", ax=None):
    if ax is None: ax = plt.gca()
    size = int(np.sqrt(len(vec)))
    ax.imshow(vec.reshape(size, size), cmap="coolwarm", interpolation="nearest")
    ax.set_title(title); ax.axis("off")

def visualize_recall(hopfield, pattern, noise_level=0.3, steps=6):
    noisy = pattern.clone()
    flip_idx = torch.randperm(len(pattern))[:int(noise_level*len(pattern))]
    noisy[flip_idx] *= -1
    traj = hopfield.recall(noisy, steps=steps, synchronous=False)
    fig, axes = plt.subplots(1, len(traj), figsize=(15, 3))
    for i, state in enumerate(traj): plot_pattern(state.numpy(), f"Step {i}", axes[i])
    plt.show()
    return traj

def energy(W, state): return -0.5 * state @ (W @ state)

def detect_cycle(traj):
    arr = traj.numpy()
    for i in range(len(arr)-1):
        for j in range(i+1, len(arr)):
            if np.array_equal(arr[i], arr[j]): return i, j
    return None

# --- Run Example ---
n_units, n_patterns = 64, 2
patterns = generate_patterns(n_units, n_patterns)
hopfield = HopfieldNet(n_units); hopfield.store_patterns(patterns)

# Recall with noise
for i, p in enumerate(patterns):
    traj = visualize_recall(hopfield, p, noise_level=0.35, steps=6)

# Energy plot
energies = [energy(hopfield.W, s) for s in traj]
plt.plot(energies, marker="o"); plt.title("Energy Landscape"); plt.xlabel("Step"); plt.ylabel("Energy"); plt.show()

# Cycle detection
rand_state = (torch.randint(0, 2, (n_units,), dtype=torch.float32) * 2 - 1)
traj = hopfield.recall(rand_state, steps=30)
cycle = detect_cycle(traj)
print("Cycle detected:" , cycle if cycle else "None")
````

````python
import torch, numpy as np, matplotlib.pyplot as plt
from torchvision import datasets, transforms

# --- Hopfield Network Class ---
class HopfieldNet:
    def __init__(self, n_units: int):
        self.n_units = n_units
        self.W = torch.zeros((n_units, n_units), dtype=torch.float32)

    def store_patterns(self, patterns: torch.Tensor):
        self.W = torch.zeros((self.n_units, self.n_units), dtype=torch.float32)
        for p in patterns:
            self.W += torch.outer(p, p)
        self.W.fill_diagonal_(0)
        self.W /= patterns.shape[0]

    def recall(self, state: torch.Tensor, steps=10, synchronous=False):
        s, history = state.clone(), [state.clone()]
        for _ in range(steps):
            if synchronous:
                s = torch.sign(self.W @ s)
            else:
                for i in torch.randperm(self.n_units):
                    raw = torch.dot(self.W[i], s)
                    s[i] = 1.0 if raw >= 0 else -1.0
            history.append(s.clone())
        return torch.stack(history)

# --- Visualization Helpers ---
def plot_pattern(vec, title="", ax=None):
    if ax is None: ax = plt.gca()
    size = int(np.sqrt(len(vec)))
    ax.imshow(vec.reshape(size, size), cmap="gray", interpolation="nearest")
    ax.set_title(title); ax.axis("off")

def visualize_digit_recall(hopfield, pattern, digit_label, noise_level=0.3, steps=6):
    noisy = pattern.clone()
    flip_idx = torch.randperm(len(pattern))[:int(noise_level*len(pattern))]
    noisy[flip_idx] *= -1
    traj = hopfield.recall(noisy, steps=steps, synchronous=False)
    fig, axes = plt.subplots(1, len(traj), figsize=(15, 3))
    for i, state in enumerate(traj):
        plot_pattern(state.numpy(), f"{digit_label}: Step {i}", axes[i])
    plt.show()
    return traj

# --- Load MNIST and Prepare Patterns ---
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Lambda(lambda x: (x.view(-1) > 0.5).float()*2 - 1) # binarize to {-1,+1}
])
mnist = datasets.MNIST(root="./data", train=True, download=True, transform=transform)

# Select one representative image for each digit 0–9
digit_patterns = []
for d in range(10):
    for i in range(len(mnist)):
        if mnist[i][1] == d:   # check digit label
            digit_patterns.append(mnist[i][0])
            break
patterns = torch.stack(digit_patterns)

# Initialize Hopfield net
n_units = 28*28
hopfield = HopfieldNet(n_units)
hopfield.store_patterns(patterns)

# --- Test recall on all 10 digits ---
for digit_label, pattern in enumerate(patterns):
    traj = visualize_digit_recall(hopfield, pattern, digit_label, noise_level=0.4, steps=6)
````

# Reported Metrics of Hopfield Network Experiments

## Energy Values

### Small Synthetic (8×8)
- Initial energy: ≈ \(-200\).  
- Sharp decline: ≈ \(-1000\) within the first update.  
- Stabilized: ≈ \(-1000\) across Steps 1–6.  

### MNIST Digits (28×28, \(N=784\))
- Initial energy: ≈ \(-25{,}000\).  
- Sharp drop: ≈ \(-190{,}000\) to \(-200{,}000\) after the first recall step.  
- Stable: constant across later steps, confirming convergence.  

---

## Convergence Dynamics

- Recall trajectories show **steep initial energy decrease** followed by stability at a minimum.  
- Stabilization typically within **1–2 steps**; no further changes beyond Step 6.  

---

## Pattern Recall

- Noisy binary images evolve into recognizable attractors.  
- Even with heavy noise (\(30\%–40\%\) pixel flips), final states reconstruct clean stored digits (e.g., “3”).  
- Across digits 0–9, recall consistently converged to the correct stored pattern.  

---

## Interpretation in Context

### Energy Function Behavior
- Sharp energy decline reflects the theoretical guarantee: asynchronous updates always reduce or maintain Lyapunov energy.  
- Plateau indicates convergence to a **local minimum attractor**.  

### Associative Memory Function
- Reconstruction of corrupted digits demonstrates associative memory: noisy/partial inputs map to complete stored patterns.  

### Capacity vs. Performance
- With \(N=784\) neurons, theoretical capacity is:

$$
0.15N \approx 120 \quad \text{patterns}.
$$

- Storing 10 digit exemplars is well below capacity → explains strong recall performance.  

---

## Implications

- **Robust Recall:** Networks reliably denoise and retrieve stored patterns.  
- **Scalability:** Larger \(N\) leads to deeper minima (e.g., \(-1000\) vs. \(-200{,}000\)).  
- **Predictable Convergence:** Stabilization in a few steps ensures computational efficiency.  

---

## Limitations

- **Generalization:** Only specific stored exemplars are retrieved; unseen variations collapse to the nearest attractor.  
- **Pattern Interference:** With many patterns (>100), spurious attractors and overlaps reduce recall quality.  
- **Binary Encoding:** Binarization discards grayscale detail, limiting realism compared to modern deep networks.  

---

# Concise Academic Summary

The experiments confirm that Hopfield’s recurrent associative network exhibits **rapid Lyapunov energy convergence**, stabilizing within 1–2 steps for both small synthetic inputs and MNIST digits. Energy minima align with stored attractors, enabling robust reconstruction from noisy states. While the results validate theoretical principles of **associative memory**, limitations in capacity, generalization, and binary encoding underscore why Hopfield networks remain primarily of **historical and conceptual significance** relative to modern deep architectures.
