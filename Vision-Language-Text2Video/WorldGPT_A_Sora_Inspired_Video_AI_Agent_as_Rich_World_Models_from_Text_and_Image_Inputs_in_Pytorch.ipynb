# WorldGPT: A Sora-Inspired Video AI Agent as Rich World Models from Text and Image Inputs

# https://arxiv.org/abs/2403.07944

## Abstract
**WorldGPT** is a multimodal **text-to-video generation framework** inspired by OpenAI’s *Sora*. It integrates **large language models (LLMs)** with **diffusion-based video synthesis** to produce temporally coherent and semantically grounded videos from either text or image inputs.  
The system merges a **Prompt Enhancer** (LLM-driven semantic expansion) and a **Full Video Translation** pipeline (GroundingDINO + Stable Diffusion + DynamiCrafter) to construct “**world models**” — coherent temporal simulations where motion, perception, and meaning evolve together.

---

## Problems

| **Challenge** | **Description** |
|:--|:--|
| **Temporal Inconsistency** | Existing text-to-video systems exhibit frame flickering and poor motion continuity. |
| **Weak Semantic Alignment** | Generated videos often diverge visually from textual intent or image context. |
| **Limited Control and Creativity** | Current pipelines lack structured prompt control and narrative coherence. |
| **Closed-Source Constraints** | Proprietary systems (e.g., Sora) prevent open research replication and transparency. |

---

## Proposed Solutions

**WorldGPT** introduces an **agent-based multimodal architecture** with two primary components:

### 1. Prompt Enhancer  
- Uses **ChatGPT** to parse and expand user prompts into structured **sub-prompts**.  
- Extracts **keywords**, **spatial/temporal cues**, and **optimization directives** for better control.  
- Ensures high **semantic precision** and **narrative coherence** across video segments.

### 2. Full Video Translation Pipeline  
- **Key Frame Generator:**  
  - Employs **GroundingDINO** for text–object detection and **Stable Diffusion** for generating high-quality, semantically aligned key frames.  
- **Video Generator:**  
  - Uses **DynamiCrafter** to interpolate between key frames.  
  - Incorporates **LLM-guided contextual refinements** for maintaining consistency across time.

Together, these modules simulate **reasoning over both text semantics and temporal dynamics**, forming a **rich, evolving world model** from static or linguistic input.

The goal of **WorldGPT** is to design an **open, interpretable, and controllable text-to-video framework** that can:

### Mathematical Formulation

**(a) LLM Objective**  
Language modeling for structured prompt generation:
$$
L = -\sum_{i=1}^{n} \log P(\text{token}_i \mid \text{tokens}_{<i}, \theta)
$$

**(b) Diffusion Process**  
Generative noise modeling over time:
$$
x_t = \alpha_t x_{t-1} + (1 - \alpha_t)\epsilon
$$

**(c) Video Synthesis Loss**  
Motion interpolation objective across frames:
$$
\min_{F,B} \sum_t L(I_t, G(I, F_t, B_t))
$$

Where \( F_t \) and \( B_t \) represent forward and backward optical flows, ensuring **temporal coherence**.

# Mathematical and Statistical Content Summary — “WorldGPT: A Sora-Inspired Video AI Agent”

This paper integrates mathematical formulations from **large language modeling**, **diffusion processes**, and **temporal video synthesis** to form a unified framework for **text-to-video generation** and **AI world modeling**.  
Below is a structured, academically concise mathematical interpretation of the system’s components.

---

## 1. Large Language Model (LLM) Objective

WorldGPT’s **Prompt Enhancer** employs a large language model (ChatGPT) that decomposes natural-language prompts into structured sub-prompts.  
This process follows the standard **autoregressive language modeling objective**:

$$
L_{\text{LLM}} = - \sum_{i=1}^{n} \log P(w_i \mid w_{<i}; \theta)
$$

Where:  
- \( w_i \): the \( i^{th} \) token (word or subword)  
- \( w_{<i} \): preceding context tokens  
- \( \theta \): model parameters  

**Role:**  
This objective ensures coherent text expansion and contextual reasoning.  
In WorldGPT, it allows decomposition of prompts like *“a child running across a field”* into **semantic structures** — *subject (child)*, *action (running)*, *setting (field)*, *emotion (freedom)* — providing linguistic constraints for the subsequent visual generation stages.

---

## 2. Diffusion Process — Visual Generation Foundation

The diffusion backbone governs the **forward (noising)** process, where clean images are gradually corrupted:

$$
x_t = \alpha_t x_{t-1} + (1 - \alpha_t)\epsilon, \quad \epsilon \sim \mathcal{N}(0, I)
$$

Where:  
- \( x_t \): image (or frame) at timestep \( t \)  
- \( \alpha_t \): noise schedule coefficient  
- \( \epsilon \): Gaussian noise  

**Role:**  
The model learns to **reverse** this stochastic process — generating high-fidelity key frames from random noise conditioned on text or image embeddings.  
This is the mathematical foundation behind **Stable Diffusion** and **DynamiCrafter** used in WorldGPT.

---

## 3. Reverse (Denoising) Process

During generation, the model performs iterative **denoising**, reconstructing frames through conditional sampling:

$$
p_{\theta}(x_{t-1} \mid x_t) = \mathcal{N}(x_{t-1}; \mu_{\theta}(x_t, t), \Sigma_{\theta}(x_t, t))
$$

Where:  
- \( \mu_{\theta}, \Sigma_{\theta} \): predicted mean and variance of the reverse distribution  

**Role:**  
Each iteration predicts cleaner frames, refining both spatial structure and semantic detail.  
WorldGPT applies this both for **key-frame generation** and **temporal interpolation**, ensuring coherent transitions between frames.

---

## 4. Temporal Video Generation Loss

The **Full Video Translation** module defines a temporal reconstruction loss over generated frames, ensuring smooth motion and temporal coherence:

$$
\min_{F, B} \sum_t L_{\text{video}} = \| I_t - G(I, F_t, B_t) \|
$$

Where:  
- \( I_t \): target video frame at time \( t \)  
- \( G(\cdot) \): generator combining base image \( I \), motion field \( F_t \), and background \( B_t \)  

**Role:**  
Minimizes differences between generated and real frames, aligning **optical flow** and **appearance consistency**.  
This is conceptually analogous to **motion vector prediction** in spatiotemporal modeling.

---

## 5. Evaluation Metrics — Quantitative Benchmarks

WorldGPT’s evaluation uses **AIGCBench metrics** to measure alignment, motion realism, and frame consistency.

| **Metric** | **Formula (Simplified)** | **Interpretation** |
|:--|:--|:--|
| **CLIP Alignment (SA)** | $$ S_{\text{align}} = \cos(E_{\text{text}}, E_{\text{video}}) $$ | Cosine similarity between text and video embeddings — measures semantic alignment. |
| **Temporal Consistency (TC)** | $$ TC = \frac{1}{T-1} \sum_{t=1}^{T-1} SSIM(I_t, I_{t+1}) $$ | Average structural similarity between consecutive frames — quantifies motion smoothness. |
| **Motion Effect (ME)** | Derived from optical flow magnitude statistics | Evaluates dynamism and realism of motion fields. |
| **Frame Quality (FQ)** | SSIM or DOVER between real and generated frames | Measures per-frame clarity and visual fidelity. |

**Reported Performance:**
- Temporal Consistency ≈ **0.992**  
- CLIP Alignment ≈ **0.883**  
- Frame Quality ≈ **0.374**

**Interpretation:**  
High alignment and motion coherence validate the system’s temporal modeling and semantic fidelity, while lower FQ reflects typical diffusion-induced softness.

## 7. System as a Probabilistic World Model

WorldGPT formalizes video generation as a **conditional probabilistic process** over time:

$$
P(\text{Video} \mid \text{Text}) = P(\text{KeyFrames} \mid \text{Text}) \prod_t P(I_t \mid I_{t-1}, \text{Text})
$$

**Interpretation:**  
Each frame depends on the previous frame (\( I_{t-1} \)) and the global text condition.  
This defines a **Markov chain** over frames, guided by language context — a mathematical abstraction of how AI “world models” evolve temporally.

````python
# =======================================================================
# Mini-WorldGPT — Text-to-Video World Model (Simplified)
# -----------------------------------------------------------------------
# This lightweight implementation demonstrates the *core ideas*:
# 1️ LLM-driven prompt enhancement
# 2️ Key-frame synthesis (vision generator)
# 3️ Temporal interpolation (motion simulation)
# 4️ Evaluation + Visualization (alignment & consistency)
# =======================================================================

import torch, torch.nn as nn, torch.nn.functional as F
import matplotlib.pyplot as plt
import numpy as np, random, time

# ----------------------------------------------------------
# 1️ TEXT ENCODER (Prompt Enhancer)
# ----------------------------------------------------------
# In the real paper, ChatGPT is used to expand a prompt into structured sub-prompts.
# Here, we simulate that with an embedding + transformer encoder.

class TextEncoder(nn.Module):
    def __init__(self, vocab_size=1000, embed_dim=128):
        super().__init__()
        self.embed = nn.Embedding(vocab_size, embed_dim)
        self.encoder = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(embed_dim, nhead=4, batch_first=True),
            num_layers=2)
    def forward(self, x):
        e = self.embed(x)
        return self.encoder(e).mean(dim=1)  # (B, embed_dim)

# Simple tokenizer for demonstration
def tokenize(texts, vocab):
    tokens = []
    for t in texts:
        idxs = [vocab.get(w, 0) for w in t.lower().split()]
        if len(idxs) < 10:
            idxs += [0]*(10-len(idxs))
        tokens.append(torch.tensor(idxs[:10]))
    return torch.stack(tokens)

vocab = {w:i+1 for i,w in enumerate("a an the red blue dog cat man running beach field girl horse car tree sky".split())}
````

````python
# ----------------------------------------------------------
# 2️ IMAGE GENERATOR (Key-Frame Synthesis)
# ----------------------------------------------------------
# This mimics Stable Diffusion’s decoder head with a simple CNN.
# It takes the text embedding and generates a "frame" (32x32 RGB).

class KeyFrameGenerator(nn.Module):
    def __init__(self, text_dim=128, img_dim=3, size=32):
        super().__init__()
        self.fc = nn.Linear(text_dim, 256)
        self.gen = nn.Sequential(
            nn.ConvTranspose2d(16,64,4,2,1), nn.ReLU(),
            nn.ConvTranspose2d(64,32,4,2,1), nn.ReLU(),
            nn.ConvTranspose2d(32,3,4,2,1), nn.Sigmoid()
        )
    def forward(self, text_emb):
        x = F.relu(self.fc(text_emb)).view(-1,16,4,4)
        return self.gen(x)  # (B,3,32,32)
````

````python
# ----------------------------------------------------------
# 3️ TEMPORAL DIFFUSION SIMULATION (Frame Interpolator)
# ----------------------------------------------------------
# DynamiCrafter models realistic motion. Here we simulate by interpolating
# latent frames through a ConvGRU (spatiotemporal memory).

class ConvGRUCell(nn.Module):
    def __init__(self, input_dim, hidden_dim, kernel_size=3):
        super().__init__()
        self.hidden_dim = hidden_dim
        padding = kernel_size//2
        self.conv = nn.Conv2d(input_dim + hidden_dim, 3*hidden_dim, kernel_size, padding=padding)
    def forward(self, x, h_prev):
        combined = torch.cat([x, h_prev], dim=1)
        gates = self.conv(combined)
        z, r, n = torch.chunk(gates, 3, dim=1)
        z, r = torch.sigmoid(z), torch.sigmoid(r)
        n = torch.tanh(n)
        h = (1 - z)*n + z*h_prev
        return h

class TemporalGenerator(nn.Module):
    def __init__(self, channels=3, hidden_dim=32, steps=5):
        super().__init__()
        self.cell = ConvGRUCell(channels, hidden_dim)
        self.out = nn.Conv2d(hidden_dim, 3, 3, padding=1)
        self.steps = steps
    def forward(self, start_frame):
        B,C,H,W = start_frame.shape
        h = torch.zeros(B,32,H,W, device=start_frame.device)
        frames = [start_frame]
        x = start_frame
        for _ in range(self.steps):
            h = self.cell(x, h)
            x = self.out(h)
            frames.append(torch.sigmoid(x))
        return torch.stack(frames, dim=1)  # (B, T, 3, H, W)
````

````python
# ----------------------------------------------------------
# 4️ FULL MODEL ASSEMBLY (WorldGPT Simplified)
# ----------------------------------------------------------
class MiniWorldGPT(nn.Module):
    def __init__(self):
        super().__init__()
        self.text_enc = TextEncoder(len(vocab)+1)
        self.keyframe_gen = KeyFrameGenerator()
        self.temporal_gen = TemporalGenerator(steps=5)
    def forward(self, text_idx):
        text_emb = self.text_enc(text_idx)
        key_frame = self.keyframe_gen(text_emb)
        video = self.temporal_gen(key_frame)
        return video  # (B, T, 3, 32, 32)
````

````python
# ----------------------------------------------------------
# 5️ TRAINING SETUP (Synthetic World Simulation)
# ----------------------------------------------------------
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = MiniWorldGPT().to(device)
opt = torch.optim.Adam(model.parameters(), lr=1e-4)

# Dummy dataset: text prompts → random color patterns
texts = ["a red dog running", "a blue sky", "a man in a field", "a car on the beach"]
labels = torch.rand(len(texts), 6, 3, 32, 32)  # pseudo "video frames"
text_idx = tokenize(texts, vocab).to(device)
labels = labels.to(device)
````

````python
# ----------------------------------------------------------
# 6️ TRAINING LOOP
# ----------------------------------------------------------
epochs = 10
train_loss = []
for epoch in range(epochs):
    model.train()
    video_pred = model(text_idx)
    loss = F.mse_loss(video_pred, labels)
    opt.zero_grad(); loss.backward(); opt.step()
    train_loss.append(loss.item())
    print(f"Epoch {epoch+1}/{epochs} - Loss: {loss.item():.4f}")
````

````python
# ----------------------------------------------------------
# 7️ VISUALIZATION — LOSS CURVE
# ----------------------------------------------------------
plt.figure(figsize=(6,4))
plt.plot(train_loss, marker='o')
plt.title("Training Loss — Mini WorldGPT")
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.grid(True)
plt.show()
````

````python
# ----------------------------------------------------------
# 8️ PREDICTION AND VIDEO FRAME VISUALIZATION
# ----------------------------------------------------------
model.eval()
with torch.no_grad():
    text = ["a red dog running"]
    x = tokenize(text, vocab).to(device)
    pred_video = model(x)[0].cpu()  # (T,3,32,32)

fig, axes = plt.subplots(1,6, figsize=(12,3))
for i in range(6):
    frame = np.transpose(pred_video[i].numpy(), (1,2,0))
    axes[i].imshow(frame)
    axes[i].set_title(f"Frame {i}")
    axes[i].axis('off')
plt.suptitle("Generated Temporal Sequence — WorldGPT (Educational Simulation)")
plt.show()
````

````python
# ----------------------------------------------------------
# 9️ EVALUATION METRICS — FIXED VERSION
# ----------------------------------------------------------
def temporal_consistency(video):
    """Compute temporal smoothness as inverse of frame-wise MSE."""
    diffs = [F.mse_loss(video[i], video[i+1]).item() for i in range(video.size(0)-1)]
    return 1 - np.mean(diffs)  # Higher = smoother motion

def clip_alignment_score(text_emb, video_emb):
    """Project both text and video to same latent space and compute cosine similarity."""
    # Reduce video to a global feature (B, 128)
    B, T, C, H, W = video_emb.shape
    video_flat = video_emb.view(B, T, -1).mean(dim=1)  # (B, C*H*W) → temporal average
    video_proj = torch.nn.Linear(video_flat.size(1), text_emb.size(1)).to(video_emb.device)
    video_feat = video_proj(video_flat)
    cos = F.cosine_similarity(text_emb, video_feat)
    return cos.mean().item()

# Evaluate (synthetic proxy metrics)
text_emb = model.text_enc(tokenize(["a red dog running"], vocab).to(device))
video_emb = model.temporal_gen(model.keyframe_gen(text_emb))

score_tc = temporal_consistency(video_emb[0])
score_align = clip_alignment_score(text_emb, video_emb)
print(f"Temporal Consistency: {score_tc:.3f} | Text-Video Alignment: {score_align:.3f}")
````
