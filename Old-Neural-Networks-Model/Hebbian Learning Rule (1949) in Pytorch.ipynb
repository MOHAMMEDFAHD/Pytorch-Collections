# Hebbian Learning Rule (1949)

# https://pubmed.ncbi.nlm.nih.gov/10643472/

---

## Concept
**“Cells that fire together, wire together.”**  
This principle captures the idea that the connection (synaptic weight) between two neurons becomes stronger if they are activated at the same time.

---

## Description
Hebbian learning is a **local learning rule** that updates synaptic weights based only on the activity of the **pre-synaptic neuron** (input) and the **post-synaptic neuron** (output).  

Mathematically, the update is:

$$
\Delta w_{ij} = \eta \, x_j \, y_i
$$

where:
- \( \Delta w_{ij} \): change in weight from input neuron \(j\) to output neuron \(i\)  
- \( \eta \): learning rate (scales the update size)  
- \( x_j \): activity of the input neuron  
- \( y_i \): activity of the output neuron  

If both \(x_j\) and \(y_i\) are active simultaneously, the weight increases.  
If one is inactive, there is no update.

---

## Relevance
- **Earliest learning rule in neural computation** (1949, Donald Hebb).  
- Provided the first **biological motivation** for how neurons strengthen their connections.  
- Forms the foundation for:
  - **Associative memory models** (e.g., Hopfield Networks).  
  - **Feature learning** in unsupervised networks.  
  - Inspiration for **Hebbian-like rules** in modern deep learning.  

---

## Connection to Markov Processes
Hebbian learning has a probabilistic flavor:
- Weight changes are **local and stochastic**, depending only on the current activation states.  
- This is similar to **Markov models**, where transitions depend only on the present state, not the entire history.  

Thus, Hebbian learning can be viewed as a **probabilistic, local mechanism** for strengthening associations.

---

## Summary
- **Proposed**: 1949 by Donald Hebb.  
- **Rule**: Strengthen synapses when neurons co-activate.  
- **Equation**: \( \Delta w_{ij} = \eta \, x_j \, y_i \).  
- **Importance**: First biologically inspired weight update rule, foundational for neural networks.  
- **Legacy**: Influenced associative memory, energy-based models, and probabilistic learning frameworks.


```python
# ===========================================================
# Hebbian Learning Rule (1949)
# -----------------------------------------------------------
# Concept: "Cells that fire together, wire together."
# Description: Strengthens connections between co-activated neurons.
# Relevance: Early local learning rule, foundational to neural networks.
# Connection: Probabilistic, local weight updates (like Markov processes).
# ===========================================================

import torch
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# -----------------------------
# Hebbian Network Class
# -----------------------------
class HebbianNetwork:
    def __init__(self, n_inputs, n_neurons):
        """
        n_inputs: number of input features
        n_neurons: number of output units (neurons)
        """
        self.n_inputs = n_inputs
        self.n_neurons = n_neurons
        self.W = torch.zeros((n_neurons, n_inputs))  # weight matrix

    def forward(self, x):
        """Compute neuron outputs (binary threshold)"""
        x = torch.tensor(x, dtype=torch.float32)
        act = self.W @ x
        return torch.where(act >= 0, torch.tensor(1.0), torch.tensor(0.0))

    def hebbian_update(self, x, y, lr=0.1):
        """
        Apply Hebbian learning update:
        Δw_ij = lr * x_j * y_i
        """
        x = torch.tensor(x, dtype=torch.float32)
        y = torch.tensor(y, dtype=torch.float32)
        dw = lr * torch.outer(y, x)
        self.W += dw

    def visualize_weights(self, title="Weight Matrix"):
        plt.figure(figsize=(6,5))
        sns.heatmap(self.W.detach().numpy(), cmap="coolwarm", annot=True, fmt=".2f")
        plt.title(title)
        plt.xlabel("Inputs")
        plt.ylabel("Neurons")
        plt.show()
```

```python
# -----------------------------
# Example: Simple Binary Patterns
# -----------------------------
patterns = [
    (torch.tensor([1,0,1,0]), torch.tensor([1,0])),   # pattern 1
    (torch.tensor([0,1,0,1]), torch.tensor([0,1])),   # pattern 2
    (torch.tensor([1,1,0,0]), torch.tensor([1,1]))    # pattern 3
]

# Create Hebbian network (4 inputs, 2 outputs)
net = HebbianNetwork(n_inputs=4, n_neurons=2)
```

```python
# -----------------------------
# Training Phase
# -----------------------------
print("Training with Hebbian rule...")
for epoch in range(1, 6):
    for x, y in patterns:
        net.hebbian_update(x, y, lr=0.2)
    print(f"Epoch {epoch}:")
    net.visualize_weights(title=f"Weights after Epoch {epoch}")
```

```python
# -----------------------------
# Testing Recall
# -----------------------------
print("\nTesting pattern recall...")
for i, (x, y) in enumerate(patterns):
    output = net.forward(x)
    print(f"Pattern {i+1} Input {x.numpy()} -> Output {output.numpy()} (Target {y.numpy()})")
```

```python
# -----------------------------
# Visualization: Evolution of Weight Strength
# -----------------------------
epochs = 20
weights_over_time = []

net2 = HebbianNetwork(n_inputs=4, n_neurons=2)
for epoch in range(epochs):
    for x, y in patterns:
        net2.hebbian_update(x, y, lr=0.1)
    weights_over_time.append(net2.W.clone().numpy())
```

```python
# Plot weight evolution for each connection
weights_over_time = np.array(weights_over_time)
fig, axs = plt.subplots(1,2, figsize=(12,5))

for i in range(net2.n_neurons):
    for j in range(net2.n_inputs):
        axs[i].plot(weights_over_time[:,i,j], label=f"W[{i},{j}]")
    axs[i].set_title(f"Neuron {i} Weight Evolution")
    axs[i].legend()

plt.show()
```

```python
# -----------------------------
# Visualization: Heatmap of Final Weights
# -----------------------------
net2.visualize_weights(title="Final Weight Matrix (Hebbian Learned)")
```

# Hebbian Learning Results (Epochs 1–20)

## 1. Weight Heatmaps (Epoch 1 → 5)

**Update Rule:**
$$
\Delta w_{ij} = \eta x_j y_i
$$

- **Epoch 1:** Small initial strengthening (max ≈ 0.4).  
- **Epoch 2–3:** Clear amplification of co-active connections (e.g., \(W[0,0]\), \(W[1,1]\) increasing).  
- **Epoch 4–5:** Continued reinforcement; dominant diagonal weights reach ≈ 2.0.  

**Interpretation:**  
- Diagonal dominance = strong **self-associations** (input and output firing together).  
- Off-diagonal weights increase only where co-activations exist; otherwise remain near 0.

---

## 2. Weight Evolution Curves (per neuron)

- **Neuron 0:** \(W[0,0]\) grows fastest (direct self-association). \(W[0,1]\), \(W[0,2]\) grow moderately.  
- **Neuron 1:** \(W[1,1]\) grows steeply; \(W[1,3]\) rises moderately.  
- **Flat lines** (e.g., \(W[0,3]\)) indicate absence of correlation.  

**Interpretation:**  
- Confirms **Hebbian locality**: only frequently co-active connections strengthen.  
- **Linear growth** reflects additive Hebbian updates (no normalization mechanism).

---

## 3. Final Weight Matrix (Epoch 20)

- **Diagonal weights:** ≈ 4.0 → maximum reinforcement of co-activation.  
- **Symmetry:** Observed between rows/columns (e.g., \(W[0,*]\) and \(W[1,*]\)) due to balanced activity.  
- **Inactive pairs:** Some weights remain ≈ 0 (no co-activation).  

**Interpretation:**  
- The matrix encodes **stable associative links**.  
- Strongest synaptic weights correspond to neurons that repeatedly co-fired.  
- Captures the principle: *“cells that fire together, wire together.”*

---

##  Overall Academic Insight
- **Locality:** Updates depend only on presynaptic (\(x_j\)) and postsynaptic (\(y_i\)) activity.  
- **Unsupervised:** No external error or supervision required.  
- **Outcome:** Correlation-based connectivity emerges naturally.  
- **Significance:** Provides the foundation for **associative memory models** (e.g., Hopfield networks) and **unsupervised feature learning** (e.g., RBMs).
