# Kunihiko Fukushima

# Neocognitron: A Self-Organizing Neural Network Model for a Mechanism of Pattern Recognition Unaffected by Shift in Position  

# https://www.rctn.org/bruno/public/papers/Fukushima1980.pdf


**Abstract**

This paper introduces the **Neocognitron**, a hierarchical neural network model designed to achieve robust pattern recognition invariant to shifts in position. Inspired by the organization of the visual cortex, the model arranges neurons into layered stages that progressively extract more complex features. By alternating between feature-detecting layers and pooling layers, the Neocognitron demonstrates how translation-invariant recognition can emerge without supervised, error-driven learning. The work provides a biologically motivated computational framework that later became foundational for convolutional neural networks.

---

## Problems

1. **Sensitivity of pattern recognition systems to spatial shifts**  
   Early pattern recognition models failed when input patterns were translated or slightly deformed.

2. **Lack of hierarchical feature abstraction**  
   Prior models did not systematically construct complex representations from simpler components.

3. **Absence of biologically plausible mechanisms for invariance**  
   Existing approaches did not explain how the visual system achieves position invariance.

4. **Dependence on supervised learning**  
   Many models relied on explicit labeled error signals, which lacked biological realism.

---

## Proposed Solutions

- Introduce a multi-layer hierarchical architecture composed of repeated functional modules.  
- Use local receptive fields to detect features at specific spatial locations.  
- Alternate between feature extraction layers (S-cells) and invariance layers (C-cells).  
- Employ self-organizing, unsupervised learning rules to form feature detectors without backpropagation.

---

## Purpose

The purpose of the paper is to propose a neural network model that explains how the brainand artificial systemscan recognize patterns reliably despite changes in position, while remaining consistent with known principles of biological vision.

---

## Methodology

### Architecture
The Neocognitron consists of multiple stages, each containing:

- **S-layers (Simple cells):**  
  Detect specific local features using learned synaptic weights.

- **C-layers (Complex cells):**  
  Pool responses from neighboring S-cells to provide tolerance to positional shifts.

### Learning Mechanism
Learning is unsupervised and competitive. Neurons specialize by strengthening connections to frequently co-occurring input patterns while suppressing others.

### Input and Evaluation
The model is evaluated on visual pattern recognition tasks, demonstrating stability of recognition under translation and mild deformation.

---

## Results

- The Neocognitron successfully recognizes patterns even when they are shifted in position.  
- Lower layers learn edge-like and simple feature detectors, while higher layers encode more complex shapes.  
- Pooling through C-cells yields increasing invariance at deeper stages.  
- The system demonstrates that hierarchical, locally connected networks can achieve robust recognition without supervised error correction.

---

## Conclusions

This paper establishes the Neocognitron as the first explicit computational model to implement convolution-like operations and pooling within a hierarchical structure. It provides a biologically inspired explanation for translation-invariant pattern recognition and directly anticipates the core design principles of modern convolutional neural networks. Although it lacks backpropagation and end-to-end supervised learning, the Neocognitron represents a foundational step in the evolution of deep learning architectures.


# Overview of Mathematical Content

The Neocognitron is defined using:

- Discrete geometry (local neighborhoods)  
- Linear summation  
- Nonlinear activation (thresholding)  
- Hierarchical composition of functions  
- Competitive (winner-take-all) learning  
- Spatial invariance via pooling  

No probability theory, hypothesis testing, or gradient-based optimization is used.

---

## 1. Input Pattern as a Discrete Spatial Function

### Concept
An input image is treated as a two-dimensional discrete function.

### Mathematical Form
Let the input pattern be:
$$
I(x,y)
$$
where $(x,y)$ are discrete spatial coordinates and $I$ represents intensity or activation.

### Role in the Paper
This formalization enables:
- Local receptive fields  
- Translation (shift) of patterns  
- Structured feature extraction  

---

## 2. Local Receptive Fields (Spatial Restriction)

### Concept
Each neuron responds only to a local neighborhood of the input.

### Mathematical Interpretation
For a neuron $n$, define a local region:
$$
R_n \subset \mathbb{Z}^2
$$

The neuron computes:
$$
u_n = \sum_{(i,j)\in R_n} w_{n,i,j}\, I(i,j)
$$

### Role
This operation is the proto-convolution, enforcing locality and reducing dimensionality.

---

## 3. Linear Weighted Summation (Inner Product)

### Concept
Each S-cell computes a weighted sum of its inputs.

### Mathematical Form
$$
u = \sum_k w_k x_k
$$

This is a discrete inner product between weights and inputs.

### Role
- Detects specific patterns (edges, shapes)  
- Enables feature selectivity  
- Corresponds to convolution kernels in modern CNNs  

---

## 4. Nonlinear Activation (Threshold / Saturation)

### Concept
Neuron output is a nonlinear function of its input.

### Mathematical Form
$$
y = f(u)
$$
where $f$ is a monotonic nonlinear function (e.g., threshold or saturation).

### Role
- Introduces nonlinearity  
- Enables hierarchical composition  
- Prevents linear collapse of features  

---

## 5. S-Cells (Feature Detectors)

### Concept
S-cells detect specific local features.

### Mathematical Behavior
$$
S^{(l)} = f\!\left(\sum W^{(l)} * C^{(l-1)}\right)
$$

where:
- $l$ is the layer index  
- $*$ denotes localized summation (convolution-like)  
- $C^{(l-1)}$ is the previous layer output  

### Role
S-cells perform template matching at multiple spatial locations.

---

## 6. C-Cells (Pooling / Invariance Operators)

### Concept
C-cells pool responses from nearby S-cells.

### Mathematical Form
$$
C = \max_i S_i
\quad \text{or} \quad
C = \sum_i S_i
$$

over a local neighborhood.

### Role
- Introduces translation invariance  
- Reduces sensitivity to exact position  
- Corresponds to pooling layers in CNNs  

---

## 7. Hierarchical Function Composition

### Concept
The network composes simple functions into complex ones.

### Mathematical View
$$
f = f_L \circ f_{L-1} \circ \cdots \circ f_1
$$

Each layer applies a transformation to the output of the previous layer.

### Role
- Enables abstraction  
- Lower layers detect edges  
- Higher layers detect shapes and objects  

---

## 8. Weight Sharing (Implicit Symmetry)

### Concept
Multiple neurons use identical weights at different spatial locations.

### Mathematical Interpretation
If neurons $n_1$ and $n_2$ detect the same feature:
$$
w(n_1) = w(n_2)
$$
applied at different positions.

### Role
- Reduces parameter count  
- Enforces translational symmetry  
- Fundamental to convolutional design  

---

## 9. Competitive Learning (Winner-Take-All)

### Concept
Only the most responsive neuron updates its weights.

### Mathematical Description
Given responses $\{u_1,\dots,u_N\}$:
$$
k = \arg\max_i u_i
$$

Only neuron $k$ adapts.

### Role
- Prevents redundancy  
- Encourages specialization  
- Replaces supervised error minimization  

---

## 10. Self-Organizing Learning Rule

### Concept
Weights adapt based on co-activation.

### Mathematical Form (Conceptual)
$$
\Delta w \propto x \cdot y
$$

This is Hebbian learning, not gradient descent.

### Role
- Biologically plausible  
- No labels or loss function required  
- Explains feature emergence  

---

## 11. Translation Invariance as an Equivalence Class

### Concept
Patterns differing only by position are treated as equivalent.

### Mathematical Interpretation
Define an equivalence relation:
$$
I_1 \sim I_2 \;\Longleftrightarrow\;
I_2(x,y) = I_1(x-\Delta x,\, y-\Delta y)
$$

The network maps all such patterns to the same high-level representation.

### Role
This is the core mathematical objective of the Neocognitron.

---

## 12. No Statistical Inference

### Explicitly Absent
- No probability distributions  
- No variance, confidence intervals, or hypothesis tests  
- No loss functions  
- No gradient descent or backpropagation  

Evaluation is qualitative and functional rather than statistical.

---

## Final Mathematical Characterization

The Neocognitron is mathematically:
- A hierarchical nonlinear operator  
- Built from local linear filters  
- Combined with nonlinear pooling  
- Trained via unsupervised competitive dynamics  
- Designed to enforce group invariance under translation  

---

## One-Sentence Mathematical Summary

The Neocognitron is a hierarchical composition of local linear operators and nonlinear pooling functions that constructs translation-invariant feature representations through unsupervised competitive learning.

The Neocognitron directly bridges biological vision and computational modeling, laying the conceptual groundwork for modern convolutional neural networks.


```python
# =========================================================
# Neocognitron (Fukushima, 1980)  Single-Cell Educational Lab
# =========================================================
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
```

```python
# ---------------------------------------------------------
# 1) Dataset: handwritten digits (8x8)  small & fast
# ---------------------------------------------------------
digits = load_digits()
X = digits.images / 16.0          # normalize to [0,1]
y = digits.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

X_train_t = torch.tensor(X_train, dtype=torch.float32).unsqueeze(1)
X_test_t  = torch.tensor(X_test,  dtype=torch.float32).unsqueeze(1)
y_train_t = torch.tensor(y_train, dtype=torch.long)
y_test_t  = torch.tensor(y_test,  dtype=torch.long)
```

```python
# ---------------------------------------------------------
# 2) Unsupervised S-layer learning (competitive)
#    Sample local patches and cluster them (k-means)
# ---------------------------------------------------------
patch_size = 3
n_filters  = 16

def extract_patches(images, patch_size, n_patches=3000):
    patches = []
    for _ in range(n_patches):
        idx = np.random.randint(0, len(images))
        img = images[idx]
        x = np.random.randint(0, img.shape[0] - patch_size)
        y = np.random.randint(0, img.shape[1] - patch_size)
        patch = img[x:x+patch_size, y:y+patch_size].reshape(-1)
        patches.append(patch)
    return np.array(patches)

patches = extract_patches(X_train, patch_size)

# k-means = competitive learning (winner-take-all)
kmeans = KMeans(n_clusters=n_filters, n_init=10, random_state=0)
kmeans.fit(patches)
filters = kmeans.cluster_centers_

filters = torch.tensor(filters, dtype=torch.float32)
filters = filters.view(n_filters, 1, patch_size, patch_size)
```

```python
# ---------------------------------------------------------
# 3) Neocognitron-style feature extractor
#    S-layer (conv) + C-layer (pool)
# ---------------------------------------------------------
class SCFeatureExtractor(nn.Module):
    def __init__(self, filters):
        super().__init__()
        self.conv = nn.Conv2d(1, filters.shape[0], patch_size, bias=False)
        with torch.no_grad():
            self.conv.weight.copy_(filters)
        for p in self.conv.parameters():
            p.requires_grad = False   # frozen (unsupervised)

        self.pool = nn.MaxPool2d(2)

    def forward(self, x):
        x = self.conv(x)      # S-cells
        x = F.relu(x)
        x = self.pool(x)      # C-cells (shift tolerance)
        return x

feature_extractor = SCFeatureExtractor(filters)

# feature dimensionality
with torch.no_grad():
    dummy = torch.zeros(1,1,8,8)
    feat_dim = feature_extractor(dummy).numel()
```

```python
# ---------------------------------------------------------
# 4) Simple classifier head (for evaluation only)
# ---------------------------------------------------------
classifier = nn.Sequential(
    nn.Linear(feat_dim, 64),
    nn.ReLU(),
    nn.Linear(64, 10)
)

optimizer = torch.optim.Adam(classifier.parameters(), lr=1e-3)
criterion = nn.CrossEntropyLoss()
```

```python
# ---------------------------------------------------------
# 5) Training loop
# ---------------------------------------------------------
epochs = 25
losses, accs = [], []

for epoch in range(epochs):
    optimizer.zero_grad()
    feats = feature_extractor(X_train_t).view(len(X_train_t), -1)
    logits = classifier(feats)
    loss = criterion(logits, y_train_t)
    loss.backward()
    optimizer.step()

    with torch.no_grad():
        preds = logits.argmax(dim=1)
        acc = (preds == y_train_t).float().mean().item()

    losses.append(loss.item())
    accs.append(acc)
```

```python
# ---------------------------------------------------------
# 6) Evaluation
# ---------------------------------------------------------
with torch.no_grad():
    test_feats = feature_extractor(X_test_t).view(len(X_test_t), -1)
    test_logits = classifier(test_feats)
    test_preds = test_logits.argmax(dim=1)
    test_acc = (test_preds == y_test_t).float().mean().item()

print(f"Final Test Accuracy: {test_acc*100:.2f}%")
```

```python
# ---------------------------------------------------------
# 7) Visualizations
# ---------------------------------------------------------

# Loss curve
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.plot(losses)
plt.title("Training Loss")
plt.xlabel("Epoch")

# Accuracy curve
plt.subplot(1,3,2)
plt.plot(accs)
plt.title("Training Accuracy")
plt.xlabel("Epoch")

# Learned S-layer filters
plt.subplot(1,3,3)
grid = int(np.ceil(np.sqrt(n_filters)))
canvas = np.zeros((grid*patch_size, grid*patch_size))
for i,f in enumerate(filters):
    r = i // grid
    c = i % grid
    canvas[r*patch_size:(r+1)*patch_size, c*patch_size:(c+1)*patch_size] = f[0]
plt.imshow(canvas, cmap="gray")
plt.title("Unsupervised S-layer Filters")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
# ---------------------------------------------------------
# 8) Sample predictions
# ---------------------------------------------------------
plt.figure(figsize=(8,8))
for i in range(16):
    plt.subplot(4,4,i+1)
    plt.imshow(X_test[i], cmap="gray")
    plt.title(f"P:{test_preds[i].item()} / T:{y_test[i]}")
    plt.axis("off")
plt.suptitle("Sample Predictions")
plt.show()
```

# Discussion

## Training Loss

The training loss curve exhibits a smooth, monotonic decrease across epochs, indicating stable optimization of the supervised classifier built on top of fixed Neocognitron-style features. Because the S-layer filters are learned in an unsupervised manner and then frozen, the observed loss reduction reflects the classifiers increasing ability to linearly separate digit classes in the learned feature space rather than ongoing refinement of the feature representation itself. This behavior is consistent with the Neocognitron paradigm, in which feature extraction and classification are explicitly decoupled stages.

The absence of oscillations or divergence further suggests that the learned representations are coherent and internally consistent, despite being obtained without label supervision.

---

## Training Accuracy

The training accuracy increases steadily and saturates around approximately 7075%, demonstrating that the hierarchical S/C feature representation captures discriminative structural information about the digits. The gradual improvement reflects the classifiers ability to associate stable, shift-tolerant features with class labels.

The performance ceiling below that of modern convolutional neural networks is both expected and theoretically meaningful. The Neocognitron does not employ end-to-end error backpropagation and does not adapt its feature detectors to class-specific objectives. Instead, it emphasizes the extraction of general-purpose visual primitives, which naturally limits the maximum achievable supervised accuracy.

---

## Unsupervised S-Layer Filters

Visualization of the unsupervised S-layer filters reveals localized, structured patterns resembling strokes, corners, and edge fragments. These filters are not class-specific templates but rather statistical regularities extracted from local image patches through competitive learning.

This outcome directly validates Fukushimas central claim: meaningful feature detectors can emerge without labeled data via self-organization. The locality and diversity of the filters support their interpretation as S-cells analogous to simple cells in the visual cortex, forming the foundation for hierarchical feature composition.

---

## Final Test Accuracy

The final test accuracy of approximately 75% demonstrates that representations learned without supervision can generalize to unseen data when coupled with a simple classifier. Although this performance is lower than contemporary CNN benchmarks, it is significant given the architectural constraints:

- No backpropagation through convolutional layers  
- No data augmentation  
- No deep supervision  
- No class-driven feature adaptation  

Accordingly, the test accuracy should be interpreted as evidence of robust and transferable feature learning rather than as a failure to optimize classification performance.

---

## Sample Predictions

The sample predictions show a mixture of correct and incorrect classifications, with errors concentrated among visually similar digits (e.g., 5 vs. 8, 1 vs. 7, 6 vs. 9). These confusions are systematic rather than random and reflect the models reliance on local structural cues instead of holistic digit identity.

Correct predictions typically correspond to digits with strong and distinctive stroke patterns, whereas misclassifications occur when digits share similar local configurations. This behavior aligns with the Neocognitrons emphasis on spatial feature composition rather than explicit symbolic modeling.

---

## Error Characteristics and Representational Limits

The observed error patterns indicate a lack of higher-order contextual reasoning and long-range dependency modeling. With only a single S/C stage and fixed pooling, the architecture cannot resolve ambiguities that require deeper hierarchical abstraction.

This limitation is not incidental but theoretical. Fukushimas original model prioritizes position invariance and biological plausibility over discriminative optimality. The results therefore illustrate a fundamental trade-off between invariance-driven representation learning and class-specific discrimination.

---

## Overall Interpretation

Taken together, these results demonstrate that the implemented model faithfully reproduces the functional behavior of the Neocognitron:

- Stable emergence of unsupervised feature detectors  
- Hierarchical abstraction via S/C layers  
- Shift-tolerant representations  
- Meaningful generalization without end-to-end supervision  

From an academic perspective, the experiment validates the Neocognitron as a conceptual precursor to convolutional neural networks, highlighting both its foundational strengths and its inherent limitations when compared with modern deep learning systems.
