**Author:** A. M. Turing
**Published:** Mind, Vol. 49, pp. 433–460, 1950

---

## Abstract

Turing's landmark paper reformulates the philosophically intractable question "Can machines
think?" into an empirically testable criterion known as the Imitation Game (later termed the
Turing Test). The paper develops a rigorous conceptual framework for machine intelligence,
establishes the theoretical basis of digital computation as a substrate for thought, and
systematically refutes nine major objections to the possibility of thinking machines. It
concludes with a proposal for learning machines as the most promising path toward artificial
intelligence.

---

## Problem

The central question — "Can machines think?" — is argued to be semantically ill-formed.
Definitions of "machine" and "think" drawn from ordinary usage lead either to circular
reasoning or to resolution by popular opinion (e.g., a Gallup poll), both of which are
scientifically unacceptable. No rigorous, operationalizable criterion for machine intelligence
existed prior to this work.

---

## Proposed Solution

Turing replaces the original question with a behaviorally grounded, empirically testable
surrogate:

> **The Imitation Game:** An interrogator communicates via text with two entities — a
> human and a machine — and attempts to identify which is which. If the machine performs
> indistinguishably from the human, it is deemed to exhibit intelligent behavior.

This reframing shifts the question from metaphysics to measurable performance, bypassing
unanswerable debates about consciousness and inner states.

---

## Purpose

- To provide a rigorous operational definition of machine intelligence.
- To establish digital computers as the appropriate class of machines for the game.
- To refute prevailing philosophical, theological, and scientific objections to machine thought.
- To outline a concrete research program — learning machines — for achieving the goal
  within approximately fifty years.

---

## Methodology

### The Imitation Game Setup

| Role | Entity | Objective |
|------|--------|-----------|
| Interrogator (C) | Human | Identify which of X, Y is machine or human |
| Player A | Machine | Deceive interrogator into wrong identification |
| Player B | Human | Assist interrogator with truthful answers |

Communication is restricted to typewritten text to eliminate non-intellectual cues (voice,
appearance).

### Digital Computers as the Target Machine Class

Turing restricts the class of candidate machines to **digital computers**, characterized by
three components:

1. **Store** — memory for data and instructions.
2. **Executive Unit** — performs individual operations.
3. **Control** — enforces instruction order and execution.

Digital computers are classified as **discrete-state machines**, whose future states are
fully determined by current state and input:

$$s_{t+1} = f(s_t, i_t)$$

A key theoretical property established is **universality**: a single sufficiently capable
digital computer can simulate any discrete-state machine, making the class of digital
computers equivalent for purposes of the imitation game.

### Refutation of Nine Objections

Turing systematically addresses:

| # | Objection | Turing's Counter-Argument |
|---|-----------|--------------------------|
| 1 | Theological | Restricts divine omnipotence; unverifiable assumption |
| 2 | Heads in the Sand | Emotional, not logical; not a valid scientific argument |
| 3 | Mathematical (Gödel) | Human intellect also subject to unprovable limitations |
| 4 | Consciousness | Leads to solipsism; behavioral test sufficient in practice |
| 5 | Various Disabilities | Based on induction from limited machine experience |
| 6 | Lady Lovelace's | Universal machines can be programmed to generate novelty |
| 7 | Continuity of Nervous System | Discrete approximation sufficient under game conditions |
| 8 | Informality of Behaviour | Confuses rules of conduct with laws of behaviour |
| 9 | Extrasensory Perception | Addressable by isolating competitors in shielded rooms |

### Learning Machines Framework

Turing proposes abandoning direct simulation of the adult mind in favor of:

1. Programming a **child machine** with minimal initial structure.
2. Subjecting it to an **education process** (rewards, punishments, unemotional
   instruction channels).
3. Allowing the machine to **modify its own programs** within a fixed rule framework.

An analogy is drawn to biological evolution:

$$\text{Child Machine Structure} \equiv \text{Hereditary Material}$$
$$\text{Machine Modification} \equiv \text{Mutation}$$
$$\text{Experimenter Judgment} \equiv \text{Natural Selection}$$

A **random element** is advocated as useful for search during learning, analogous to
stochastic exploration in evolutionary processes.

---

## Results

Turing does not present empirical experimental results. Instead, the paper establishes the
following theoretical and predictive claims:

- **Prediction (1950):** Within fifty years (by ~2000), a machine with storage capacity of
  approximately $10^9$ binary digits could fool an average interrogator more than 30% of
  the time after five minutes of questioning.
- **Storage estimates:** Brain capacity estimated between $10^{10}$ and $10^{15}$ binary
  digits; the imitation game was conjectured to require no more than $10^9$.
- **Universality:** All sufficiently capable digital computers are computationally equivalent,
  reducing the question of machine intelligence to a programming problem rather than an
  engineering one.
- **Learning machines:** Demonstrated in preliminary experiments that simple reward-and-
  punishment machines could acquire rudimentary behaviors, though results were not
  conclusive.

---

## Conclusions

- The question "Can machines think?" is meaningless as posed; the Imitation Game provides
  a well-defined, testable replacement criterion.
- Digital computers, by virtue of universality, are the appropriate and sufficient substrate
  for exploring machine intelligence.
- No objection surveyed — theological, mathematical, philosophical, or empirical — provides
  a decisive proof that machines cannot achieve intelligent behavior under the game's
  criteria.
- The most promising path forward is the development of **learning machines** modeled on
  the child mind, subjected to structured education, rather than attempts to directly
  simulate the adult brain.
- Turing acknowledges the problem is primarily one of **programming**, with hardware
  advances expected to be sufficient, and anticipates that collaborative, long-term effort
  will be required.

# Mathematical & Statistical Content
## Computing Machinery and Intelligence — A. M. Turing (1950)

---

## 1. Discrete-State Machines

### Concept
A discrete-state machine is a system that moves between a **finite number of
distinct, well-separated states** in response to inputs. Turing defines them
formally using state-transition tables.

### Example Machine (Clicking Wheel)

| Last State | Input | Next State |
|------------|-------|------------|
| q1 | i0 | q2 |
| q1 | i1 | q1 |
| q2 | i0 | q3 |
| ... | ... | ... |

**Output table:**

| State | q1 | q2 | q3 |
|-------|----|----|----|
| Output | o0 | o0 | o1 |

### Role in the Paper
This formalism establishes that digital computers belong to the class of
discrete-state machines, whose future behavior is fully and precisely predictable
from their current state and input — a crucial theoretical foundation for the
entire argument.

---

## 2. State Prediction & Determinism

### Concept
Given the transition table and initial state, all future states are computable:

$$s_{t+1} = f(s_t,\ i_t)$$

where:
- $s_t$ = current internal state
- $i_t$ = current input signal
- $f$ = deterministic transition function

### Contrast with Laplace's Universe
Turing draws an analogy to **Laplacian determinism** — the idea that knowing
the complete state of the universe at one moment allows prediction of all future
states. However, Turing notes a critical difference:

- In physical systems, tiny errors compound catastrophically (sensitivity to
  initial conditions, an early notion of what is now called **chaos**).
- In discrete-state machines, small perturbations do **not** propagate — the
  system remains predictable.

### Role in the Paper
Establishes that digital computers are **reliably simulable**, making the
universality argument (Section 5) mathematically sound.

---

## 3. Storage Capacity

### Concept
The **storage capacity** of a machine is defined as the base-2 logarithm of its
total number of possible internal states:

$$C = \log_2(N)$$

where $N$ = total number of distinct states the machine can occupy.

### Worked Example
Turing considers a computer with:
- 100 sheets of paper
- 50 lines per sheet
- 30 decimal digits per line
- Symbols: digits 0–9

$$N = 10^{100 \times 50 \times 30} = 10^{150{,}000}$$

$$C = \log_2(10^{150{,}000}) \approx 150{,}000 \times 3.32 \approx 498{,}000 \text{ bits}$$

### Manchester Machine (Historical)
Turing states the Manchester computer had approximately:

$$N \approx 2^{165{,}000} \approx 10^{50{,}000}$$

$$C = 165{,}000 \text{ binary digits}$$

Broken down as:
| Component | Capacity (bits) |
|-----------|----------------|
| 64 magnetic tracks × 2560 | 163,840 |
| 8 electronic tubes × 1280 | 10,240 |
| Miscellaneous | ~300 |
| **Total** | **~174,380** |

### Additivity Property
When two machines are combined:

$$C_{\text{total}} = C_1 + C_2$$

### Role in the Paper
Gives a rigorous, quantitative measure of computational power and memory,
directly linking storage size to the machine's ability to simulate human
intelligence in the Imitation Game.

---

## 4. Universality of Digital Computers

### Concept
A **universal machine** is one that can simulate any other discrete-state machine,
provided it has sufficient storage and speed. Formally:

$$\forall\ M_i \in \mathcal{D},\ \exists\ \text{program } P_i\ \text{such that}\ U(P_i) \equiv M_i$$

where:
- $\mathcal{D}$ = class of all discrete-state machines
- $U$ = universal digital computer
- $\equiv$ = behavioral equivalence

### Role in the Paper
This is the central mathematical result of the paper. It reduces the question
"Can machines think?" to a **programming problem** rather than an engineering
one — any sufficiently large and fast digital computer can, in principle, simulate
any other machine, including one that passes the Imitation Game.

---

## 5. Gödel's Incompleteness Theorem & Related Results

### Concept
Turing references a family of formal mathematical results that place theoretical
limits on what discrete-state machines can compute or prove:

| Result | Author | Year | Core Claim |
|--------|---------|------|------------|
| Incompleteness Theorem | Gödel | 1931 | In any sufficiently powerful consistent system, there exist statements neither provable nor disprovable within it |
| Undecidability | Church | 1936 | No algorithm can decide all mathematical propositions |
| Undecidability | Kleene | 1935 | Related recursive function limits |
| Halting Problem | Turing | 1937 | No machine can decide whether an arbitrary machine halts |

### The Halting Problem (Turing's Own Result)
The most directly relevant result: for any machine $M$ running a program $P$,
there is no general algorithm $H$ such that:

$$H(M, P) = \begin{cases} \text{YES} & \text{if } M \text{ halts on } P \\ \text{NO} & \text{if } M \text{ runs forever} \end{cases}$$

This means there exist questions a machine **must** either answer wrongly or
fail to answer at all.

### Role in the Paper
Turing uses these results honestly — acknowledging machines have provable
limitations — but counters that **no equivalent proof exists showing humans are
exempt** from similar limitations. The mathematical objection is therefore
inconclusive.

---

## 6. Probabilistic & Statistical Reasoning

### Interrogator Success Rate
Turing's central empirical prediction is stated in probabilistic terms:

$$P(\text{correct identification}) \leq 0.70 \text{ after 5 minutes of questioning}$$

meaning a well-programmed machine should be able to **fool the interrogator at
least 30% of the time**, which Turing takes as a meaningful threshold of success.

### ESP Statistical Argument
Turing references experimental ESP data:

$$\text{Human (telepathic receiver): } \frac{130}{400} = 0.325 \text{ correct}$$

$$\text{Machine (random guessing): } \frac{104}{400} = 0.260 \approx \frac{1}{4} \text{ (chance level)}$$

This is used to illustrate that if ESP is real, the Imitation Game would need
experimental controls (shielded rooms) — a point about the **boundary conditions**
of valid statistical testing.

---

## 7. Random Elements in Machines

### Concept
Turing proposes incorporating a **stochastic component** into learning machines.
A random number generator introduces controlled non-determinism:

$$s_{t+1} = f(s_t,\ i_t,\ r_t), \quad r_t \sim \text{Uniform}$$

### Search Efficiency Argument
Turing gives a concrete mathematical illustration: finding a number
$n \in [50, 200]$ such that:

$$n = \left(\sum \text{digits of } n\right)^2$$

- **Systematic search:** scan $n = 51, 52, 53, \ldots$ — risks large barren
  regions with no solutions.
- **Random search:** sample $n$ uniformly at random — avoids tracking visited
  states and performs better when solutions are densely distributed.

### Role in the Paper
Anticipates **stochastic optimization** and **random search** methods central to
modern machine learning (e.g., stochastic gradient descent, random initialization).

---

## 8. Brain Storage Capacity Estimates

### Concept
Turing provides order-of-magnitude estimates to make the engineering target
concrete:

| Estimate | Value |
|----------|-------|
| Brain capacity (upper bound) | $10^{15}$ binary digits |
| Brain capacity (lower bound, Turing's preference) | $10^{10}$ binary digits |
| Required for Imitation Game (Turing's estimate) | $\leq 10^{9}$ binary digits |
| Encyclopaedia Britannica (11th ed.) | $\approx 2 \times 10^9$ binary digits |
| Practicable storage (1950 technology) | $\approx 10^7$ binary digits |

### Role in the Paper
Grounds the theoretical discussion in quantitative engineering reality, showing
that the storage gap between available and required capacity was finite and
closeable — making machine intelligence an engineering challenge, not an
impossibility.

---

## 9. Evolutionary Analogy as a Formal Mapping

### Concept
Turing establishes a precise structural correspondence between machine learning
and biological evolution:

$$\text{Child Machine Structure} \longleftrightarrow \text{Hereditary Material}$$

$$\text{Modifications to Child Machine} \longleftrightarrow \text{Mutation}$$

$$\text{Experimenter's Judgment} \longleftrightarrow \text{Natural Selection}$$

### Key Mathematical Insight
Turing argues that **directed mutation** (experimenter-guided) is strictly more
efficient than **random mutation** (biological evolution) because:

- The experimenter can identify the cause of failure and target mutations.
- No need to exhaustively explore the combinatorial space of all possible
  genotypes.

This is an early informal argument for what modern ML calls **gradient-based
optimization over random search**.

# Structured Problem–Gap–Solution Analysis
## Computing Machinery and Intelligence — A. M. Turing (1950)

---

| 1 | **Ill-defined central question:** The question "Can machines think?" lacks a rigorous, operationalizable definition of both "machine" and "think," making it resistant to scientific investigation. | Prior discourse relied on ordinary language and popular intuition, rendering any answer dependent on subjective definition or statistical opinion (e.g., a Gallup poll) rather than empirical evidence. | Replace the question with the **Imitation Game**: a behaviorally grounded, text-based test in which an interrogator attempts to distinguish a machine from a human, converting an unanswerable philosophical question into a measurable performance criterion. |
| 2 | **Absence of a standard machine class:** No agreed-upon class of machines was specified as the appropriate candidate for intelligent behavior, leading to ambiguity about what kind of system the question even concerns. | Without a well-defined target class, arguments about machine intelligence were either too broad (including biological organisms) or too narrow (restricted to a single device), preventing generalizable conclusions. | Restrict the candidate class to **digital computers**, justified by their universality property: any discrete-state machine can be simulated by a sufficiently capable digital computer, making the class both precise and maximally general. |
| 3 | **No formal model of machine computation:** The theoretical properties of computing machines — their states, transitions, and capacities — had not been connected systematically to the question of intelligence. | Without a formal computational model, claims about what machines could or could not do remained intuitive and anecdotal, with no basis for principled comparison between machines or between machines and humans. | Formalize digital computers as **discrete-state machines** defined by state-transition tables, storage capacity $C = \log_2(N)$, and deterministic state prediction $s_{t+1} = f(s_t, i_t)$, providing a rigorous mathematical substrate for the intelligence argument. |
| 4 | **Theological objection — soul as prerequisite for thought:** The claim that thinking requires an immortal soul, divinely granted only to humans, was treated as a serious barrier to machine intelligence. | This objection placed the question of machine intelligence outside the domain of empirical science entirely, making it immune to experimental resolution and discouraging technical research programs. | Counter theologically on its own terms: restricting God's freedom to confer souls arbitrarily contradicts divine omnipotence. More decisively, dismiss the objection as historically unreliable — theological arguments have repeatedly been overturned by scientific progress (e.g., the Copernican case). |
| 5 | **Mathematical objection — Gödelian limits on machines:** Results from mathematical logic (Gödel 1931, Church 1936, Turing 1937) establish that discrete-state machines cannot answer certain classes of questions correctly, suggesting a provable ceiling on machine intelligence. | These formal results were used to argue that machines are fundamentally inferior to human intellect, which supposedly faces no equivalent limitations, thus rendering machine intelligence permanently incomplete. | Acknowledge the limits honestly, but note the **asymmetry of proof**: while machine limitations are formally established, no equivalent proof demonstrates that human intellect is exempt from analogous constraints. The objection is therefore inconclusive rather than decisive. |
| 6 | **Consciousness objection — behavioral output insufficient for genuine thought:** The argument (associated with Professor Jefferson) holds that unless a machine experiences genuine emotion and self-awareness, any output it produces is mere symbol manipulation, not thought. | This position, taken to its logical conclusion, collapses into **solipsism** — one can only verify thought by being the thinker — making it impossible to attribute intelligence to any entity other than oneself, human or machine alike. | Demonstrate that the consciousness criterion is self-defeating: in practice, human intelligence is also assessed behaviorally (e.g., viva voce examinations). The Imitation Game operationalizes exactly this convention. If the objection is pressed, it refutes human-to-human attribution of thought as much as it refutes machine intelligence. |
| 7 | **Lady Lovelace's objection — machines can only do what they are programmed to do:** Based on observations of Babbage's Analytical Engine, it was argued that machines originate nothing and are incapable of genuine novelty or surprise. | This objection treats the fixed instruction set of a machine as a permanent ceiling on its output, ignoring the possibility of emergent complexity arising from programmed rules operating on novel inputs. | Counter with the **universality argument**: a universal machine can be programmed to simulate any discrete-state machine, including ones that produce outputs surprising even to their designers. Turing also offers a personal empirical rebuttal — machines frequently produce results that surprise him — attributable to insufficient prior calculation rather than absence of machine creativity. |
| 8 | **Continuity objection — nervous system is not discrete:** The continuous nature of neural dynamics (where small input variations can produce large output changes) was argued to make discrete-state machines fundamentally inadequate models of biological intelligence. | This objection implied that any digital simulation of neural processes would introduce irrecoverable approximation errors, making human-level intelligence unreachable via discrete computation. | Note that under **Imitation Game conditions**, the interrogator cannot exploit continuous versus discrete distinctions through text alone. A digital computer can produce statistically appropriate approximations (e.g., sampling plausible values of $\pi$) that are indistinguishable from continuous machine output in practice. |
| 9 | **Informality objection — human behavior cannot be fully rule-governed:** It was argued that because no finite rulebook can cover all conceivable human situations, humans cannot be machines, and therefore machines cannot replicate human behavior. | The argument conflates **rules of conduct** (explicit, conscious precepts) with **laws of behaviour** (physical laws governing a system), committing the logical fallacy of the undistributed middle, yet this confusion was rarely identified explicitly. | Distinguish sharply between the two senses of "rule." Laws of behaviour governing a machine are not the same as consciously followed rules of conduct. The absence of explicit conduct rules does not preclude the existence of underlying behavioural laws, and no scientific method can confirm their absence. |
| 10 | **No concrete research program for achieving machine intelligence:** Even granting that machine intelligence is theoretically possible, no systematic methodology existed in 1950 for actually building such a system. | Attempts to directly simulate the adult human mind faced an intractable complexity barrier: the full cognitive architecture of a mature mind was (and remains) far beyond direct specification or replication. | Propose the **learning machine paradigm**: begin with a minimal "child machine," subject it to a structured education process using rewards, punishments, and unemotional instruction channels, and allow it to modify its own programs. This is formalized via an evolutionary analogy — child machine structure $\leftrightarrow$ hereditary material; machine modification $\leftrightarrow$ mutation; experimenter judgment $\leftrightarrow$ natural selection — and is supplemented by a **random element** to enable efficient search over the space of possible behaviors. |

```python
# ============================================================
# Turing's Imitation Game — Educational PyTorch Implementation
# Paper: "Computing Machinery and Intelligence" (Turing, 1950)
# Dataset: MNIST (grayscale digits 0–9)
# ============================================================
#
# CONCEPTUAL MAPPING TO TURING (1950):
# ┌─────────────────────────────────────────────────────────┐
# │ Turing's Concept        │ Implementation                │
# ├─────────────────────────┼───────────────────────────────┤
# │ Digital Computer Store  │ Learned feature embeddings    │
# │ Executive Unit          │ FC processing layers          │
# │ Control Unit            │ Classification head           │
# │ Initial state (birth)   │ Random weight initialisation  │
# │ Education process       │ Supervised training           │
# │ Reward signal           │ Confidence bonus loss         │
# │ Punishment signal       │ Entropy penalty loss          │
# │ Imitation objective     │ Cross-entropy loss            │
# │ Learning machine        │ Backpropagation + Adam        │
# └─────────────────────────┴───────────────────────────────┘
#
# WHY MNIST?
# Turing's paper concerns symbolic, discrete reasoning —
# not complex visual patterns. MNIST's clean, low-dimensional
# grayscale digits (28×28) closely mirror the discrete, text-
# like input space Turing envisioned for his learning machine.
# ============================================================

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, Subset
import torchvision
import torchvision.transforms as transforms
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.patches import FancyBboxPatch
import warnings
warnings.filterwarnings('ignore')
```

```python
# ─────────────────────────────────────────────────────────────
# 0.  REPRODUCIBILITY & DEVICE
#     Turing: "The initial state of the mind at birth"
#     — fixed seed ensures reproducible 'birth state'
# ─────────────────────────────────────────────────────────────
torch.manual_seed(42)
np.random.seed(42)
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"[Device] Using: {DEVICE}")
```

```python
# ─────────────────────────────────────────────────────────────
# 1.  HYPER-PARAMETERS
# ─────────────────────────────────────────────────────────────
NUM_CLASSES   = 10       # MNIST: digits 0–9
EPOCHS        = 5        # Short training run (educational)
BATCH_SIZE    = 128      # Mini-batch size
LR            = 1e-3     # Adam learning rate
SUBSET_TRAIN  = 8000     # Training subset size
SUBSET_VAL    = 2000     # Validation subset size

# ── Turing Loss Weights ──────────────────────────────────────
# Turing described three learning signals:
#  (1) Imitation  : primary task — match the correct answer
#  (2) Reward     : reinforce confident correct predictions
#  (3) Punishment : penalise vague / uncertain outputs
W_IMITATION  = 1.0      # weight for cross-entropy loss
W_REWARD     = 0.4      # weight for reward signal loss
W_PUNISHMENT = 0.3      # weight for entropy penalty loss

MNIST_CLASSES = [str(i) for i in range(10)]   # '0' … '9'
```

```python
# ─────────────────────────────────────────────────────────────
# 2.  DATASET PREPARATION
#     "The child machine must be subjected to an appropriate
#      course of education." — Turing (1950), §7
#
#     MNIST: 28×28 grayscale handwritten digits.
#     We normalise to zero mean / unit variance and use
#     a small subset to stay fast and educational.
# ─────────────────────────────────────────────────────────────
mnist_transform = transforms.Compose([
    transforms.ToTensor(),
    # MNIST mean=0.1307, std=0.3081 (standard values)
    transforms.Normalize((0.1307,), (0.3081,)),
])

# Download MNIST automatically
train_full = torchvision.datasets.MNIST(
    root='./data', train=True,  download=True,
    transform=mnist_transform)
val_full   = torchvision.datasets.MNIST(
    root='./data', train=False, download=True,
    transform=mnist_transform)

# Take stratified subsets for speed
def stratified_subset(dataset, n_per_class, num_classes=10):
    """
    Return indices with exactly n_per_class samples per digit.
    This mirrors Turing's idea of balanced 'education' input.
    """
    targets  = np.array(dataset.targets)
    indices  = []
    for c in range(num_classes):
        cls_idx = np.where(targets == c)[0]
        chosen  = cls_idx[:n_per_class]
        indices.extend(chosen.tolist())
    np.random.shuffle(indices)
    return indices

n_train_per_class = SUBSET_TRAIN  // NUM_CLASSES   # 800 per digit
n_val_per_class   = SUBSET_VAL    // NUM_CLASSES   # 200 per digit

train_idx = stratified_subset(train_full, n_train_per_class)
val_idx   = stratified_subset(val_full,   n_val_per_class)

train_dataset = Subset(train_full, train_idx)
val_dataset   = Subset(val_full,   val_idx)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE,
                          shuffle=True,  num_workers=0, pin_memory=True)
val_loader   = DataLoader(val_dataset,   batch_size=BATCH_SIZE,
                          shuffle=False, num_workers=0, pin_memory=True)

print(f"[Data]   Train: {len(train_dataset)} samples "
      f"| Val: {len(val_dataset)} samples")
```

```python
# ─────────────────────────────────────────────────────────────
# 3.  MODEL DEFINITION — "The Imitation Machine"
#
#     We implement Turing's three-component digital computer
#     as a fully-connected neural network (appropriate for
#     MNIST's flattened pixel vectors):
#
#       STORE      → input embedding layer
#       EXECUTIVE  → hidden reasoning layers
#       CONTROL    → output / decision layer
#
#     "It is not the complexity of the machine that matters,
#      but the programme." — Turing (1950), §7
# ─────────────────────────────────────────────────────────────
class ImitationMachine(nn.Module):
    """
    Fully-connected network embodying Turing's three-part
    digital computer architecture for MNIST digit recognition.

    Input  : 28×28 = 784 flattened pixel values
    Output : 10-class logits (digits 0–9)
    """

    def __init__(self, input_dim=784, hidden_dims=(512, 256, 128),
                 num_classes=10, dropout_rate=0.3):
        super(ImitationMachine, self).__init__()

        # ── STORE: Compress raw pixels into a dense representation
        #    Analogous to the machine's "paper" / short-term memory
        self.store = nn.Sequential(
            nn.Linear(input_dim, hidden_dims[0]),
            nn.BatchNorm1d(hidden_dims[0]),
            nn.ReLU(inplace=True),
            nn.Dropout(dropout_rate),
        )

        # ── EXECUTIVE: Hierarchical reasoning
        #    Each layer refines the stored representation,
        #    like a human computer working through sub-problems
        self.executive = nn.Sequential(
            nn.Linear(hidden_dims[0], hidden_dims[1]),
            nn.BatchNorm1d(hidden_dims[1]),
            nn.ReLU(inplace=True),
            nn.Dropout(dropout_rate),

            nn.Linear(hidden_dims[1], hidden_dims[2]),
            nn.BatchNorm1d(hidden_dims[2]),
            nn.ReLU(inplace=True),
            nn.Dropout(dropout_rate * 0.7),   # less dropout near output
        )

        # ── CONTROL: Final decision / classification
        #    "The control sees that instructions are obeyed
        #     correctly and in the right order." — Turing (1950)
        self.control = nn.Linear(hidden_dims[2], num_classes)

        # Weight initialisation — the "initial state at birth"
        self._init_weights()

    def _init_weights(self):
        """Xavier initialisation — principled random starting state."""
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                nn.init.zeros_(m.bias)

    def forward(self, x):
        # Flatten 28×28 image → 784-dim vector
        x        = x.view(x.size(0), -1)          # (B, 784)
        stored   = self.store(x)                   # (B, 512)
        executed = self.executive(stored)          # (B, 128)
        logits   = self.control(executed)          # (B, 10)
        # Return logits and intermediate features for auxiliary losses
        return logits, executed
```

```python
# ─────────────────────────────────────────────────────────────
# 4.  LOSS FUNCTIONS
#     Implementing Turing's three learning signals:
#
#     L_total = w1 * L_imitation
#             + w2 * L_reward
#             + w3 * L_punishment
# ─────────────────────────────────────────────────────────────

def loss_imitation(logits, targets):
    """
    IMITATION LOSS (Primary Objective)
    ───────────────────────────────────
    Turing: the machine must correctly imitate the human response.
    Implementation: standard cross-entropy — penalise wrong answers.

        L_imit = -log P(correct class)
    """
    return F.cross_entropy(logits, targets)


def loss_reward(logits, targets):
    """
    REWARD SIGNAL LOSS
    ───────────────────────────────────
    Turing: "A reward signal increases the probability of
             repetition of the events which led up to it."
    Implementation: maximise log-confidence on correct class.
    The machine is rewarded for being both correct AND confident.

        L_rew = -mean[ log P(y_true) ]   for all samples
    """
    probs         = F.softmax(logits, dim=1)
    # Extract probability assigned to the true class
    correct_probs = probs[torch.arange(len(targets)), targets]
    return -torch.log(correct_probs + 1e-8).mean()


def loss_punishment(logits):
    """
    PUNISHMENT / ENTROPY PENALTY LOSS
    ───────────────────────────────────
    Turing: "Punishment makes events unlikely to be repeated."
    The machine is punished for giving vague, spread-out answers.
    High entropy = uncertain = punished.

        L_pun = H(p) = -sum( p * log p )   averaged over batch
    """
    probs   = F.softmax(logits, dim=1)
    entropy = -(probs * torch.log(probs + 1e-8)).sum(dim=1).mean()
    return entropy


def total_loss_fn(logits, targets):
    """
    Combined Turing Loss:
        L = w_imit * L_imitation
          + w_rew  * L_reward
          + w_pun  * L_punishment

    Returns total loss and each component separately for logging.
    """
    l_imit = loss_imitation(logits, targets)
    l_rew  = loss_reward(logits,    targets)
    l_pun  = loss_punishment(logits)

    l_total = (W_IMITATION  * l_imit
             + W_REWARD     * l_rew
             + W_PUNISHMENT * l_pun)

    return l_total, l_imit, l_rew, l_pun
```

```python
# ─────────────────────────────────────────────────────────────
# 5.  MODEL, OPTIMISER & SCHEDULER
# ─────────────────────────────────────────────────────────────
model = ImitationMachine(
    input_dim   = 784,
    hidden_dims = (512, 256, 128),
    num_classes = NUM_CLASSES,
).to(DEVICE)

# Adam optimiser — modern analogue of Turing's gradient-based learning
optimizer = optim.Adam(model.parameters(), lr=LR, weight_decay=1e-4)

# Cosine annealing — gradually reduce LR (like a teacher reducing
# intervention as the pupil becomes more capable)
scheduler = optim.lr_scheduler.CosineAnnealingLR(
    optimizer, T_max=EPOCHS, eta_min=1e-5)

total_params = sum(p.numel() for p in model.parameters()
                   if p.requires_grad)
print(f"[Model]  Trainable parameters: {total_params:,}")
print(f"[Model]  Architecture: Store(784→512) → "
      f"Executive(512→256→128) → Control(128→10)")
```

```python
# ─────────────────────────────────────────────────────────────
# 6.  TRAINING & VALIDATION LOOP
# ─────────────────────────────────────────────────────────────

# History containers — one value per epoch
history = {
    # Aggregate losses
    'train_loss':  [], 'val_loss':  [],
    # Component losses
    'train_imit':  [], 'val_imit':  [],
    'train_rew':   [], 'val_rew':   [],
    'train_pun':   [], 'val_pun':   [],
    # Accuracy
    'train_acc':   [], 'val_acc':   [],
    # Per-class accuracy per epoch → shape (EPOCHS, NUM_CLASSES)
    'per_class_acc': [],
}


def run_epoch(loader, model, optimizer=None, training=True):
    """
    Single forward (and optionally backward) pass over a DataLoader.

    Args:
        loader    : DataLoader for train or val split
        model     : ImitationMachine instance
        optimizer : pass optimizer for training, None for validation
        training  : bool — enables gradients and weight updates

    Returns:
        dict of aggregated metrics for this epoch
    """
    model.train() if training else model.eval()

    # Accumulators
    sum_loss = sum_imit = sum_rew = sum_pun = 0.0
    n_correct = n_total = 0
    class_correct = np.zeros(NUM_CLASSES)
    class_total   = np.zeros(NUM_CLASSES)
    all_conf      = []     # max predicted probability per sample
    all_correct   = []     # bool: was prediction correct?

    ctx = torch.enable_grad() if training else torch.no_grad()

    with ctx:
        for images, labels in loader:
            images, labels = images.to(DEVICE), labels.to(DEVICE)

            # Forward pass through Store → Executive → Control
            logits, _ = model(images)

            # Compute combined Turing loss
            l_tot, l_imit, l_rew, l_pun = total_loss_fn(logits, labels)

            if training:
                optimizer.zero_grad()
                l_tot.backward()
                # Gradient clipping — prevents exploding gradients
                nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
                optimizer.step()

            # ── Accumulate losses (weighted by batch size)
            bs          = images.size(0)
            sum_loss   += l_tot.item()  * bs
            sum_imit   += l_imit.item() * bs
            sum_rew    += l_rew.item()  * bs
            sum_pun    += l_pun.item()  * bs

            # ── Accuracy
            probs             = F.softmax(logits, dim=1)
            confidence, preds = probs.max(dim=1)
            n_correct        += (preds == labels).sum().item()
            n_total          += bs

            # ── Per-class accuracy tracking
            for c in range(NUM_CLASSES):
                mask              = (labels == c)
                class_total[c]   += mask.sum().item()
                class_correct[c] += (preds[mask] == labels[mask]).sum().item()

            # ── Confidence tracking
            all_conf.extend(confidence.cpu().detach().numpy())
            all_correct.extend((preds == labels).cpu().numpy())

    return {
        'loss':          sum_loss / n_total,
        'imit':          sum_imit / n_total,
        'rew':           sum_rew  / n_total,
        'pun':           sum_pun  / n_total,
        'acc':           n_correct / n_total,
        'class_correct': class_correct,
        'class_total':   class_total,
        'confidences':   np.array(all_conf),
        'correct_mask':  np.array(all_correct, dtype=bool),
    }


# ── Main Training Loop ────────────────────────────────────────
print("\n" + "="*62)
print("  TRAINING THE IMITATION MACHINE  (Turing, 1950)")
print("  Dataset: MNIST | Epochs: 5 | Device:", DEVICE)
print("="*62)

best_val_acc    = 0.0
best_epoch      = 1
final_val_conf  = None   # confidence data from last epoch (for Panel G)

for epoch in range(1, EPOCHS + 1):

    # ── Training pass
    tr = run_epoch(train_loader, model,
                   optimizer=optimizer, training=True)

    # ── Validation pass (no gradients)
    va = run_epoch(val_loader, model,
                   optimizer=None, training=False)

    # ── LR scheduler step
    scheduler.step()

    # ── Record history
    history['train_loss'].append(tr['loss'])
    history['val_loss'].append(va['loss'])
    history['train_imit'].append(tr['imit'])
    history['val_imit'].append(va['imit'])
    history['train_rew'].append(tr['rew'])
    history['val_rew'].append(va['rew'])
    history['train_pun'].append(tr['pun'])
    history['val_pun'].append(va['pun'])
    history['train_acc'].append(tr['acc'])
    history['val_acc'].append(va['acc'])

    # Per-class accuracy this epoch
    pc_acc = va['class_correct'] / (va['class_total'] + 1e-8)
    history['per_class_acc'].append(pc_acc)

    # Save best model weights
    if va['acc'] > best_val_acc:
        best_val_acc = va['acc']
        best_epoch   = epoch
        torch.save(model.state_dict(), 'best_imitation_machine.pt')

    # Keep confidence data from last epoch
    if epoch == EPOCHS:
        final_val_conf = va

    print(f"  Epoch [{epoch}/{EPOCHS}]  "
          f"Train Loss: {tr['loss']:.4f}  "
          f"Val Loss: {va['loss']:.4f}  "
          f"Train Acc: {tr['acc']*100:5.2f}%  "
          f"Val Acc: {va['acc']*100:5.2f}%")

print(f"\n  Best Val Accuracy: {best_val_acc*100:.2f}% "
      f"at Epoch {best_epoch}")
print("="*62)
```

```python
# ─────────────────────────────────────────────────────────────
# 7.  PREDICTION PIPELINE
#     Load best weights and run full inference on val set
# ─────────────────────────────────────────────────────────────
print("\n[Inference] Loading best weights …")
model.load_state_dict(
    torch.load('best_imitation_machine.pt', map_location=DEVICE))
model.eval()

all_preds   = []
all_labels  = []
all_probs   = []

with torch.no_grad():
    for images, labels in val_loader:
        images = images.to(DEVICE)
        logits, _ = model(images)
        probs     = F.softmax(logits, dim=1)
        _, preds  = probs.max(dim=1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.numpy())
        all_probs.extend(probs.cpu().numpy())

all_preds  = np.array(all_preds)
all_labels = np.array(all_labels)
all_probs  = np.array(all_probs)

# ── Confusion matrix (normalised)
conf_mat = np.zeros((NUM_CLASSES, NUM_CLASSES), dtype=np.float32)
for t, p in zip(all_labels, all_preds):
    conf_mat[t, p] += 1.0
conf_mat_norm = conf_mat / (conf_mat.sum(axis=1, keepdims=True) + 1e-8)

# ── Per-class final accuracy
final_pc_acc = conf_mat_norm.diagonal()

# ── Confidence statistics
max_conf         = all_probs.max(axis=1)
correct_mask     = (all_preds == all_labels)
mean_conf_corr   = max_conf[correct_mask].mean()
mean_conf_incorr = max_conf[~correct_mask].mean() \
                   if (~correct_mask).sum() > 0 else 0.0

# ── Top / bottom 3 classes by accuracy
sorted_idx  = np.argsort(final_pc_acc)[::-1]
top_classes = [MNIST_CLASSES[i] for i in sorted_idx[:3]]
low_classes = [MNIST_CLASSES[i] for i in sorted_idx[-3:]]

print(f"[Eval]   Overall Val Accuracy : {correct_mask.mean()*100:.2f}%")
print(f"[Eval]   Mean Conf (Correct)  : {mean_conf_corr:.3f}")
print(f"[Eval]   Mean Conf (Incorrect): {mean_conf_incorr:.3f}")
print(f"[Eval]   Top digits    : {', '.join(top_classes)}")
print(f"[Eval]   Weakest digits: {', '.join(low_classes)}")
```

## Panel A — Total Loss Curves

### Overview
Tracks the aggregate weighted loss across all three Turing objectives
(Imitation + Reward + Punishment) over 5 training epochs for both the
training and validation splits.

### Key Findings
- Training loss undergoes a steep initial descent from approximately 1.2
  to below 0.4 within the first two epochs, then continues declining
  gradually to near 0.2 by epoch 5.
- Validation loss begins lower (~0.55) than training loss at epoch 1,
  converges rapidly, and plateaus at approximately 0.343 from epoch 3
  onward.
- The two curves converge and remain closely tracked after epoch 2,
  with no observable divergence through epoch 5.

### Discussion
The rapid early descent is characteristic of a well-initialised network
learning the dominant statistical structure of MNIST digits quickly —
consistent with Turing's intuition that a "child machine" can absorb
basic patterns efficiently from structured input. The fact that
validation loss is initially lower than training loss is attributable to
dropout regularisation being active only during training, which
artificially inflates training loss in early epochs. The stable
post-epoch-2 convergence without divergence indicates the absence of
overfitting, supporting the adequacy of weight decay and dropout for
this dataset scale (8,000 samples). The plateau in validation loss
suggests the model approaches its representational capacity limit within
5 epochs on this subset, which is expected given MNIST's relative
simplicity.

### Limitations
Five epochs on an 8,000-sample subset provides an incomplete picture of
asymptotic behaviour. The aggregate loss conflates three qualitatively
different objectives, making it difficult to attribute improvement to
any single Turing learning signal without reference to Panel B.

## Panel B — Component Loss Curves

### Overview
Decomposes the total loss into its three constituent Turing-motivated
objectives — Imitation (cross-entropy), Reward (negative log-confidence
on correct class), and Punishment (entropy penalty) — tracked separately
across epochs for train and validation splits.

### Key Findings
- **Imitation Loss (L_imit):** Descends from ~0.65 to ~0.21, with train
  and val curves closely aligned throughout. Final val value: 0.2122.
- **Reward Signal (L_rew):** Follows a near-identical trajectory to
  L_imit (final val: 0.2122), which is mathematically expected since
  both are functions of the predicted probability on the correct class.
- **Entropy Penalty (L_pun):** Begins higher (~0.65) and descends more
  steeply, reaching ~0.15 by epoch 5 (final val: 0.1542). Train and val
  curves diverge slightly more than for the other two objectives.

### Discussion
The near-perfect co-movement of L_imit and L_rew confirms their
mathematical relatedness: both measure the model's confidence on the
correct class, differing only in their derivation (log-loss versus
negative log-probability). This redundancy is pedagogically motivated —
it mirrors Turing's distinct conceptual framing of "imitation" versus
"reward" — but practically these two signals provide correlated
gradient information. The steeper decline and lower final value of
L_pun reflects the model rapidly concentrating probability mass onto
fewer classes as training progresses, consistent with a well-calibrated
classifier on a relatively easy dataset like MNIST. The slight train/val
divergence in L_pun suggests mild overconfidence on training data,
which is a known property of networks trained with entropy-minimising
auxiliary objectives.

### Limitations
The high correlation between L_imit and L_rew raises the question of
gradient redundancy: the reward signal may provide limited additional
learning signal beyond the imitation loss. A formal ablation study
removing individual loss components would be necessary to quantify each
term's independent contribution.

A best validation accuracy of 93.15% on an 8,000-sample MNIST subset

## Panel G — Confidence Distribution Histogram

### Overview
Plots the distribution of maximum predicted class probabilities
(confidence scores) separately for correctly and incorrectly classified
validation samples, with vertical lines marking mean confidence for

### Key Findings
- Correct predictions are overwhelmingly concentrated in the high-
  confidence region (0.9–1.0), with a mean confidence of 0.969.
- Incorrect predictions are broadly distributed across the full
  confidence range (0.2–1.0), with a mean confidence of 0.709.
- A substantial fraction of incorrect predictions are made with high
  confidence (>0.8), suggesting the model is miscalibrated for its
  error cases.
- The separation between the two mean confidence values (Δ = 0.260) is
  meaningful but not absolute.

### Discussion
The high mean confidence on correct predictions (0.969) confirms that
the entropy penalty and reward signal are functioning as intended —
encouraging the model to concentrate probability mass on its chosen
class. However, the non-trivial density of high-confidence incorrect
predictions (the right tail of the red histogram) is a calibration
concern: the model assigns near-certainty to wrong answers in a
meaningful fraction of cases. This is a well-documented failure mode of
cross-entropy-trained classifiers, often attributed to the combination
of softmax saturation and insufficient regularisation of the output
distribution. Ideally, an entropy penalty should suppress this
behaviour, but the relatively low punishment weight (W=0.3) appears
insufficient to fully eliminate overconfident errors. From a Turing
perspective, this mirrors the paper's discussion of "errors of
conclusion" — the machine commits confidently to a wrong answer rather
than expressing appropriate uncertainty.

### Limitations
Confidence calibration cannot be fully assessed from histograms alone;
reliability diagrams or Expected Calibration Error (ECE) metrics would
provide a more rigorous evaluation. The 2,000-sample validation set
limits the density estimation quality in the tails of the incorrect
predictions distribution.
