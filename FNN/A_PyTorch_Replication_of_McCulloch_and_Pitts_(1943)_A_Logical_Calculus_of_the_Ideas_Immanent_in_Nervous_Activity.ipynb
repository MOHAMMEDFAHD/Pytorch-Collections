# McCulloch & Pitts (1943) — *A Logical Calculus of the Ideas Immanent in Nervous Activity*

## Abstract
McCulloch and Pitts formalize neural activity as logical propositions, introducing a **logical calculus of neural nets**. They demonstrate that, because of the all-or-none firing property, neurons can be modeled as binary threshold units governed by propositional logic. Networks of such formal neurons can compute any function realizable by a **Turing machine**, laying the foundation of **artificial neural networks** and theoretical computational neuroscience.

---

## Problems
- Absence of a **formal mathematical model** of nervous activity.  
- Need to describe how **neural networks produce complex behaviors**.  
- Difficulty reconciling **inhibition, facilitation, extinction, and learning** in one framework.  
- No systematic method to determine when **logical propositions** are realizable by neural circuits.  

---

## Proposed Solutions
- Model neurons as **binary threshold devices** (all-or-none firing).  
- Map neuron states directly to **propositional logic**.  
- Build networks as **logical circuits**: conjunction (AND), disjunction (OR), and negation (NOT).  
- Introduce **equivalent nets** (different structures yielding identical logical behavior).  
- Use **recursive functions** and propositional calculus to model **cyclic networks** (feedback loops).  

---

## Purpose
To establish a **formal, logical, and mathematical foundation** for nervous activity, showing that the brain could be understood as a system of logical elements. This bridged **neurophysiology, symbolic logic, and computation**, providing a mechanistic explanation of cognition.  

---

## Methodology

### Assumptions
1. Neurons are **all-or-none** devices.  
2. A fixed number of simultaneous excitations is required to fire.  
3. **Synaptic delays** are the only significant temporal factors.  
4. **Inhibitory synapses** block firing absolutely.  
5. Network structure is fixed.  

### Formal Calculus
- Represent neuron firing as **logical propositions**.  
- Define **realizability** of logical expressions in terms of neural nets.  
- Develop theorems for **acyclic networks** and extend to **cyclic ones**.  
- **Equivalence theorems:** relative inhibition ≈ absolute inhibition; extinction ≈ inhibition; temporal ≈ spatial summation.  
- **Computability link:** neural nets can compute exactly what a Turing machine can.  

---

## Results
- Any logical expression (under realizability conditions) can be implemented by a neural net.  
- Proved the **theorem of realizability** for propositional expressions and cyclic nets.  
- Demonstrated equivalence of multiple biological assumptions (e.g., inhibition vs. extinction).  
- Showed nets with circles (feedback) can implement **recursive functions**.  
- Established that neural nets possess **Turing-complete computational power**, equivalent to Church’s λ-definability and Kleene’s primitive recursiveness.  

---

## Conclusions
McCulloch and Pitts introduced the **first mathematical model of neural networks**, treating the brain as a computational system implementing logic. Their framework unified **neurophysiology, symbolic logic, and computability theory**, proving that neural nets can realize **universal computation**. While biologically simplified, this work became one of the **most influential foundations** of computational neuroscience, artificial intelligence, and cognitive science.


# McCulloch & Pitts (1943): A Logical Calculus of the Ideas Immanent in Nervous Activity

## 1. Binary Neuron Assumption

**Mathematical Formulation:**

$$
V_i \in \{0,1\}
$$

- \(V_i = 1\): neuron fires.  
- \(V_i = 0\): neuron does not fire.  

**Role:** Simplifies biological neurons into logical variables representing truth values.

---

## 2. Excitation and Inhibition

A neuron fires when the weighted sum of its excitatory inputs exceeds a threshold, unless any inhibitory input is active (absolute inhibition).

**Logical Expression:**

$$
V_i = f(x_1, x_2, \dots, x_k, y_1, y_2, \dots, y_m)
$$

where \(f\) is a Boolean function combining AND, OR, NOT.  

**Role:** Neurons directly implement logical connectives.

---

## 3. Temporal Assumptions

Activity evolves in **discrete time** steps due to synaptic delay.

**Rule:**

$$
V_i(t+1) = f\big(V_1(t), V_2(t), \dots, V_n(t)\big)
$$

**Role:** Frames neural networks as discrete dynamical systems.

---

## 4. Logical Propositions and Realizability

- Any Boolean formula (using AND, OR, NOT) can be represented as a neural net.  
- **Realizability Theorem:** if a logical formula meets structural constraints, it can be realized as a neural circuit.  

**Role:** Links neural nets with propositional logic.

---

## 5. Cyclic Networks and Recursion

- Feedback loops enable representation of **recursive functions**.  
- Example: a neuron with delayed self-feedback acts as a memory element.  

**Result:** Cyclic networks have the same computational power as **Turing machines**.

---

## 6. Equivalence Theorems

- Relative inhibition ≈ absolute inhibition.  
- Extinction ≈ inhibition.  
- Temporal summation ≈ spatial summation.  

**Role:** Different biological assumptions yield equivalent mathematical behavior.

---

## 7. Computability

McCulloch & Pitts proved their networks can compute all functions that are:

- **Turing-computable**  
- **λ-definable** (Church)  
- **Primitive recursive** (Kleene)  

**Role:** Situates neural networks firmly within computability theory.

---

## 8. Statistical Aspect

- The model ignores stochasticity.  
- Neurons are treated as **deterministic logical devices**.  

**Role:** Establishes a purely logical-mathematical framework, unlike later probabilistic models.

---

# Concise Mathematical Summary

- Neurons are **binary threshold devices**:

  $$
  V_i \in \{0,1\}
  $$

- Networks update in **discrete time**:

  $$
  V_i(t+1) = f(V_1(t), \dots, V_n(t))
  $$

- Inhibition is absolute (logical NOT).  
- Networks implement any **Boolean function** (combinational logic).  
- **Feedback** enables recursion and memory → universal computation.  
- **Equivalence theorems** simplify assumptions without reducing expressiveness.  

**Conclusion:** Neural nets are **logical machines with Turing equivalence**, unifying brain-like activity with formal logic and computability.

```python
# McCulloch & Pitts (1943) Replication Lab
# A simple educational pipeline showing how threshold-based neurons implement logic functions.

import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np

# ======================================================
# 1. DATA PREPARATION
# ======================================================
# Define truth tables for AND, OR, XOR (inputs are {0,1})
X = torch.tensor([[0,0],[0,1],[1,0],[1,1]], dtype=torch.float32)
y_and = torch.tensor([[0],[0],[0],[1]], dtype=torch.float32)
y_or  = torch.tensor([[0],[1],[1],[1]], dtype=torch.float32)
y_xor = torch.tensor([[0],[1],[1],[0]], dtype=torch.float32)  # cannot be solved with 1-layer perceptron

datasets = {"AND": y_and, "OR": y_or, "XOR": y_xor}

# ======================================================
# 2. MODEL DEFINITION
# ======================================================
# McCulloch & Pitts used threshold units.
# We approximate them using a sigmoid (for smooth gradients) and apply threshold at inference.
class MCPNetwork(nn.Module):
    def __init__(self, input_dim, hidden_dim=0):
        super().__init__()
        if hidden_dim == 0:
            # Single-layer perceptron
            self.model = nn.Sequential(
                nn.Linear(input_dim, 1),
                nn.Sigmoid()
            )
        else:
            # Add hidden layer to solve XOR
            self.model = nn.Sequential(
                nn.Linear(input_dim, hidden_dim),
                nn.Sigmoid(),
                nn.Linear(hidden_dim, 1),
                nn.Sigmoid()
            )
    def forward(self, x):
        return self.model(x)

# ======================================================
# 3. TRAINING FUNCTION
# ======================================================
def train_model(X, y, hidden_dim=0, epochs=2000, lr=0.1):
    model = MCPNetwork(input_dim=2, hidden_dim=hidden_dim)
    criterion = nn.BCELoss()
    optimizer = optim.SGD(model.parameters(), lr=lr)

    losses, accs = [], []
    for epoch in range(epochs):
        optimizer.zero_grad()
        outputs = model(X)
        loss = criterion(outputs, y)
        loss.backward()
        optimizer.step()

        # Accuracy calculation
        preds = (outputs.detach() > 0.5).float()
        acc = (preds == y).float().mean().item()

        losses.append(loss.item())
        accs.append(acc)
    return model, losses, accs
```

```python
# ======================================================
# 4. VISUALIZATION UTILITIES
# ======================================================
def plot_training(losses, accs, title):
    fig, ax = plt.subplots(1,2, figsize=(10,4))
    ax[0].plot(losses); ax[0].set_title(f"{title} Loss"); ax[0].set_xlabel("Epoch"); ax[0].set_ylabel("Loss")
    ax[1].plot(accs); ax[1].set_title(f"{title} Accuracy"); ax[1].set_xlabel("Epoch"); ax[1].set_ylabel("Accuracy")
    plt.show()

def evaluate_model(model, X, y, title):
    with torch.no_grad():
        outputs = model(X)
        preds = (outputs > 0.5).float()
    print(f"Results for {title}:")
    for inp, target, pred in zip(X, y, preds):
        print(f" Input: {inp.numpy()} | Target: {int(target.item())} | Pred: {int(pred.item())}")
    print()
```

```python
# ======================================================
# 5. RUN EXPERIMENTS
# ======================================================

for name, y in datasets.items():
    hidden = 2 if name=="XOR" else 0   # XOR needs hidden layer
    model, losses, accs = train_model(X, y, hidden_dim=hidden)

    # Plot training curves
    plot_training(losses, accs, title=name)

    # Show predictions
    evaluate_model(model, X, y, title=name)
```

# Statistical Analysis of Logical Gate Experiments (McCulloch–Pitts / Perceptron Replication)

## Extracted Metrics

### OR Gate
- **Loss**: ≈ 0.6 → 0.05 by 2000 epochs.  
- **Accuracy**: ≈ 0.75 → **1.0** (perfect).  
- **Predictions**: Correctly reproduces OR truth table.  

### AND Gate
- **Loss**: ≈ 0.8 → 0.1 by 2000 epochs.  
- **Accuracy**: ≈ 0.25 → **1.0** (perfect).  
- **Predictions**: Matches AND truth table.  

### XOR Gate
- **Loss**: Plateau ≈ 0.70 → 0.693.  
- **Accuracy**: ≈ 0.50 (chance-level).  
- **Predictions**: Fails to reproduce XOR truth table.  

---

## Interpretation

1. **Linear Separability**  
   - OR and AND are **linearly separable**.  
   - Single-layer perceptron converges with monotonic loss decrease → perfect accuracy.  

2. **XOR Limitation**  
   - XOR is **not linearly separable**.  
   - Perceptron cannot implement nonlinear decision boundaries → stagnation in loss/accuracy.  

3. **Energy of Training Curves**  
   - OR/AND: smooth gradient descent convergence.  
   - XOR: plateau due to representational limitation, not optimization failure.  

---

## Implications

- Confirms McCulloch & Pitts (1943) and Rosenblatt’s perceptron findings.  
- Single-layer perceptrons can model **simple logical functions** (AND, OR, NOT).  
- Failure on XOR highlights the **historical bottleneck**, leading to the rise of **multi-layer perceptrons (MLPs)** and **backpropagation**.  

---

## Limitations

- Inputs restricted to **binary truth tables**.  
- Single-layer model only; multi-layer would solve XOR.  
- Metrics: limited to **loss/accuracy curves**, no variance or multiple run analysis.  

---

## Concise Academic Summary

The replication shows that perceptrons **solve linearly separable gates (OR, AND)** but fail on **non-linear XOR**, reproducing a key historical result in AI. This experiment illustrates both the **expressive power** and **limitations** of early neural network models, motivating the invention of deeper architectures.
